## NumPy's Arithmetic Binary Functions.

```python
import numpy as np
```

##### 1.	Add two arrays a = [1, 2, 3] and b = [4, 5, 6].

```python
a1 = np.array([1,2,3])
b1 = np.array([4,5,6])
print(a1+b1)
print(np.add(a1,b1))
```

##### 2.	Subtract array b = [4, 5, 6] from a = [10, 15, 20].

```python
a2 = np.array( [4, 5, 6])
b2 = np.array( [10, 15, 20])
print(b2-a2)
print(np.subtract(b2,a2))
```

##### 3.	Multiply a = [2, 3, 4] with b = [5, 6, 7].

```python
a3 = np.array( [2, 3, 4])
b3 = np.array(  [5, 6, 7])
print(a3*b3)
print(np.multiply(a3,b3))
```

##### 4.	Divide a = [10, 20, 30] by b = [2, 4, 5] using np.divide().

```python
a4 = np.array([10, 20, 30])
b4 = np.array([2, 4, 5])
print(a4/b4)
print(np.divide(a4,b4))
```

##### 5.	Use np.floor_divide() on a = [10, 20, 30], b = [3, 6, 7].

```python
a5 = np.array( [10, 20, 30])
b5 = np.array( [3, 6, 7])
print(a5//b5)
print(np.floor_divide(a5,b5))
```

#### 16.	Multiply arrays with different shapes using broadcasting:
##### a = [[1], [2], [3]], b = [10, 20].


```python
a16 = np.array([[1], [2], [3]])
b16 = np.array( [10, 20])
print(a16*b16)
print()
print(np.multiply(a16,b16))
print()
print(np.multiply(b16,a16))
```
