# Comparison Binary Function 

```python
import numpy as np
```

##### 1.	Compare a = [1, 2, 3] and b = [1, 0, 3] using np.equal().

```python
a1 = np.array([1,2,3])
b1 = np.array([1,0,3])
print(a1==b1)
print(np.equal(a1,b1))
```
