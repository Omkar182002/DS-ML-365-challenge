# NUMPY COPY AND VIEW

```python
import numpy as np
```

```python
array = np.arange(10)
array
```

#### 1. Create a copy of an array using .copy() and verify that changes do not affect the original array

```python
ar1 = array.copy()
ar1
```

```python
ar1[0] =999
```

```python
print(array)
print(ar1)
```

#### 2. Create a copy of an array using np.copy() and demonstrate independence from the original array.

```python
ar2 = np.copy(array)
ar2[4] = 999
print("Original array:", array)
print("Copied array  :", ar2)
```

#### 3.Create a new array using slicing ([:]) and check whether it behaves as a copy or view.

```python
ar3 = array[:]
ar3[0] =999
array
```


#### 4. Create a copy of an array using np.array() and modify it to confirm original data remains unchanged.

```python
ar4 = np.array(array).copy()
ar4[0] = 999
print("Original array:", array)
print("Copied array  :", ar4)
```

#### 5. Copy data from one array to another using np.copyto() and verify independence.

```python
ar5 =np.empty_like(array)
np.copyto(ar5,array)
ar5
```

#### 9. Convert an array to a list and back to an array to create a copy and verify independence.

#### 11. Create a view and modify it to observe changes in original

```python
new_array = np.array(range(1,11))
new_array
```

```python
b1 = new_array.view()
b1[1] = 999
print(new_array)
```
