### REAL PROBLEM-BASED ASSIGNMENTS

```python
sales = np.array([
    [120, 150, 100],   # Day 1 (Product A, B, C)
    [90,  200, 130],   # Day 2
    [80,  170, 160],   # Day 3
    [200, 220, 180],   # Day 4
    [160, 140, 190]    # Day 5
])
```

#### 1. Find total sales across all days and products.

```python
total_sales = sales.sum()
print(total_sales)
```

#### 2. Find total sales for each day.

```python
print("Total Sales per Day :")
i =1
for t_sales in sales.sum(axis=1):
    print(f"Day {i} : {t_sales}")
    i+=1
```

#### 3. Find total sales of each product (A, B, C).

```python
print("Total Sales per Product : ")
for product ,t_sales in zip(("A","B","C"),np.sum(sales,axis=0)):
    print(f" Product {product} : {t_sales}")
```

#### 4. Find which day had highest total sales.

```python
highest_total_sale = np.sum(sales,axis=1).max() 
higest_seling_day= np.sum(sales,axis=1).argmax()+1
print(f"Day {higest_seling_day} : {highest_total_sale}")
```

#### 5. Find which product sold the most overall.

```python
total_sale = np.sum(sales,axis =0).max()
high_sale_p = np.sum(sales,axis =0).argmax()
product = ["A","B","C"]
print("Highest Saling Poroduct :",product[high_sale_p])
print("Total saling :",total_sale)
```

#### 6. Find which day had minimum total sales.

```python
less_sale_day = np.sum(sales,axis=1).argmin()
print("Less saling Day:",less_sale_day+1)
print("Total sale on that Day : ",np.sum(sales,axis=1).min())
```

### 7. Find mean sales for each day.

```python
day =1
for avg in np.mean(sales,axis=1):
    print(f"Day {day} : {avg}")
    day+=1
```

```python
day =1
for avg in np.mean(sales,axis=1):
    print(f"Day {day} : {np.round(avg,2)}")
    day+=1
```

```python
day =1
for avg in np.mean(sales,axis=1):
    print(f"Day {day} : {avg:.2f}")
    day+=1
```

#### 8. Find highest single sale value in dataset.

```python
high_single_sale = np.max(sales)
print("Highest Saling Value :",high_single_sale)
```

#### 9, Find difference between highest and lowest sales.

```python
range_value = np.max(sales)-np.min(sales)
print(range_value)
```

```python
print(np.ptp(sales))
```

#### 10. Find standard deviation of each product to see which is most stable.

```python
product =["A","B","C"]
print("standard deviation of each product ")
for p, st in zip(product,np.std(sales,axis=0)):
    print(f"Product {p} : {st:.2f}")
```
