# NumPy Missing Values Operations

### FUNCTION 1: np.isnan()
Returns boolean array where NaN values are True.

```python
import numpy as np

# Employee salaries with missing values
salaries = np.array([45000, np.nan, 52000, 38000, np.nan, 61000, 49000, np.nan, 55000, 47000])
print("Salaries:", salaries)
```

#### Q1. Find all indices where salary is missing (NaN) using np.isnan().



```python
np.isnan(salaries)
```

```python
np.where(np.isnan(salaries))[0]
```

#### Q2. Count how many missing values are in the array.



```python
np.where(np.isnan(salaries))[0].size
```

#### Q3. Extract all non-missing salaries into a new array.



```python
array2 = np.extract(~np.isnan(salaries),salaries)
array2
```

```python
arr3 = salaries[~np.isnan(salaries)]
arr3
```

#### Q4. Replace all missing salaries with the median of non-missing values.



```python
arr4 = salaries.copy()
arr4
```

```python
median_data = np.nanmedian(arr4)
median_data
```

```python
arr4[np.isnan(arr4)] = median_data
arr4
```

#### Q5. Check if any missing values exist in the array using np.isnan()



```python
np.isnan(salaries).any()
```

```python
np.isnan(arr4).any()
```

## FUNCTION 2: np.nan_to_num()
Replaces NaN with 0 and inf with large numbers

```python
readings = np.array([25.5, np.nan, 30.2, np.inf, 28.7, np.nan, -np.inf, 26.3, 29.1, np.nan])
print("Readings:", readings)
```

#### Q6. Replace all NaN values with 0 using np.nan_to_num().

```python
readings2 = readings.copy()
readings2
```

```python
readings2 = np.nan_to_num(readings2,nan=0)
readings2.astype(int)
```

#### Q7. Replace NaN with 100 and inf with 999 using custom parameters.

```python
readings3 = readings.copy()
readings3
```

```python
readings3 = np.nan_to_num(readings3,nan= 100, posinf= 999, neginf= -999)
readings3
```

#### Q8. Replace only NaN (keep inf as is) by setting posinf=None, neginf=None.

```python
readings4 = readings.copy()
readings4
```

```python
readings4 = np.nan_to_num(readings4,nan= 100, posinf= None, neginf= None)
readings4
```

```python
np.set_printoptions(suppress=True, precision=2)
```

```python
readings4
```

### REAL-WORLD SCENARIO 1: Customer Purchase Data

```python
np.random.seed(42)

# 100 customers × 12 months
purchases = np.random.randint(0, 1000, size=(100, 12)).astype(float)

# Randomly insert 15% missing values
mask = np.random.random(purchases.shape) < 0.15
purchases[mask] = np.nan

print("Purchases shape:", purchases.shape)
print("First 5 customers, first 6 months:\n", purchases[:5, :6])
print(f"Missing values: {np.isnan(purchases).sum()} out of {purchases.size}")
```

### Q41. Calculate average monthly purchase per customer ignoring NaN.

```python
np.nanmean(purchases,axis=1)[:5]
```

### Q42. Replace missing values with each customer's mean (use broadcasting).

```python
purchases2 = purchases.copy()
purchases2[:5,:6]
```

```python
cust_mean = np.nanmean(purchases2,axis=1)
purchases2[np.isnan(purchases2)] = cust_mean
purchases2[:5,:5]
```

```python
indexs = np.where(np.isnan(purchases2))
purchases2[indexs] = cust_mean[indexs[1]]
purchases2[:6,:5]
```

### Q43. Find customers with more than 3 missing months.

```python
miss_count = np.sum(np.isnan(purchases),axis=1)
customers = np.where(miss_count > 3)[0]
print(customers)
```

### Q44. Calculate total yearly purchase for each customer (replace NaN with 0 first).

```python
purchases3 = purchases.copy()
purchases3
```

```python
purchases3[np.isnan(purchases3)] = 0
```

```python
purchases3
```

### Q45. Find which month has highest total sales (after filling NaN with median).

```python
purchases4 = purchases.copy()
med_data = np.nanmedian(purchases4,axis=1)
ind_data = np.where(np.isnan(purchases4))
purchases4[ind_data] = med_data[ind_data[0]]
```

```python
purchases4
```

```python
total_sale_per_month = np.sum(purchases4,axis=0)
print("Month has highest sales : ",np.argmax(total_sale_per_month)+1)
print("Sale : ",np.max(total_sale_per_month))
```
