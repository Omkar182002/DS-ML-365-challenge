# 2. Creating array using linspace()

```python
aa1 = np.linspace(1,10)
aa1.size
```

```python
aa2 = np.linspace(1,10,10)
print(aa2.size)
aa = np.linspace(1,10,num = 10)
print(aa.size)
```

```python
aa3 = np.linspace(1,100,num = 10, dtype=int)
aa4 = np.linspace(1,100,  num= 10, endpoint = False,  dtype = int , retstep = True)
print(aa3)
print(aa4)
```

#### 0 से 10 तक 5 values बनाओ और साथ में step size भी print करो (retstep=True)।


```python
ab1, step  = np.linspace(0,10,5, retstep= True)
print(ab1)
print("Step : ",step)
```

#### 10 से 1 तक reverse order में linspace से array बनाओ।

```python
ab2 = np.linspace(10,1,10).astype(int)
ab2
```

#### एक array बनाओ जिसमें 0 से 1 तक के बीच 1000 points हों।

```python
ab3= np.linspace(0,1,1000)
print(ab3.size)
```

#### linspace() से बने array को np.sin() में input दो।

```python
x = np.linspace(0, np.pi, 5)
y = np.sin(x)
print(y)
```

#### Two array eleemnt - wise sum 

```python
a1 = np.linspace(0, 100, 10).astype(int)
a2 = np.linspace(10, 0, 10).astype(int)
sum_array = a1 + a2
print(sum_array)

```

####  Gradient-style 2D array (shape 10×1):

```python
a = np.linspace(0, 1, 10).reshape(10, 1)
print(a)
```

### 14. Create a matrix of shape (4, 4) where the diagonal elements form an arithmetic sequence [2, 4, 6, 8] using broadcasting

```python
e11 = np.zeros((4, 4), dtype=int)
e11[np.arange(4), np.arange(4)] = [2, 4, 6, 8]
print(e11)
```

#### 15. Create a matrix where the main diagonal is 1, upper diagonal is 2, and lower diagonal is 3.



```python
ee1 = np.zeros((5, 5), dtype=int)
ee1[np.arange(5), np.arange(5)] = 1             # main
ee1[np.arange(4), np.arange(1, 5)] = 2           # upper
ee1[np.arange(1, 5), np.arange(4)] = 3           # lower
print(ee1)
```
