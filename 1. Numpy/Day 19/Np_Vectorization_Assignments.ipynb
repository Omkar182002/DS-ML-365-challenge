# NumPy Vectorization – Real-World Problems


```python
import numpy as np
```

### PROBLEM SET 1: E-Commerce Pricing


```python

# 50,000 products with [original_price, discount_percent, tax_rate]
products = np.random.rand(50000, 3)
products[:, 0] = products[:, 0] * 1000 + 10  # Price: $10-$1010
products[:, 1] = np.random.randint(0, 50, 50000)  # Discount: 0-50%
products[:, 2] = np.random.uniform(0.05, 0.15, 50000)  # Tax: 5-15%

print("Products shape:", products.shape)
print("First 5 products [price, discount%, tax%]:\n", products[:5])
```

```python
print(products[1,0])
```

#### Q1. Calculate discounted price: price * (1 - discount/100) for all 50,000 products at once.

```python
discount_price = products[:,0] * (1-products[:,1]/100)
print("Fist 5 product discount price : ")
for dp in np.nditer(discount_price[:5]):
    print(np.round(dp,2))
```

#### Q2. Calculate final price after tax: discounted price * (1 + tax rate).

```python
tax = products[:, 2]
final_price = discount_price * (1 + tax)
final_price
```

#### Q3. Apply flash sale: products under $50 get additional 20% off. Update discounted price.

```python
flash_mask = products[:,0] < 50
discount_price[flash_mask] *= 0.8
print(discount_price[:5])
```

#### Q4. Mark products as "High Value" if final price > $500, "Medium" if $100-500, "Low" if < $100.

```python
ch = np.array(["High Value","mediaum value","Low value"])

condi = [ final_price > 500,
    (final_price >= 100) & (final_price <= 500),
    final_price < 100 ]
category = np.select(condi,ch,default="Unk")
category
```

#### Q5. Calculate total revenue if all products sell 1 unit each.

```python
total_revenue = np.sum(final_price)
print(total_revenue)
```

### PROBLEM SET 2: Weather Station Analysis

```python
# 1000 weather stations × 365 days × 4 measurements
# Measurements: [temp_min, temp_max, humidity, rainfall_mm]
weather = np.random.rand(1000, 365, 4)
weather[:, :, 0] = weather[:, :, 0] * 30 + (-5)  # Min temp: -5 to 25°C
weather[:, :, 1] = weather[:, :, 1] * 25 + 10   # Max temp: 10-35°C
weather[:, :, 2] = weather[:, :, 2] * 70 + 20   # Humidity: 20-90%
weather[:, :, 3] = weather[:, :, 3] * 50        # Rainfall: 0-50mm

print("Weather shape:", weather.shape)
print("Station 0, Day 0 [min_temp, max_temp, humidity, rain]:", weather[0, 0])
```

#### Q6. Calculate daily average temperature: (min_temp + max_temp) / 2 for all stations, all days.

```python
avg_temp = (weather[...,0] + weather[...,1]) /2
np.round(avg_temp,2)
```

#### Q7. Flag heatwave days: max_temp > 30 AND min_temp > 20 for any station.

###### When both day and night are hot for multiple days, it is called a heatwave.

```python
max_temp = weather[...,0]
min_temp = weather[...,1]
heatwave = (max_temp > 30) & (min_temp > 20)

print(heatwave.shape)
print(heatwave[0][:10])
```

#### Q8. Calculate total annual rainfall per station (sum across days).

###### Rainfall means how much rain falls in a place, measured in millimeters

```python
rainfall = weather[...,3]
annual_rainfall = np.sum(rainfall,axis=1)
```

```python
print(annual_rainfall[:5])    # first 5 stations
```

```python
rainfall[:5]
```

#### Q9. Find stations with average humidity > 75%.

```python
humidity = weather[:, :, 2]
avg_humidity = np.mean(humidity, axis=1)
high_humidity_stations = np.where(avg_humidity > 75)[0]

print("Stations with high humidity:", high_humidity_stations)
print("Count:", len(high_humidity_stations))
```

### PROBLEM SET 3: Student Grades Processing

```python
# 10,000 students × 6 subjects × 3 exams
# Exams: [quiz1, quiz2, final]
grades = np.random.randint(40, 100, size=(10000, 6, 3))
weights = np.array([0.2, 0.2, 0.6])  # Quiz1 20%, Quiz2 20%, Final 60%

print("Grades shape:", grades.shape)
print("First student's grades (6 subjects × 3 exams):\n", grades[0])
```

#### Q11. Calculate weighted final grade for each student, each subject using vectorization.

```python
weighted = grades * weights
final_scores = np.sum(weighted, axis=2)
print(final_scores[0]) # student 1st
```

#### Q12. Calculate overall grade (average across all subjects) for each student.

```python
overall_grade = np.mean(final_scores,axis=1)
overall_grade
```

#### Q13. Identify students who failed any subject (grade < 60).

```python
failed_st = np.where(final_scores<60)[0]
print("First 10 Failed studensts : ",failed_st[:10])
```

```python
failed_st = np.unique(np.where(final_scores < 60)[0])
print("First 10 Failed studensts : ",failed_st[:10])
```

#### Q14. Calculate class average for each subject (mean across all students).

```python
subject_average = np.mean(final_scores, axis=0)

print(subject_average.shape)  
print(subject_average)
```
