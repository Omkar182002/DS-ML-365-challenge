# 4. Data Cleaning

#### Replace all NaN values with 0 in an array.

```python
array_d = np.array([100,110,99,10,np.nan, 20,120,np.nan,111,np.nan])
array_d
```

```python
np.where(np.isnan(array_d),0,array_d)
```

```python
array_d
```

```python
array_d = np.where(np.isnan(array_d),0,array_d)
```

```python
array_d
```

#### Replace all values > 100 with 100 (cap values).

```python
array_d[array_d>100] = 100
```

```python
array_d
```

#### Replace all values < 0 with mean of array.

```python
np.random.seed(42)
new_arr = np.random.randint(-10,10,10)
```

```python
new_arr
```

```python
new_arr[new_arr<0] = np.mean(new_arr[new_arr>0])
```

```python
new_arr
```

#### Remove negative values using filtering.

```python
np.random.seed(42)
new_arr2 = np.random.randint(-10,10,10)
new_arr2
```

```python
result = new_arr2[new_arr2>=0]
result
```
