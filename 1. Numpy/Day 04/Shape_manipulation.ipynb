##  Shape Manipulation in NumPy

```python
import numpy as np
```

##### 1. Create a 1D array of numbers from 1 to 12 and reshape it to (3, 4).

```python
a1 = np.arange(1,12+1)
b1 = a1.reshape(3,4)
print(a1)
print("_"*20)
print(b1)
```

##### 2. Reshape a 1D array of 16 elements into shape (4, 4).

```python
a2 = np.arange(16)
print(a2)
b2 = np.reshape(a2,(4,4))
print("_"*20)
print(b2)
```

##### 3 . Convert a 2D array [[1, 2], [3, 4]] into a 1D array using ravel().

```python
a3 = np.array( [[1, 2], [3, 4]] )
print(a3)
b3= a3.ravel()
print(b3)
```

##### 4. Flatten a 3D array into 1D using flatten().

```python
a4 = np.zeros((2,2,2))
print(a4)
print("_"*20)
print(a4.flatten())
print("_"*20)
print(a4)
```


##### 8 . Create a 2D array and add a new axis using expand_dims() at axis=0.

```python
a8 = np.linspace(1,12,12).reshape(3,4)
print(a8)
print("_"*20)
b8 = np.expand_dims(a8,0)
print(b8)
print(b8.shape)

print("_"*20)
c8 = np.expand_dims(a8,1)
print(c8)
print(c8.shape)
print("_"*20)
d8 = np.expand_dims(a8,2)
# d8 = a8.expand_dims()  # AttributeError: 'numpy.ndarray' object has no attribute 'expand_dims'
print(d8)
print(d8.shape)
```
