# NumPy Sorting Functions –  Simple Assignments

## 1: np.sort()

```python
import numpy as np

marks = np.array([85, 42, 97, 63, 78, 51, 89, 34, 76, 91])
print("Original marks:", marks)
```

#### Q1. Sort the marks in ascending order using np.sort().

```python
np.sort(marks)
```

#### Q2. Sort the marks in descending order (use np.sort() then reverse).



```python
np.sort(marks)[::-1]
```

#### Q3. Sort the marks but keep the original array unchanged. Prove it by printing both.



```python
sorted_mark = np.sort(marks)
print("marks        : ",marks)
print("Sorted marks : ",sorted_mark)
```

#### Q4. Sort only the first 5 marks (indices 0-4) using slicing + np.sort().



```python
# help(np.sort)
np.sort(marks[:5])
```

#### Q5. Find the 3 highest marks using np.sort() and slicing.



```python
h1,h2,h3 = np.sort(marks)[::-1][:3]
print("Highest mark :",h1)
print("Second Highest m :",h2)
print("Third Highest m :",h3)
```

## 2: ndarray.sort()
Sorts the array in-place (modifies original).



```python
prices = np.array([299, 150, 450, 125, 350, 200, 500, 75, 425, 180])
print("Original prices:", prices)
```

#### Q6. Sort the prices in ascending order using in-place sort.

```python
price2 = prices.copy()
price2.sort()
print(price2)
```

#### Q7. After sorting, print the original array to confirm it changed

```python
prices3 = prices.copy()
prices3.sort()
```

```python
prices3
```

#### Q8. Sort the prices in descending order using in-place sort.

```python
prices4 = prices.copy()
prices4.sort()
prices4[::-1]
```

#### Q9. Sort only the last 6 elements in-place (leave first 4 unchanged).

```python
prices9 = prices.copy()
prices9[4:] = np.sort(prices9[4:])

print(f"After sorting : {prices9}")
print(f"First 4 elements unchanged: {prices9[:4]}")
print(f"After sorting last 6 elements: {prices9[4:]}")
```

#### Q10. Sort the array in-place then find the median (middle value).

```python
prices6 = prices.copy()
prices6.sort()
median_value = np.median(prices6)
print(prices6)
print('median value : ',median_value)
```

## 3: np.argsort()
Returns indices that would sort the array.

```python
heights = np.array([165, 180, 155, 172, 168, 190, 158, 175, 182, 160])
print("Heights:", heights)
print("Student IDs (index):", np.arange(10))
```

#### Q11. Find the indices that would sort heights from shortest to tallest.

```python
sorted_index = np.argsort(heights)
print('sorted array index :',sorted_index)
```

#### Q12. Use those indices to get the sorted heights array.

```python
heights[sorted_index]
```

#### Q13. Find which student (index) is the tallest using np.argsort().

```python
st_id = np.argsort(heights)[-1]
print("Tallest Student id : ",st_id)
print("Height : ",heights[st_id])
```

#### Q14. Find which student (index) is the shortest using np.argsort().

```python
st_id = np.argsort(heights)[0]
print("Smallest Student id : ",st_id)
print("Height : ",heights[st_id])
```

#### Q15. Get the top 3 tallest students' indices using np.argsort().

```python
st1,st2,st3 = np.argsort(heights)[::-1][:3]
print("Top 3 highest height student ids :")
print("St id :",st1)
print("St id :",st2)
print("St id :",st3)
```

### 4: np.partition()
Partial sort - places kth smallest element in correct position.

```python
salaries = np.array([45000, 32000, 68000, 28000, 52000, 41000, 75000, 38000, 61000, 49000])
print("Salaries:", salaries)
```

#### Q16. Partition the array so that the 3rd smallest (index 2) is in correct position.

```python
partition_array = np.partition(salaries,2)
partition_array
```

```python
print(partition_array[2])
```

#### Q17. Get the 4 smallest salaries using np.partition().

```python
sorted_salaries = np.partition(salaries,4)
print(sorted_salaries[:4])
```

#### Q18. Get the 3 largest salaries using np.partition() with negative index.

```python
sorted_array = np.partition(salaries,4)
```

```python
print(sorted_array[-3])
```

#### Q19. Compare speed: find top 5 salaries using np.sort() vs np.partition() on large array.

```python
import time
start = time.time()
sorted_all = np.sort(salaries)
top_5_sort = sorted_all[-5:]
time_sort = time.time() - start

start = time.time()
partitioned = np.partition(salaries, -5)
top_5_partition = partitioned[-5:]
time_partition = time.time() - start
```

```python
print(f"Array size: {len(salaries):,}")
print(f"\nnp.sort() time: {time_sort:.4f}s")
print(f"Top 5 via sort: {np.sort(top_5_sort)}")
print(f"\nnp.partition() time: {time_partition:.4f}s")
print(f"Top 5 via partition: {np.sort(top_5_partition)}")
print("-" * 60)
```

### 5: np.argpartition()
Returns indices of partial sort.

```python
scores = np.array([88, 45, 92, 67, 79, 56, 94, 38, 81, 73, 96, 62, 85, 49, 91])
print("Scores:", scores)
print("Student indices:", np.arange(15))
```

#### Q21. Find indices of the 3 smallest scores using np.argpartition().

```python
score_index = np.argpartition(scores,3)
print(score_index[:3])
```

#### Q22. Find indices of the 5 largest scores using np.argpartition().

```python
score_max = np.argpartition(scores,3)[-5:]
print(score_max)
```

#### Q23. Use the indices to extract the actual 5 largest scores.

```python
print("Top 5 scores : ",scores[score_max])
```

#### Q24. Find which student has the 2nd highest score using np.argpartition().

```python
print("Second Higest scoes : ",scores[score_max][-2])
```

#### Q25. Find which student has the lowest score using np.argpartition().

```python
print(scores[score_index[0]])
```
