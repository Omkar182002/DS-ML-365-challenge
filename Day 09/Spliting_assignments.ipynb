# 1. np.split()

### 1D ARRAY

```python
A = np.array([10, 20, 30, 40, 50, 60])
print("Array  A : ",A)
```

#### BASIC

#### 1. Split A into 2 equal parts.

```python
data1, data2 = np.split(A,2)
print("Dataset 1 : ",data1)
print("Dataset 2 : ",data2)
```

#### 2. Split A into 3 equal parts.

```python
data1 , data2, data3 = np.split(A,3)
print("Dataset 1 : ",data1)
print("Dataset 2 : ",data2)
print("Dataset 3 : ",data3)
```

#### 3.Split A at index [2].

```python
data1, data2 = np.split(A,[2])
print("Dataset 1 : ",data1)
print("Dataset 2 : ",data2)
```

#### 4. Split A at indices [2, 4].

```python
data1 , data2 , data3 = np.split(A,[2,4])
print("Dataset 1 : ",data1)
print("Dataset 2 : ",data2)
print("Dataset 3 : ",data3)
```

#### 5. Split A into 6 parts.

```python
np.split(A,6)
```

#### ADVANCED 

#### 1. Split A into 3 parts and compare sums.

```python
data1 , data2, data3 = np.split(A,3)
print("Dataset 1  Sum : ",data1.sum())
print("Dataset 2  sum : ",data2.sum())
print("Dataset 3  sum  :",data3.sum())
```

#### 2. Split A and identify segment containing 40.

```python
data1, data2 = np.split(A,2)
if np.isin(data1,40).any():
    print("Contain in 1st part")
    print("Dataset 1 : ",data1)
elif np.isin(data2,40).any() :
    print("Contain in 2nd part")
    print("Dataset 2 : ",data2)
else :
    print("Not coianing in both parts")
```

#### 3. Split A into halves and compare max values.

```python
data1, data2 = np.split(A,2)
max_data1 = np.max(data1)
max_data2 = np.max(data2)
if max_data1 > max_data1:
    print("Max value[data1]:",max_data1)
elif max_data1 < max_data2:
    print("Max value[data2] : ",max_data2)
else :
    print("Both Are common :",data1)
```

#### 4.Split A and calculate length of each segment.

```python
data1,data2 = np.split(A,[4])
print("Length of data1 :",data1.size)
print("Length of data2 : ",data2.size)
```

#### 5.Split A and reverse each segment.

```python
data1 , data2 = np.split(A,2)
print("Dataset Reverse :",data1[::-1])
print("Dataset Reverse : ",data2[::-1])
```
