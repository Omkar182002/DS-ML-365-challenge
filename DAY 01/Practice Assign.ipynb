#### 10. Create Fibonacci-like first 20 values (array form).

```python
fib = np.empty(20, dtype=int)
fib[0] = 0
fib[1] = 1
for i in range(2, 20):
    fib[i] = fib[i-1] + fib[i-2]

print(fib)
```
