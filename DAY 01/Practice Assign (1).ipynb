#### 26. Create array sorted in descending order without using sort().

```python
np.random.seed(42)
a26 = np.random.randint(1,100,15)
n = a26.size
print(a26)
for i in range(n-1):
    k=i
    for j in range(i+1,n):
        if a26[k] >a26[j]:
            k = j
    temp = a26[k]
    a26[k]=a26[i]
    a26[i]= temp

print(a26)
    
```

### 27.Create array of prime numbers up to 200.

```python
prime_numbers = []
for num in range(2,201):
    Is_prime = True 
    for i in range(2,int(num//2)+1):
        if num %i ==0:
            Is_prime = False
            break
    if Is_prime :
        prime_numbers.append(num)
a27 = np.array(prime_numbers)
a27
```
