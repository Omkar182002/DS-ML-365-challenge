### chained condition (&,|)

##### In below we wnat only greater than 10 and less than 50 

```python
x1 = np.random.randint(1,100,50)
print(x1)
```

```python
bv = (x1>10 )&( x1<50)
print(bv)
print(x1[bv])
```

```python
x1.any(x1>10&x1<55)
```

#### Operator precedence (गलत तरीके से & use kiya gaya hai)
#####  Galat likha:  
######  1. x1.any(x1 > 10 & x1 < 55)   2. Yeh Python is tarah samajhta hai: x1 > (10 & x1) < 55   Jo completely galat hai — isiliye error deta hai.
##### Sahi Tarika: Use brackets properly, and call .any() on the result, not with arguments.
###### 1. (x1 > 10) & (x1 < 55)
###### 2. ((x1 > 10) & (x1 < 55)).any()
###### Galat : x1.any(condition)	   |||||||||               sahi : condition.any()  and also shahi : np.any(array)

######  Jab hum NumPy ke np.sqrt() function ka use karte hain kisi array ke upar, to ye function array ke har element ka square root nikalta hai. Lekin mathematical rule ke mutabiq, kisi bhi negative number ka square root real number system mein exist nahi karta. NumPy is situation ko handle karne ke liye nan (Not a Number) return karta hai. Iska matlab hota hai ki result mathematically invalid hai.
###### Agar aapke array (jaise u1) mein kuch negative values hain aur aap unka square root nikalne ki koshish karte hain, to NumPy aapko ek RuntimeWarning deta hai:

```python
print(np.sqrt(u1).astype(int))
```

###### Jab aap NumPy mein np.sqrt(u1).astype(int) likhte ho, to aap har element ka square root nikalne ki koshish kar rahe ho aur usse integer mein convert kar rahe ho.

###### Lekin agar u1 array mein kuch negative numbers hote hain (jaise -1, -5), to unka square root real numbers mein possible nahi hota. NumPy in values ka square root NaN (Not a Number) return karta hai.

###### Jab aap astype(int) lagate ho, to NumPy NaN ko integer mein convert nahi kar paata. Is wajah se aapko warning milti hai:

##### Final Note :
###### Negative numbers ka sqrt real number mein valid nahi hota.
###### NaN ko int mein convert karna galti hai, warning deta hai.
