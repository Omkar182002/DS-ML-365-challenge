# Business Problem

#### Company: E-commerce Logistics Company
- The company introduced a new AI-based routing system to reduce delivery time.

- Earlier, deliveries were managed using the old routing system. After introducing the new system, the company wants to check whether deliveries are actually faster.

#### Business Question:

- Has the new AI routing system significantly reduced the average delivery time compared to the old routing system?

# 

#### Calculate Population Parameters

```python
p_mean = popu_df["delivery_time_hours"].mean()
p_std = popu_df["delivery_time_hours"].std(ddof=0)
```

```python
p_mean
```

```python
p_std
```

# 

```python
sample_mean = sample_df["delivery_time_hours"].mean()
```

```python
sample_mean
```

```python
n = len(sample_df)
```

```python
n
```

# 

##### Z-TEST 

```python
z = (sample_mean - p_mean)/(p_std / np.sqrt(n))
```

```python
z
```

```python
alpha = 0.05
critical_z = stats.norm.ppf(alpha)      # Left-tailed test
```

```python
critical_z
```

```python
if z  < critical_z:
    print("\nDecision: Reject H₀")
    print("Conclusion: The new routing system significantly reduced delivery time.")
else:
    print("\nDecision: Fail to Reject H₀")
    print("Conclusion: There is insufficient evidence that the new routing system reduced delivery time.")
```
