# Hypothesis Testing Using Z-Test

```python
import numpy as np
import scipy.stats as stats
```

# 

## Example 1: Student Study Time
### Problem :  
- A school knows that students score an average of 70 marks in a math test. The population standard deviation is 8 marks.
- After introducing a new teaching method, a sample of 25 students has an average score of 73 marks.
- The school wants to know if the new teaching method increased the average score

```python
# given : 
p_mean_score = 70 
p_std = 8 
sample_size = 25
sample_mean = 73
```

##### Hypotheses
- H0 = 70   - Null Hypothesis
- H1 > 70   - Alternative Hypothesis   
Test : Right-tailed Z-test

```python
Z_test = (sample_mean - p_mean_score) / (p_std/ np.sqrt(sample_size))
```

```python
Z_test
```

 - The critical Z-value is the cutoff point that determines whether you reject or fail to reject the null hypothesis.
- alpha = significance level : represents the probability of making a Type I Error

```python
alpha = 0.05 
critical_z = stats.norm.ppf(1-alpha)
```

```python
critical_z
```

```python
print(f"Calculated Z-statistic : {Z_test:.3f}")
print(f"Critical Z-value       : {critical_z:.3f}")
```

```python
if Z_test > critical_z:
    print("Decision: Reject Null Hypothesis H0")
    print("Conclusion: The new teaching method significantly increased the average student score.")
else:
    print("Decision: Fail to Reject the Null Hypothesis (H0)")
    print("Conclusion: There is insufficient evidence that the new teaching method increased the average student score.")
```

## Example 2: Bakery Bread Weight
### Problem
- A bakery makes loaves of bread with an average weight of 500 g. The population standard deviation is 20 g.
- A sample of 36 loaves has an average weight of 495 g.
- The bakery wants to know if the average weight has decreased.

```python
# given
p_mean2 = 500
p_std2 = 20
s_mean = 495
s_size = 36
```

#### Hypotheses
- H0 = 500  - Null Hypothesis
- H1 < 500   - Alternative Hypothesis   
Test : Left-tailed Z-test

```python
z_test2 = (s_mean - p_mean2 )/ (p_std2 / np.sqrt(s_size)) 
```

```python
z_test2
```

```python
alpha = 0.05
critical_z2  = stats.norm.ppf(alpha)
```

```python
critical_z2
```

```python
print(f"Calculated Z-statistic : {z_test2:.3f}")
print(f"Critical Z-value       : {critical_z2:.3f}")
```

```python
if z_test2 < critical_z2:
    print("Decision: Reject Null Hypothesis H0")
    print("Conclusion: The average weight of the bread loaves has significantly decreased.")
else:
    print("Decision: Fail to Reject the Null Hypothesis (H0)")
    print("Conclusion: There is insufficient evidence to conclude that the average weight has decreased.")
```

## Example 3: Coffee Cup Fill
### Problem : 
- A coffee shop claims each cup contains 250 ml of coffee. The population standard deviation is 10 ml.
- A random sample of 64 cups has an average fill of 247 ml.
- The manager wants to determine whether the average amount of coffee served is different from 250 ml.

```python
# given :
p_mean3 = 250
p_std3 = 10
s_size3 = 64
s_mean3 = 247
```

#### Hypotheses
H0 = 250 - Null Hypothesis   
H1 !=  250 - Alternative Hypothesis  
Test : Two - tailed Z-test  

```python
z_test3 = (s_mean3 - p_mean3 )/ (p_std3 / np.sqrt(s_size3)) 
```

```python
z_test3
```

```python
alpha = 0.05
left_critical = stats.norm.ppf(alpha/2)
right_critical = stats.norm.ppf(1 - alpha/2)
```

```python
if z_test3 < left_critical or z_test3 > right_critical:
    print("Decision: Reject Null Hypothesis H0")
    print("Conclusion: The average amount of coffee served is significantly different from 250 ml.")
else:
    print("Decision: Fail to Reject the Null Hypothesis (H0)")
    print("Conclusion: There is insufficient evidence to conclude that the average amount of coffee served is different from 250 ml.")
```
