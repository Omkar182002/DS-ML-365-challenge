## Discrete Probability Distribution
- A discrete probability distribution is a type of probability distribution that is used only for countable (discrete) values.

### Example 1: Dice Roll 🎲

- Possible outcomes: 1, 2, 3, 4, 5, 6
- These are countable values, so this is a discrete probability distribution.

```python
#Possible values
x = [1, 2, 3, 4, 5, 6]
n= len(x)
# Probability of each value
probability = [1/n] * n
distribution = pd.DataFrame({"Random Variable (X)": x,"Probability": probability})
print(distribution)
```

#### Problem: When the number of possible outcomes is large, a probability distribution table becomes difficult to use.
#### So, we use mathematical functions : 
- Probability Mass Function → for discrete (countable) values
- Probability Density Function → for continuous (measurable) values

### Probability Mass Function (PMF)

#### Example 1: Coin Toss
- A fair coin is tossed 10 times. What is the probability of getting exactly 6 Heads?

```python
n = 10      # coin is tossed 10 times
```

```python
p = 1/2     # Probability of Head
```

```python
x = 6  #  getting exactly 6 Heads     
```

```python
print('Number of time coin tossed : ',n)
print('Probability of Head : ',p)
print('Randome Variable(Head) :',x)
```

```python
probability = stats.binom.pmf(x, n, p)
```

```python
print(probability)
```

#### Binomial PMF
Use when:
- Fixed number of trials
- Only two outcomes: Success or Failure

### Probability Density Function (PDF)

##### Example 1: Student Heights (Normal PDF)
- A school records the heights of 100 students. Find the probability density for a student whose height is 170 cm.

```python
heights = pd.DataFrame({
    "Height": [165,168,170,172,169,
               171,173,167,170,169]
})
```

```python
heights
```

```python
mean_height = heights["Height"].mean()
```

```python
mean_height
```

```python
std_height = heights["Height"].std()
```

```python
std_height
```

```python
stu_H_density = stats.norm.pdf(x=170,loc=mean_height,scale=std_height)
```

```python
print(stu_H_density)
```

### Cumulative Distribution Function (CDF )
- gives the probability that a random variable is less than or equal to a given value.

##### Example 1: Delivery Time (Continuous)
- Find the probability that a delivery takes 30 minutes or less.

```python
delivery = pd.DataFrame({
    "Time": [28, 32, 30, 29, 31, 27, 33, 30, 29, 31]})
```

```python
delivery
```

```python
mean_time =  delivery["Time"].mean()
std_time =  delivery["Time"].std()
```

```python
prob = stats.norm.cdf(x=30,loc=mean_time,scale=std_time)
```

```python
print("Mean:", mean_time)
print("Standard Deviation:", std_time)
print("Probability:", prob)
```
