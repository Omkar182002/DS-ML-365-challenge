# Bernoulli Distribution Analysis

##### A Bernoulli distribution is a probability distribution that describes a single experiment with only two possible outcomes:
- Success
- Failure

#### Gender 

#### Q1. What is the probability that a randomly selected customer is Male?

```python
p = (df["Gender"] == "Male").mean()
```

```python
print(stats.bernoulli.pmf(1, p))
```

###### Another Way 

```python
p1 = (df['Gender_Binary']==1).mean()
```

```python
print(stats.bernoulli.pmf(1, p1))
```
