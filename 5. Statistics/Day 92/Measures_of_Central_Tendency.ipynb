# Measures of Central Tendency

```python
import pandas as pd
import numpy as np
import statistics 
```

#### Retail Sales Analysis

```python
# load dataset 
data = pd.read_excel('retail_sales_dummy_500.xlsx')
```

```python
data.head()
```

```python
# Explore the Dataset
data.describe().T
```

```python
# average sales amount
data['Sales'].mean()
```

```python
statistics.mean(data['Sales'])
```

```python
# typical customer spending
data['Sales'].median()
```

```python
statistics.median(data['Sales'])
```

```python
print(data["Sales"].mean())
print(data["Sales"].median())
```

##### Mean > Median	Positively skewed [Right-skewed]

```python
# Most sold category
data['Category'].mode()
```

```python
statistics.mode(data['Category'])
```

```python
# Most common payment method
data['Payment_Method'].mode()
```

```python
statistics.mode(data['Payment_Method'])
```

```python
# Average selling price considering quantity sold.
# waighted mean
np.average(data["Unit_Price"],weights=data["Quantity"])
```

```python
data["Unit_Price"].mean()
```

```python
data[["Unit_Price","Quantity"]].head(10)
```

```python
# Some observations are more important than others
# so that the reason  weighted mean  is this better than mean 
```

# 

##### Trimmed Mean

```python
from scipy.stats import trim_mean
```

```python
trim_mean(data["Sales"],0.10)
```

```python
trim_mean(data["Sales"],0.20)
```

```python
# removing a fixed percentage of the smallest and largest observations and calculate mean 
# useful when outlier are presant in dataset
```

##### Category-wise Mean

```python
data.groupby("Category")["Sales"].mean()
```

```python
# we can understand which category have better sales or performance 
```

##### Category-wise Median

```python
data.groupby("Category")["Sales"].median()
```

```python
# usefull to check performance when data contain outliers
```
