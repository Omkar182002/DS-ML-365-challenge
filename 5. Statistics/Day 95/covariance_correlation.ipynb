# 

## Covariance
- indicates how two numerical variables change together.

```python
data['Math'].cov(data['Attendance'])
```

```python
data['Science'].cov(data['Attendance'])
```

```python
data['English'].cov(data['Attendance'])
```

- Math and Attendance: Positive relationship .    -  attendance increases, Math scores generally increase as well.
- Science and Attendance: Positive relationship.  -  attendance increases, Science scores generally increase as well.
- English and Attendance: Negative relationship.  -  attendance increases, English scores tend to decrease slightly.

#### - Which relationship is stronger ?

- Science & Attendance (Covariance = 7.12) > Math & Attendance (Covariance = 2.09)


#### - Is  Science & Attendance  is stronger relationship than others

```python
data['Science'].cov(data['Attendance'])
```

```python
(data['Science']*2).cov(data['Attendance']*2)
```

- Using covariance, we cannot determine whether a relationship is strong or weak. We can only determine the direction of the relationship—whether it is positive, negative, or there is no linear relationship.

# 

```python
x1 = data['Math'].head(15)
x2 = data['Math'].head(15)*2

y1 = data['Attendance'].head(15)
y2 = data['Attendance'].head(15)*2
```

```python
fig, ax = plt.subplots(1,2,figsize=(10, 5))
ax[0].scatter(x=x1,y=y1)
ax[0].set_title(f'Covariance: {x1.cov(y1):.2f}')
ax[0].set_xlabel('Math')
ax[0].set_ylabel('Attendance')


ax[1].scatter(x=x2,y=y2)
ax[1].set_title(f'Covariance: {x2.cov(y2):.2f}')
ax[1].set_xlabel('Math')
ax[1].set_ylabel('Attendance')
plt.show()
```

# 

```python
fig, ax = plt.subplots(1,2,figsize=(10, 5))
ax[0].scatter(x=x1,y=x1)
ax[0].set_title(f'Covariance: {x1.cov(y1):.2f}')
ax[0].set_xlabel('Math')
ax[0].set_ylabel('Math')


ax[1].scatter(x=x2,y=x2)
ax[1].set_title(f'Covariance: {x2.cov(y2):.2f}')
ax[1].set_xlabel('Math')
ax[1].set_ylabel('Math')
plt.show()
```

##### Both charts look the same, but the covariance values are different. This is because covariance depends on the scale of the data, not just the relationship.

## Correlation
-  tells us both the direction and the strength of the relationship between two numerical variables.
-  We use correlation to understand how strongly two variables are related.

```python
data.corr(numeric_only=True)
```

 -  +1.00 → Perfect Positive
-   +0.70 to +0.99 → Strong Positive
-   +0.30 to +0.69 → Moderate Positive
-   +0.01 to +0.29 → Weak Positive
-  0.00 → No Relationship
-  -0.01 to -0.29 → Weak Negative
-  -0.30 to -0.69 → Moderate Negative
-  -0.70 to -1.00 → Strong Negative
-  -1.00 → Perfect Negative

```python
data['Math'].corr(data['Attendance'])
```

- Weak Positive Relationship

```python
fig, ax = plt.subplots(2,2,figsize=(7, 7))
ax[0,0].scatter(x=x1,y=y1)
ax[0,0].set_title(f'Corrlation: {x1.corr(y1):.2f}')
ax[0,0].set_ylabel('Attendance')


ax[0,1].scatter(x=x2,y=y2)
ax[0,1].set_title(f'Correlation: {x2.corr(y2):.2f}')

ax[1,0].scatter(x=x1,y=y1)
ax[1,0].set_title(f'Covariance: {x1.cov(y1):.2f}')
ax[1,0].set_xlabel('Math')
ax[1,0].set_ylabel('Attendance')


ax[1,1].scatter(x=x2,y=y2)
ax[1,1].set_title(f'Covariance: {x2.cov(y2):.2f}')
ax[1,1].set_xlabel('Math')

plt.tight_layout()
plt.show()
```

###### Both charts show the same relationship. The correlation remains the same because it is scale-independent, while the covariance changes because it is scale-dependent

### So Correlation is preferred because it is not affected by the scale of the data and always ranges from -1 to +1.
