#### 3. Using that same p, what's the probability that out of 100 visitors, more than 20 actually purchase? (Use binomial formula or Normal approximation since n is large.)

###### What is the chance that at least 20 out of 100 visitors will make a purchase?

```python
binom.sf(20,100,purchase_rate)
```
