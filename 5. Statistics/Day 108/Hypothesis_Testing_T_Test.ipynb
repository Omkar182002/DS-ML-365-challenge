# One-Sample T-Test: Business Case Studies

```python
import numpy as np
import pandas as pd 
import scipy.stats as stats
```

### Case 1 – FastTrack Logistics

#### Business Problem

- FastTrack Logistics has always taken 4.5 days on average to deliver orders. Recently, the company introduced an AI-powered route optimization system to help drivers choose faster routes and reduce delivery time.

- The management wants to know if the new system is actually improving delivery speed before using it across all regions.

#### Business Question
- Has the average delivery time become less than 4.5 days after implementing the AI routing system?

```python
data_df = pd.read_csv('delivery_sample_after_ai_routing.csv')
```

```python
data_df.info()
```

```python
data_df.head()
```

```python
data_df.duplicated().sum()
```

```python
sample1 = data_df['Delivery_Time_Days']
```

# 

#### Hypotheses
- H0 = 4.5 days - Null Hypothesis
- H1 < 4.5 days - Alternative Hypothesis   
Test : Left-Tailed T-test

```python
mean1 = 4.5
```

```python
sample_mean1 = sample1.mean()
```

```python
sample_std1 = sample1.std()
sample_size1 = len(sample1)
```

```python
print('Sample Mean : ',sample_mean1)
print('Sample Std  : ', sample_std1)
print('Sample Sizxe : ', sample_size1)
```

```python
T1 = (sample_mean1 - mean1)/(sample_std1/np.sqrt(sample_size1))
```

```python
T1
```

```python
alpha = 0.05
dof = sample_size1 - 1
t_critical1 = stats.t.ppf(alpha, dof) 
```

```python
t_critical1
```

```python
if T1 < t_critical1:
    print("Decision: REJECT H0")
    print("Conclusion: Sufficient evidence that mean delivery time < 4.5 days")
else:
    print("Decision: FAIL TO REJECT H0")
    print("Conclusion: Not enough evidence that mean delivery time < 4.5 days")
```

### Case 3 – Banking (Meridian National Bank)

#### Business Problem

- Meridian National Bank's service level agreement (SLA) states that loan applications should be processed in 6 days on average. Recently, the internal audit team received complaints that processing times may have changed due to increased workload or a new digital workflow.

- The bank wants to check whether the current average processing time is still 6 days or if it has changed.

#### Business Question

- Is the average loan processing time different from the SLA benchmark of 6 days?

```python
data_df3 = pd.read_csv('loan_processing_time_audit_sample.csv')
```

```python
data_df3.head()
```

```python
data_df3.info()
```

```python
data_df3.duplicated().sum()
```

```python
sample3 = data_df3['Processing_Time_Days']
```

```python
sample3.head()
```

#### Hypotheses
- H0 = 6 days - Null Hypothesis
- H1 != 6 days - Alternative Hypothesis   
Test : Two-Tailed T-test

```python
mean3 = 6
```

```python
sample_mean3 = sample3.mean()
sample_std3 = sample3.std()
sample_size3 = len(sample3)
```

```python
print('Sample Mean : ', sample_mean3)
print('Sample Std  : ', sample_std3)
print('Sample Size : ', sample_size3)
```

```python
T3_test = (sample_mean3 - mean3) / (sample_std3 / np.sqrt(sample_size3))
```

```python
T3_test
```

```python
alpha = 0.05
dof3 = sample_size3 -1
left_critical = stats.t.ppf(alpha/2, dof3)
right_critical = stats.t.ppf(1 - alpha/2, dof3)
```

```python
left_critical
```

```python
right_critical
```

```python
if T3_test < left_critical or T3_test > right_critical:
    print("Reject H0")
    print("Average loan processing time is different from 6 days.")
else:
    print("Fail to Reject H0")
    print("Average loan processing time is not significantly different from 6 days.")
```
