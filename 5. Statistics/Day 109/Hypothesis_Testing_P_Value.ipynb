## Hypothesis Testing : 

```python
import scipy.stats as stats
import numpy as np
```

## P-Value Using Z-Test

### Problem 1: Cricket Bating Performance Analysis (One-Tailed)
 A cricket academy claims that its advanced batting training program helps players achieve an average batting score of 45 runs per innings. To verify this claim, a sports analyst randomly selects 36 players who completed the training program and records their batting scores over recent matches. The sample has an average batting score of 47.8 runs per innings. Based on historical performance data, the population standard deviation is known to be 7 runs. Using a significance level of 0.05, determine whether there is sufficient statistical evidence to conclude that the average batting score of players in the training program is higher than the academy's claimed average.

```python
# Given Data :
mean1 = 45             # population mean
s_mean1 = 47.8         # sample Mean
p_std1 = 7             # population standard deviation
n = 36                 # sample size
alpha = 0.05           # significance level
```

###### Hypothesis:
- Null Hypothesis H0 : mean1 = 45
- Alternative Hypothesis H1 : mean1 > 45

```python
# Calculate Z statistic
z1 = (s_mean1 - mean1) / (p_std1 / np.sqrt(n))
```

```python
z1
```

```python
# Calculate p-value for one-tailed test
p_value1 = 1 - stats.norm.cdf(z1)
```

```python
p_value1
```

```python
# Also we can calculate p-value for one-tailed test
stats.norm.sf(z1)
```

```python
print("Z - statistic:", round(z1, 2))
print("P-value:", round(p_value1, 4))
```

```python
# Decision
if p_value1 < alpha:
    print("Reject Null Hypothesis")
    print("There is sufficient evidence that the average batting score is higher than 45 runs.")
else:
    print("Fail to Reject Null Hypothesis")
    print("There is not enough evidence that the average batting score is higher than 45 runs.")
```

### Problem 2: Fast Bowling Speed Analysis   (Two-Tailed)
A cricket academy claims that the average bowling speed of its professional fast bowlers is 140 km/h. A sports analyst wants to verify whether the actual average bowling speed is different from this claimed value. A random sample of 45 fast bowlers is selected, and their average bowling speed is found to be 137 km/h. The population standard deviation of bowling speeds is known to be 12 km/h. Using a significance level of 0.05, determine whether there is sufficient statistical evidence to conclude that the average bowling speed differs from the academy's claim.

```python
# Given Data :
mean2 = 140             # population mean
s_mean2 = 137         # sample Mean
p_std2 = 12            # population standard deviation
n2 = 45                # sample size
alpha = 0.05           # significance level
```

###### Hypothesis:
Null Hypothesis H0 : mean2 = 140    
Alternative Hypothesis H1 : mean2 != 140   

```python
z2 = (s_mean2 - mean2) / (p_std2 / np.sqrt(n2))
```

```python
z2
```

```python
# Two-tailed p-value
p_value2 = 2 * (stats.norm.sf(abs(z2)))
```

```python
print("Z-statistic:", round(z2, 2))
print("P-value:", round(p_value2, 4))
```

```python
if p_value2 < alpha:
    print("Reject H0: Average bowling speed is significantly different from 140 km/h")
else:
    print("Fail to Reject H0: No significant evidence that average speed differs from 140 km/h")
```

## P-Value Using T-Test 

### Problem 3: Cricket Fitness Program Evaluation (One-Tailed)
A cricket training center claims that its fitness program helps players complete a running drill in an average time of 12 minutes. To evaluate the program, a coach selects 18 players who completed the training and records their drill completion times. The sample has an average completion time of 11.2 minutes with a sample standard deviation of 1.5 minutes. Using a significance level of 0.05, determine whether there is sufficient statistical evidence to conclude that the fitness program reduces the average drill completion time.

```python
# given data : 
mean3 = 12      
s_mean3 = 11.2         # sample Mean
s_std3 = 1.5          # sample standard deviation
n3 = 18               # sample size
alpha = 0.05
```

###### Hypothesis:
- Null Hypothesis H0 : mean3 = 12
- Alternative Hypothesis H1 : mean3 < 12

```python
# Calculate t-statistic
T1 = (s_mean3 - mean3) / (s_std3 / np.sqrt(n3))
```

```python
T1
```

```python
# Degrees of freedom
df1 = n3 - 1
```

```python
# One-tailed p-value
p_value3 = stats.t.cdf(T1, df1)
```

```python
p_value3
```

```python
print("t-statistic:", round(T1, 2))
print("p-value:", round(p_value3, 4))
print("Degrees of freedom:", df1)
```

```python
if p_value3 < alpha:
    print("Reject H0")
    print("Fitness program significantly reduces drill completion time.")
else:
    print("Fail to Reject H0")
    print("Not enough evidence that the program reduces completion time.")
```

### Problem 4: Batting Practice Duration Analysis (Two-Tailed)
A cricket analytics company claims that professional players spend an average of 35 minutes practicing batting techniques during each training session. To verify this claim, a researcher randomly selects 20 players and records their practice duration. The sample has an average practice time of 37 minutes with a sample standard deviation of 6 minutes. Using a significance level of 0.05, determine whether there is sufficient statistical evidence to conclude that the actual average practice duration differs from the company's claimed value.

```python
# given data  
mean4 = 35      
s_mean4 = 37        # sample Mean
s_std4 = 6          # sample standard deviation
n4 = 20             # sample size
alpha = 0.05
```

###### Hypothesis:
- Null Hypothesis H0 : mean4 = 35
- Alternative Hypothesis H1 : mean4 != 35

```python
T2 = (s_mean4 - mean4) / (s_std4 / np.sqrt(n4))
```

```python
T2
```

```python
df2 = n4 -1
p_value4 = 2 * stats.t.sf(T2, df2) # also  p_value4 = 2 * stats.t.sf(abs(T2), df2) when t statistics is negative
```

```python
print("t-statistic:", round(T2, 2))
print("p-value:", round(p_value4, 4))
print("Degrees of freedom:", df2)
```

```python
if p_value4 < alpha:
    print("Reject H0")
    print("Average practice duration differs from 35 minutes.")
else:
    print("Fail to Reject H0")
    print("There is no significant evidence that the average practice duration differs from 35 minutes.")
```
