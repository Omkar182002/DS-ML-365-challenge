# Uniform Distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

#### Generate 100,000 observations between 0 and 10.

```python
uniform_dist = stats.uniform.rvs(loc=0,scale=10,size=100000,random_state=42)
```

```python
uniform_dist[:10]
```

#### Convert into DataFrame

```python
df = pd.DataFrame({'Uniform_Data': uniform_dist})
```

```python
df.head()
```

#### Summary Statistics

```python
df.describe().T
```

#### Density Plot

```python
sns.kdeplot(data=df)
plt.show()
```

```python
sns.kdeplot(data=df,x='Uniform_Data')
plt.show()
```

```python
pd.Series(uniform_dist).plot(kind="density",figsize=(8,5))
plt.title("Uniform Distribution Density")
plt.show()
```

#### Probability that a value is less than or equal to 6.

```python
stats.uniform.cdf(x=6,loc=0,scale=10)
```

```python
stats.uniform.cdf(x=1,loc=0,scale=10)
```

```python
stats.uniform.cdf(x=10,loc=0,scale=10)
```

# 

#### Percent Point Function (PPF)

##### Find the value corresponding to 40% probability.

```python
stats.uniform.ppf(q=0.4,loc=0,scale=10)
```

###### 0.4 means 40% of observations lie below 4.

# 

#### Verify CDF and PPF are Inverse

```python
x = 7
prob = stats.uniform.cdf(x=x,loc=0,scale=10)
value = stats.uniform.ppf(q=prob,loc=0,scale=10)

print(prob)
print(value)
```

# 

#### Probability Between Two Values

```python
probability = (stats.uniform.cdf(8, loc=0, scale=10)-stats.uniform.cdf(3, loc=0, scale=10))
```

```python
probability
```

###### 50% observations lie between 3 and 8.
