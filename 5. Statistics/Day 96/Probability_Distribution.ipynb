###### We cannot predict whether the result will be Head or Tail, but we know the result will be either Head or Tail. 
###### Therefore, tossing a coin is a random experiment

- Experiment: Toss a coin.
- Possible outcomes: Head, Tail.
- Exact outcome: Unknown until the coin is tossed.
- That's why it's called a random experiment.

#### Probability Distribution 

```python
students 
```

```python
# Count Pass students
npass_student = len(students[students['Result'] == 'Pass'])
```

```python
npass_student
```

```python
# Count Fail students
nfail_student = len(students[students['Result'] == 'Fail'])
```

```python
nfail_student
```

```python
total_students
```

```python
# Calculate probabilities
pass_probability = npass_student / total_students
fail_probability = nfail_student / total_students
```

```python
pass_probability
```

```python
fail_probability
```

```python
# Random variable values
pass_x = 1
fail_x = 0
```

###### Create probability distribution table

```python
distribution = pd.DataFrame({
    'Random Variable (X)': [pass_x, fail_x],
    'Outcome': ['Pass', 'Fail'],
    'Probability': [pass_probability, fail_probability]
})
```

```python
distribution
```

###### A probability distribution is a table that shows each possible value of a random variable and its probability.

# 

##### Visulization

```python
import seaborn as sns
```

```python
sns.barplot(data=distribution, x='Outcome', y='Probability')
```

```python
import matplotlib.pyplot as plt 
```

```python
sns.barplot(data=distribution, x='Outcome', y='Probability')
plt.title('Probability Distribution')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.show()
```

# 

##### Probability Distribution using Value Counts

```python
rolls = [random.randint(1,6) for _ in range(1000)]
```

```python
rolls[:10]
```

```python
df3 = pd.DataFrame({'Dice':rolls})
```

```python
df3.head()
```

```python
probability_distribution = (df3['Dice'].value_counts(normalize=True).sort_index())
```

```python
probability_distribution
```

##### Bar Chart

```python
probability_distribution.plot(kind='bar')
plt.title("Probability Distribution")
plt.xlabel("Dice Number")
plt.ylabel("Probability")
plt.show()
```

```python
probability_distribution = probability_distribution.reset_index()
```

```python
probability_distribution 
```

```python
sns.barplot(data= probability_distribution, x='Dice', y='proportion')
plt.title("Probability Distribution")
plt.xlabel("Dice Number")
plt.ylabel("Probability")
plt.show()
```

#### Verify Probability Distribution

###### A valid probability distribution must satisfy two conditions:
- Each probability must be between 0 and 1.
- The sum of all probabilities must be 1.

```python
probability_distribution.sum()
```
