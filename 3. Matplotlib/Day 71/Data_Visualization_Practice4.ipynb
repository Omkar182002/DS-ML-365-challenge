# Case Study: Supermarket Sales Analysis

### Business Problem

A supermarket manager wants to analyze sales performance and answer:

- Which product category sells the most?
- Which day generated the highest sales?
- What is the distribution of bill amounts?
- Are more items sold leading to higher revenue?

## Dataset

```python
data = {
    'Date': [
        '2025-06-01','2025-06-01',
        '2025-06-02','2025-06-02',
        '2025-06-03','2025-06-03',
        '2025-06-04','2025-06-04',
        '2025-06-05','2025-06-05'
    ],

    'Category': [
        'Groceries','Beverages',
        'Groceries','Snacks',
        'Beverages','Snacks',
        'Groceries','Beverages',
        'Snacks','Groceries'
    ],

    'Items_Sold': [
        25,15,
        30,20,
        18,22,
        35,20,
        25,40
    ],

    'Sales': [
        2500,1200,
        3000,1800,
        1500,2000,
        3500,1600,
        2300,4000
    ]
}
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

# 

### Task 1: Sales by Category
- Which category generates the highest revenue?

```python
category_sales = (df.groupby('Category')['Sales'].sum().reset_index())
```

```python
plt.figure(figsize=(8,5))
plt.bar(category_sales['Category'],category_sales['Sales'])
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Revenue')
for i, v in enumerate(category_sales['Sales']):
    plt.annotate(v, xy=(i,v))
plt.show()
```

# 

### Task 2: Daily Sales Trend
- How does revenue change day by day?

```python
daily_sales = (df.groupby('Date')['Sales'].sum().reset_index())
```

```python
plt.figure(figsize=(8,5))
plt.plot(daily_sales['Date'],daily_sales['Sales'],marker='o')
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()
```

# 

### Task 3: Revenue Distribution
- How are sales amounts distributed?

```python
plt.figure(figsize=(8,5))
plt.hist(df['Sales'],bins=5)
plt.title('Sales Distribution')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.show()
```

### Task 4: Items Sold vs Revenue
- Do more items sold generate higher revenue?

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Items_Sold'],df['Sales'])
plt.title('Items Sold vs Revenue')
plt.xlabel('Items Sold')
plt.ylabel('Revenue')
plt.show()
```

# 

### Task 5: Category Revenue Share
- What percentage of revenue comes from each category?

```python
plt.figure(figsize=(6,6))
plt.pie(category_sales['Sales'],labels=category_sales['Category'],autopct='%1.1f%%')
plt.title('Revenue Share by Category')
plt.show()
```
