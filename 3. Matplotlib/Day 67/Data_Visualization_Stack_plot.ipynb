# Data Visualization

## Stack Plots

```python
import matplotlib.pyplot as plt
```

### Revenue by Product Category

```python
months = ['Jan','Feb','Mar','Apr','May']

electronics = [100,120,110,160,180]
fashion = [80,90,110,100,120]
home = [50,55,60,70,80]
```

# 

```python
plt.figure(figsize=(10,5))
plt.stackplot(months,electronics)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.stackplot(months,fashion)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.stackplot(months,home)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.stackplot(months,electronics,fashion,home)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.stackplot(months,electronics,fashion,home,labels=['Electronics','Fashion','Home'])
plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.stackplot(months,electronics,fashion,home,labels=['Electronics','Fashion','Home'])
plt.title('Revenue by Category')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.legend(loc='upper left')
plt.grid(alpha=0.3)
```

## Website Traffic Sources
- Is total traffic growing?
- Which source contributes most?
- Is paid traffic increasing?

```python
months = ['Jan','Feb','Mar','Apr','May']

organic = [1000,1200,1300,1500,1700]
paid = [500,600,700,900,1000]
social = [300,350,400,500,600]
```

```python
plt.figure(figsize=(10,5))
plt.stackplot(months,organic,paid,social,labels=['Organic','Paid','Social'])
plt.title('Website Traffic Sources')
plt.xlabel('Month')
plt.ylabel('Visitors')
plt.legend()
plt.show()
```

## Employee Count by Department

```python
years = [2020,2021,2022,2023,2024]

it = [20,25,30,40,50]
sales = [15,18,20,25,30]
hr = [5,6,7,8,10]
```

```python
plt.figure(figsize=(10,5))
plt.stackplot(years,it,sales,hr,labels=['IT','Sales','HR'])
plt.title('Employee Count by Department')
plt.xlabel('Year')
plt.ylabel('Employees')
plt.legend()
plt.show()
```

```python
years = ['2020','2021','2022','2023','2024']
plt.figure(figsize=(10,5))
plt.stackplot(years,it,sales,hr,labels=['IT','Sales','HR'])
plt.title('Employee Count by Department')
plt.xlabel('Year')
plt.ylabel('Employees')
plt.legend()
plt.show()
```
