## Histogram 

#### 1 — E-commerce Order Value Distribution

###### The Sales Director wants to understand how customer order values are distributed to identify:
- ###### Most common order ranges  
- ###### High-value customers  
- ###### Potential outliers  
- ###### Pricing and promotion opportunities  

```python
order_values = [
120,145,160,180,190,200,210,220,230,240,
250,260,270,280,290,300,310,320,330,340,
350,360,370,380,390,400,410,420,430,440,
450,460,470,480,490,500,510,520,530,540,
550,560,570,580,590,600,610,620,630,640,
650,660,670,680,690,700,710,720,730,740,
750,760,770,780,790,800,810,820,830,840,
850,860,870,880,890,900,920,940,960,980,
1000,1050,1100,1150,1200,1300,1400,1500,1700,2000
]
```

```python
order_values[:10]
```

```python
plt.figure(figsize=(10,6))
plt.hist(order_values,5 )
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(order_values,4, edgecolor='black' )
plt.title('Distribution of Customer Order Values')
plt.xlabel('Order Value ($)')
plt.ylabel('Number of Orders')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(order_values,bins=10,color='steelblue',edgecolor='black')

plt.title('Distribution of Customer Order Values')
plt.xlabel('Order Value ($)')
plt.ylabel('Number of Orders')

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(order_values,bins=[0,200,400,600,800,1000,1200,1400,1600,1800,2000],color='steelblue',edgecolor='black')

plt.title('Distribution of Customer Order Values')
plt.xlabel('Order Value ($)')
plt.ylabel('Number of Orders')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks([0,200,400,600,800,1000,1200,1400,1600,1800,2000])

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
binrange = [0,200,400,600,800,1000,1200,1400,1600,1800,2000]
plt.hist(order_values,bins=binrange,color='steelblue',edgecolor='black')

plt.title('Distribution of Customer Order Values')
plt.xlabel('Order Value ($)')
plt.ylabel('Number of Orders')

bin_labels = [f'${binrange[i]}-${binrange[i+1]}' for i in range(len(binrange)-1)]

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(binrange[1:],bin_labels, rotation = 60)

plt.tight_layout()
plt.show()
```
