# Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

## Box plots

### 1: Employee Salary Analysis

###### HR wants to understand salary distribution and identify unusually high or low salaries.

```python
df = pd.DataFrame({
    'Salary':[35000,38000,40000,42000,45000,
              47000,50000,52000,55000,120000]
})
```

```python
df
```

```python
plt.figure(figsize=(10,8))
plt.boxplot(df['Salary'])
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(df['Salary'])
plt.title('Employee Salary Distribution')
plt.ylabel('Salary')
plt.grid(True, alpha=0.3)
plt.show()
```

### 2: Customer Spending by Membership Type

###### An e-commerce company wants to compare spending behavior of Premium vs Regular customers.

```python
premium = [500,550,600,650,700,750,800]
regular = [100,150,200,250,300,350,400]
data = [premium, regular]
```

```python
data
```

```python
plt.figure(figsize=(7,4))
plt.boxplot(data,labels=['Premium','Regular'])
plt.show()
```

```python
plt.figure(figsize=(7,4))
plt.boxplot(data,labels=['Premium','Regular'])
plt.title('Customer Spending by Membership Type')
plt.ylabel('Amount Spent')
plt.grid(True, alpha=0.3)
plt.show()
```

### 3: Delivery Time Analysis

###### A logistics company tracks delivery times.

```python
delivery_time = [
    20,22,24,25,26,27,28,30,32,
    35,38,40,42,45,120
]
```

```python
delivery_time
```

```python
plt.figure(figsize=(10,8))
plt.boxplot(delivery_time)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.boxplot(delivery_time, notch=True)
plt.show()
```

```python
plt.boxplot(delivery_time, vert=False)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.boxplot(delivery_time, widths=0.5)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.boxplot(delivery_time, patch_artist=True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.boxplot(delivery_time, showmeans=True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.boxplot(delivery_time, showmeans=True , whis=0.5)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.boxplot(delivery_time,labels=['Delivery Time'],showmeans=True,notch=True,patch_artist=True,widths=0.5)
plt.title('Delivery Time Distribution Analysis')
plt.ylabel('Minutes')
plt.grid(True, alpha=0.3)
plt.show()
```

### 4: Regional Sales Comparison

###### Management wants to compare sales distributions across regions.
###### -  Create a box plot with 4 categories.

```python
north = [100,120,130,140,150,160,170]
south = [80,90,100,110,120,130,140]
east = [150,180,200,220,250,270,300]
west = [90,100,105,110,115,120,125]
```

```python
plt.figure(figsize=(10,8))
plt.boxplot([north,south,east,west],labels=['North','South','East','West'])
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.boxplot( [north, south, east, west],labels=['North','South','East','West'])
plt.title('Regional Sales Comparison')
plt.ylabel('Sales')
plt.grid(True, alpha=0.3)
plt.show()
```
