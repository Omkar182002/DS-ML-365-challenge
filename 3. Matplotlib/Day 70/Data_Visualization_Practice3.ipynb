### Case Study: Student Performance Analysis

### Business Problem
A school wants to analyze student performance and answer:
- Which student scored the highest?
- What is the average score?
- How are scores distributed?
- Is attendance related to marks?

# Dataset

```python
data = {
    'Student': [
        'Amit','Rahul','Priya','Neha','Rohan',
        'Anjali','Vikas','Sneha','Karan','Pooja'
    ],

    'Marks': [
        85,72,95,88,67,
        91,78,84,69,93
    ],

    'Attendance': [
        92,85,98,94,80,
        97,88,90,82,99
    ]
}
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data)
```

```python
df.info()
```

```python
df.head()
```

#### Task 1: Student Marks Comparison
- Who scored the highest marks?

```python
plt.figure(figsize=(8,5))
plt.bar(df['Student'],df['Marks'])
plt.title('Marks by Student')
plt.xlabel('Student')
plt.ylabel('Marks')

plt.show()
```

# 

# 

#### Task 2: Marks Distribution
- How are marks distributed?

```python
plt.figure(figsize=(8,5))
plt.hist(df['Marks'],bins=5)
plt.title('Marks Distribution')
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.show()
```

# 

#### Task 3: Attendance Comparison
- Which students attend classes regularly?

```python
plt.figure(figsize=(8,5))
plt.bar(df['Student'],df['Attendance'])
plt.title('Attendance by Student')
plt.xlabel('Student')
plt.ylabel('Attendance (%)')
plt.show()
```

# 

#### Task 4: Attendance vs Marks
- Does attendance affect marks?

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Attendance'],df['Marks'])
plt.title('Attendance vs Marks')
plt.xlabel('Attendance (%)')
plt.ylabel('Marks')
plt.show()
```

# 

#### Task 5: Marks Spread
- Are there any unusual scores?

```python
plt.figure(figsize=(6,5))
plt.boxplot(df['Marks'])
plt.title('Marks Distribution')
plt.show()
```
