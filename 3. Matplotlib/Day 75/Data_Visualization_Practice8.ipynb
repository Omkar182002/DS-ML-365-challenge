# Case Study: Coffee Shop Sales Analysis

### Business Problem
A coffee shop owner wants to answer:
- Which drink generates the highest revenue?
- Which day has the highest sales?
- How are bill amounts distributed?
- Does selling more cups increase revenue?
- What percentage of revenue comes from each drink?

#### Dataset

```python
data = {
    'Date': [
        '2025-06-01','2025-06-01',
        '2025-06-02','2025-06-02',
        '2025-06-03','2025-06-03',
        '2025-06-04','2025-06-04',
        '2025-06-05','2025-06-05'
    ],

    'Drink': [
        'Coffee','Tea',
        'Coffee','Sandwich',
        'Tea','Sandwich',
        'Coffee','Tea',
        'Sandwich','Coffee'
    ],

    'Cups_Sold': [
        80,50,
        90,40,
        60,45,
        100,55,
        50,110
    ],

    'Revenue': [
        12000,5000,
        13500,8000,
        6000,9000,
        15000,5500,
        10000,16500
    ]
}
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

# 

### Task 1: Revenue by Drink

```python
# data preparation
drink_revenue = (df.groupby('Drink')['Revenue'].sum().reset_index())
```

```python
drink_revenue
```

```python
plt.figure(figsize=(8,5))
plt.bar(drink_revenue['Drink'],drink_revenue['Revenue'])
plt.title('Revenue by Drink')
plt.xlabel('Drink')
plt.ylabel('Revenue')
plt.show()
```

# 

### Task 2: Daily Revenue Trend

```python
# data preparation
daily_revenue = (df.groupby('Date')['Revenue'].sum().reset_index())
```

```python
plt.figure(figsize=(8,5))
plt.plot(daily_revenue['Date'],daily_revenue['Revenue'],marker='o')
plt.title('Daily Revenue')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.grid(True)
plt.xticks(daily_revenue['Date'])
plt.show()
```

# 

### Task 3: Revenue Distribution
- How are bill amounts distributed?

```python
plt.figure(figsize=(8,5))
plt.hist(df['Revenue'],bins=5)
plt.title('Revenue Distribution')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.show()
```

# 

# 

### Task 4: Cups Sold vs Revenue
- Do more cups sold generate more revenue?

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Cups_Sold'],df['Revenue'])
plt.title('Cups Sold vs Revenue')
plt.xlabel('Cups Sold')
plt.ylabel('Revenue')
plt.show()
```

# 

### Task 5: Revenue Share by Drink

```python
plt.figure(figsize=(6,6))
plt.pie(drink_revenue['Revenue'],labels=drink_revenue['Drink'],autopct='%1.1f%%')
plt.title('Revenue Share by Drink')
plt.show()
```
