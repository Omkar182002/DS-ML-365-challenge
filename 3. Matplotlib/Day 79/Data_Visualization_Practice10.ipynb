# Case Study: Bookstore Sales Analysis 📚

### Business Problem
A bookstore owner wants to answer:
- Which book category generates the highest revenue?
- Which month had the highest sales?
- How are sales amounts distributed?
- Does selling more books increase revenue?
- What percentage of revenue comes from each category?

#### Dataset

```python
data = {
    'Month': [
        'Jan','Jan',
        'Feb','Feb',
        'Mar','Mar',
        'Apr','Apr',
        'May','May'
    ],

    'Category': [
        'Fiction','Academic',
        'Fiction','Comics',
        'Academic','Comics',
        'Fiction','Academic',
        'Comics','Fiction'
    ],

    'Books_Sold': [
        120,80,
        140,60,
        100,70,
        160,110,
        90,180
    ],

    'Revenue': [
        36000,24000,
        42000,18000,
        30000,21000,
        48000,33000,
        27000,54000
    ]
}
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.info()
```

# 

### Task 1: Revenue by Category

```python
# Data Preparation
category_revenue = (df.groupby('Category')['Revenue'].sum().reset_index())
```

```python
category_revenue
```

```python
plt.figure(figsize=(8,5))
plt.bar(category_revenue['Category'],category_revenue['Revenue'])
plt.title('Revenue by Category')
plt.xlabel('Book Category')
plt.ylabel('Revenue')
plt.show()
```

# 

### Task 2: Monthly Revenue Trend

```python
monthly_revenue = (df.groupby('Month')['Revenue'].sum().reset_index())
```

```python
monthly_revenue
```

```python
plt.figure(figsize=(8,5))
plt.plot(monthly_revenue['Month'],monthly_revenue['Revenue'],marker='o')
plt.title('Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()
```

### Task 3: Revenue Distribution
- How are sales amounts distributed?

```python
plt.figure(figsize=(8,5))
plt.hist(df['Revenue'],bins=5)
plt.title('Revenue Distribution')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.show()
```

### Task 4: Books Sold vs Revenue
- Do more books sold generate higher revenue?

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Books_Sold'],df['Revenue'])
plt.title('Books Sold vs Revenue')
plt.xlabel('Books Sold')
plt.ylabel('Revenue')
plt.show()
```

### Task 5: Revenue Share by Category

```python
plt.figure(figsize=(6,6))
plt.pie(category_revenue['Revenue'],labels=category_revenue['Category'],autopct='%1.1f%%')
plt.title('Revenue Share by Category')
plt.show()
```

# 

## Complete Dashboard

```python

# Create canvas (2 rows, 2 columns)

fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(14, 8))

# Chart 1 : Revenue by Category
ax[0, 0].bar(category_revenue['Category'],category_revenue['Revenue'])
ax[0, 0].set_title('Revenue by Category',fontsize=12,fontweight='bold')
ax[0, 0].set_xlabel('Category')
ax[0, 0].set_ylabel('Revenue')


# Chart 2 : Monthly Revenue Trend
ax[0, 1].plot(monthly_revenue['Month'],monthly_revenue['Revenue'],marker='o',linewidth=2)
ax[0, 1].set_title('Monthly Revenue Trend',fontsize=12,fontweight='bold')
ax[0, 1].set_xlabel('Month')
ax[0, 1].set_ylabel('Revenue')
ax[0, 1].grid(True)


# Chart 3 : Revenue Distribution
ax[1, 0].hist(df['Revenue'],bins=5)
ax[1, 0].set_title('Revenue Distribution',fontsize=12,fontweight='bold')
ax[1, 0].set_xlabel('Revenue')
ax[1, 0].set_ylabel('Frequency')


# Chart 4 : Books Sold vs Revenue
ax[1, 1].scatter(df['Books_Sold'],df['Revenue'])
ax[1, 1].set_title('Books Sold vs Revenue',fontsize=12,fontweight='bold')
ax[1, 1].set_xlabel('Books Sold')
ax[1, 1].set_ylabel('Revenue')
ax[1, 1].grid(True)


# Main Dashboard Heading
fig.suptitle('Bookstore Sales Dashboard',fontsize=18,fontweight='bold')

plt.tight_layout()
plt.show()
```
