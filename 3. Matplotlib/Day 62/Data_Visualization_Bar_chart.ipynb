### Bar chart

### 1 — E-commerce Sales Analysis
The business team wants to identify the highest revenue-generating product categories.

```python
df = pd.DataFrame({
    "Category": [
        "Electronics","Electronics","Electronics",
        "Furniture","Furniture",
        "Clothing","Clothing","Clothing",
        "Groceries","Groceries"
    ],
    "Revenue": [
        120000, 150000, 130000,
        90000, 110000,
        70000, 85000, 95000,
        60000, 65000
    ]
})
```

```python
df
```

```python
plt.figure(figsize=(10,5))
plt.bar(df['Category'], df['Revenue'], color='blue')
plt.xlabel('Category')
plt.title("Total Revenue by Category")
plt.ylabel('Revenue')
plt.grid(True, alpha=0.3)
plt.show()
```

```python

revenue_by_category = (
    df.groupby("Category", as_index=False)["Revenue"]
      .sum()
      .sort_values(by="Revenue", ascending=False)
)

plt.figure(figsize=(10,5))
plt.bar(revenue_by_category['Category'],revenue_by_category['Revenue'], color='blue')
plt.xlabel('Category')
plt.title("Total Revenue by Category")
plt.ylabel('Revenue')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.bar(revenue_by_category['Category'],revenue_by_category['Revenue'], color='blue')
plt.xlabel('Category')
plt.title("Total Revenue by Category")
plt.ylabel('Revenue')
plt.grid(True, alpha=0.3)
for i , value in enumerate(revenue_by_category['Revenue']):
    plt.annotate(value, xy=(i,value), xytext=(1,10), textcoords='offset points', ha='center')
plt.show()
```

### 3 — Sales Team Performance
A company wants to compare the number of deals closed by each sales representative.

```python
df3 = pd.DataFrame({
    "Sales_Rep": [
        "Amit","Priya","Rahul","Sneha",
        "Karan","Neha","Arjun","Pooja"
    ],
    "Deals_Closed": [
        42, 55, 38, 61,
        47, 52, 35, 58
    ]
})
```

```python
df3
```

```python
df3 = df3.sort_values(by='Deals_Closed',ascending=False)
```

```python
df3
```

```python
plt.figure(figsize=(10,6))
plt.bar(df3['Sales_Rep'], df3['Deals_Closed'])
plt.title('Sales Team Performance')
plt.xlabel('Sales Representative')
plt.ylabel('No of closed deals')
plt.grid(axis='y', alpha = 0.5)
for i , value in enumerate(df3['Deals_Closed']):
    plt.annotate(value, xy=(i, value), xytext=(1,6), textcoords='offset points')
plt.show()
```

```python
colors = [ 'orange' if value == df3['Deals_Closed'].max() else 'steelblue'for value in df3['Deals_Closed']]

plt.figure(figsize=(10,6))
plt.bar(df3['Sales_Rep'], df3['Deals_Closed'], color=colors)
plt.title('Sales Team Performance')
plt.xlabel('Sales Representative')
plt.ylabel('No of closed deals')
plt.grid(axis='y', alpha = 0.5)
for i , value in enumerate(df3['Deals_Closed']):
    plt.annotate(value, xy=(i, value), xytext=(1,6), textcoords='offset points')
plt.show()
```

### 4 — Customer Complaints Analysis
The Customer Experience team wants to identify the most common complaint types so they can prioritize improvement efforts.

```python
 df4 = pd.DataFrame({
    "Complaint_Type": [
        "Network Issue",
        "Billing Error",
        "Call Drop",
        "Slow Internet",
        "App Issue",
        "Network Issue",
        "Call Drop",
        "Network Issue",
        "Billing Error",
        "Slow Internet",
        "Network Issue",
        "Call Drop",
        "App Issue",
        "Network Issue",
        "Slow Internet"
    ]
})
```

```python
df4
```

```python
cmp_count = pd.DataFrame({
    'Complaint_Type': df4['Complaint_Type'].value_counts().index,
    'Count': df4['Complaint_Type'].value_counts().values
})
```

```python
plt.figure(figsize=(10,6))
plt.bar(cmp_count['Complaint_Type'], cmp_count['Count'])
plt.title('Customer Complaints by Type')
plt.xlabel('Complaint Type')
plt.ylabel('Number of Complaints')
plt.grid(axis='y', alpha = 0.5)
for i , value in enumerate(cmp_count['Count']):
    plt.annotate(value, xy=(i, value), xytext=(1,6), textcoords='offset points')
plt.show()
```

### 5 — Employee Count by Department
How many employees are in each department?`

```python
df5 = pd.DataFrame({
    "Department": [
        "Sales",
        "HR",
        "IT",
        "Sales",
        "IT",
        "Sales",
        "Finance",
        "HR",
        "IT",
        "Finance"
    ]
})
```

```python
emp_ct = (
    df5['Department']
    .value_counts()
    .reset_index()
)
```

```python
emp_ct
```

```python
plt.figure(figsize=(10,6))
plt.bar(emp_ct['Department'], emp_ct['count'])
plt.title("Employee Count by Department")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.grid(axis='y', alpha = 0.5)
for i , value in enumerate(emp_ct['count']):
    plt.annotate(value, xy=(i, value), xytext=(1,6), textcoords='offset points')
plt.show()
```
