### Exercise 1: plt.style.use()

### Create a line chart of monthly sales.
- Tasks
- - Apply the ggplot style.
- - Plot sales.
- Add:
- - Title: "Monthly Sales"
- - X-label: "Month"
- - Y-label: "Sales"

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [120, 150, 180, 170, 210]
```

```python
plt.figure(figsize=(8,5))
plt.plot(months, sales, marker='o')
plt.show()
```

```python
plt.style.available
```

```python
plt.figure(figsize=(8,5))
plt.style.use('ggplot')
plt.plot(months, sales, marker='o')
plt.title("Monthly Sales")
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.style.use('fivethirtyeight')
plt.plot(months, sales, marker='o')
plt.title("Monthly Sales")
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.style.use( 'dark_background')
plt.plot(months, sales, marker='o')
plt.title("Monthly Sales")
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.style.use('tableau-colorblind10')
plt.plot(months, sales, marker='o')
plt.title("Monthly Sales")
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.style.use('grayscale')
plt.plot(months, sales, marker='o')
plt.title("Monthly Sales")
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

### Exercise 2: Temporary Style

### Create two line charts using:
- Tasks
- - First chart → dark_background
- - Second chart → default style
- - Use plt.style.context().

```python
x = [1,2,3,4,5]
y = [5,8,6,10,12]
```

```python
with plt.style.context('dark_background'):
    plt.figure(figsize=(6,4))
    plt.plot(x, y, marker='o')
    plt.title('Dark Background Style')
    plt.show()

plt.figure(figsize=(6,4))
plt.plot(x, y, marker='o')
plt.title('Default Style')
plt.show()
```

### Exercise 3: fill_between()

### Tasks
- Plot sales.
- Fill the area between the line and x-axis.
- Use:
- - color = skyblue
- - alpha = 0.4

```python
months = [1,2,3,4,5]
sales = [100,130,120,170,160]
```

```python
plt.figure(figsize=(8,5))
plt.plot(months, sales, marker='o')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(months, sales, marker='o')
plt.fill_between(months, sales, color='skyblue', alpha=0.4)
plt.show()
```

### Exercise 4: Confidence Band

### Tasks
- Plot mean.
- Create a confidence interval band.

```python
x = [1,2,3,4,5]
mean = [20,25,30,35,40]
lower = [18,23,28,33,38]
upper = [22,27,32,37,42]
```

```python
plt.figure(figsize=(8,5))
plt.plot(x, mean, label='Mean')
plt.fill_between(x, lower, upper, alpha=0.3, label='95% Confidence Interval')
plt.legend()
plt.show()
```

### Exercise 5: plt.colorbar()

### Tasks
- Create scatter plot.
- Use profit for colors.
- Add colorbar.
- Label colorbar:
- - Profit (₹)

```python
import numpy as np
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
profit = np.random.randint(1000,10000,100)
```

```python
sc = plt.scatter(x, y, c=profit)
plt.colorbar(sc, label='Profit (₹)')
plt.show()
```
