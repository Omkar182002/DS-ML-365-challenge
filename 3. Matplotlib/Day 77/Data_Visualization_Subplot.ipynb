### Exercise 2:    2×2 Business Dashboard

#### Tasks
Create:
- - Revenue    |  Orders
- - Customers  |  Returns
- Requirements :
- - Revenue → Line plot
- - Orders → Bar plot
- - Customers → Scatter plot
- - Returns → Line plot with markers
- Add titles to all plots.

```python
# dataset
days = [1, 2, 3, 4, 5]

revenue = [1000, 1200, 900, 1400, 1500]
orders = [50, 60, 45, 70, 75]
customers = [120, 140, 130, 160, 180]
returns = [5, 7, 4, 6, 8]
```

```python
plt.subplot(2,2,1)
plt.plot(days, revenue, marker ='^')
plt.title('Revenue by Day')
plt.xlabel('Days')
plt.ylabel('Revenue')
plt.grid()
```

```python
plt.subplot(2,2,2)
plt.bar(days, orders)
plt.title('Orders by Day')
plt.xlabel('Days')
plt.ylabel('Orders')
```

```python
plt.subplot(2,2,3)
plt.scatter(days, customers)
plt.title('Customers vs days')
plt.xlabel('Days')
plt.ylabel('Customers')
plt.grid(True,alpha=0.2)
```

```python
plt.subplot(2,2,4)
plt.plot(days, returns, marker ='o')
plt.title('Returns by Day')
plt.xlabel('Days')
plt.ylabel('Returns')
plt.grid()
```

```python
plt.subplot(2,2,1)
plt.plot(days, revenue, marker ='^')
plt.title('Revenue by Day')
plt.xlabel('Days')
plt.ylabel('Revenue')
plt.grid()

plt.subplot(2,2,2)
plt.bar(days, orders)
plt.title('Orders by Day')
plt.xlabel('Days')
plt.ylabel('Orders')

plt.subplot(2,2,3)
plt.scatter(days, customers)
plt.title('Customers vs days')
plt.xlabel('Days')
plt.ylabel('Customers')
plt.grid(True,alpha=0.2)

plt.subplot(2,2,4)
plt.plot(days, returns, marker ='o')
plt.title('Returns by Day')
plt.xlabel('Days')
plt.ylabel('Returns')
plt.grid()
```

```python
# Use tight_layout().
```

```python
plt.subplot(2,2,1)
plt.plot(days, revenue, marker ='^')
plt.title('Revenue by Day')
plt.xlabel('Days')
plt.ylabel('Revenue')
plt.tight_layout()
plt.grid()

plt.subplot(2,2,2)
plt.bar(days, orders)
plt.title('Orders by Day')
plt.xlabel('Days')
plt.tight_layout()
plt.ylabel('Orders')

plt.subplot(2,2,3)
plt.scatter(days, customers)
plt.title('Customers vs days')
plt.xlabel('Days')
plt.tight_layout()
plt.ylabel('Customers')
plt.grid(True,alpha=0.2)

plt.subplot(2,2,4)
plt.plot(days, returns, marker ='o')
plt.title('Returns by Day')
plt.xlabel('Days')
plt.tight_layout()
plt.ylabel('Returns')
plt.grid()
```

```python
plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
plt.plot(days, revenue, marker ='^')
plt.title('Revenue by Day', weight='bold')
plt.xlabel('Days')
plt.ylabel('Revenue')
plt.grid()

plt.subplot(2,2,2)
plt.bar(days, orders)
plt.title('Orders by Day', weight='bold')
plt.xlabel('Days')
plt.ylabel('Orders')

plt.subplot(2,2,3)
plt.scatter(days, customers)
plt.title('Customers vs days', weight='bold')
plt.xlabel('Days')
plt.ylabel('Customers')
plt.grid(True,alpha=0.2)

plt.subplot(2,2,4)
plt.plot(days, returns, marker ='o')
plt.title('Returns by Day', weight='bold')
plt.xlabel('Days')
plt.ylabel('Returns')
plt.grid()
```

### Exercise 3: Compare Regional Performance

### Tasks
Create:  
North  | South   
East  |  West 
- Requirements:
- - All charts should be line plots.
- - Add markers.
- - Add grid.
- - Give each subplot an appropriate title.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']

north = [100, 120, 140, 150]
south = [90, 100, 110, 130]
east = [80, 95, 105, 120]
west = [110, 125, 145, 160]
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(months, north, marker='o')
plt.title('North')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(months, south, marker='o')
plt.title('South')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(months, east, marker='o')
plt.title('East')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(months, west, marker='o')
plt.title('West')
plt.grid(True)

plt.tight_layout()
plt.show()
```

### Exercise 5: Executive Dashboard (Uneven Layout)

### Tasks
 -Create:  
Revenue Trend  
Orders   | Profit   
- Requirements:
- - Revenue chart should occupy the entire first row.
- - Orders and Profit should be side by side.
- - Add titles and grids.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']

revenue = [1000, 1200, 1300, 1500, 1700]
orders = [50, 55, 60, 70, 75]
profit = [200, 250, 300, 350, 400]
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
plt.plot(months, revenue, marker='o')
plt.title('Revenue Trend')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.bar(months, orders)
plt.title('Orders')
plt.grid(True, alpha=0.2)

plt.subplot(2, 2, 4)
plt.plot(months, profit, marker='o')
plt.title('Profit')
plt.grid(True)

plt.tight_layout()
plt.show()
```
