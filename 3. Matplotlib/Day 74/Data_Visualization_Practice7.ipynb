# Case Study: Mobile Phone Store Sales Analysis 

### Business Problem
A mobile phone store wants to answer:
- Which smartphone brand generates the highest revenue?
- Which month has the highest sales?
- How are sale amounts distributed?
- Do more units sold lead to higher revenue?
- What percentage of revenue comes from each brand?

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Month': [
        'Jan','Jan',
        'Feb','Feb',
        'Mar','Mar',
        'Apr','Apr',
        'May','May'
    ],

    'Brand': [
        'Samsung','Apple',
        'OnePlus','Samsung',
        'Apple','OnePlus',
        'Samsung','Apple',
        'OnePlus','Samsung'
    ],

    'Units_Sold': [
        25,18,
        20,30,
        22,24,
        35,28,
        26,40
    ],

    'Revenue': [
        500000,900000,
        400000,600000,
        1100000,480000,
        700000,1400000,
        520000,800000
    ]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.info()
```

```python
# Data Preparation
brand_revenue = df.groupby('Brand')['Revenue'].sum().reset_index()
```

```python
brand_revenue
```

```python
plt.figure(figsize=(8,5))
plt.bar( brand_revenue['Brand'], brand_revenue['Revenue'])
plt.title('Revenue by Brand')
plt.xlabel('Brand')
plt.ylabel('Revenue')
plt.show()
```

# 

### Task 2: Monthly Revenue Trend
- How does revenue change over time?

```python
# data preapartion
monthly_reveneu = df.groupby('Month')['Revenue'].sum().reset_index()
```

```python
monthly_reveneu
```

```python
plt.figure(figsize=(10,8))
plt.plot(monthly_reveneu['Month'], monthly_reveneu['Revenue'],marker='o')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()
```

# 

### Task 3: Revenue Distribution
- How are sales amounts distributed?

```python
plt.figure(figsize=(8,5))
plt.hist(df['Revenue'], bins=5)
plt.title('Revenue Distribution')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Revenue'], bins=5)

plt.ticklabel_format(style='plain', axis='x')

plt.title('Revenue Distribution')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.show()
```

# 

### Task 4: Units Sold vs Revenue
- Do higher sales volumes generate more revenue?

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Units_Sold'],df['Revenue'])
plt.title('Units Sold vs Revenue',weight='bold')
plt.xlabel('Units Sold')
plt.ylabel('Revenue')
plt.show()
```

# 

### Task 5: Revenue Share by Brand
- How much revenue does each brand contribute?

```python
brand_revenue
```

```python
plt.figure(figsize=(7,7))
plt.pie(brand_revenue['Revenue'], labels=brand_revenue['Brand'], autopct='%0.1f%%')
plt.title('Revenue Share by Brand', weight='bold')
plt.show()
```
