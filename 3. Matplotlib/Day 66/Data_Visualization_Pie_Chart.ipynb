# Data Visualization

## Pie Chart

```python
import matplotlib.pyplot as plt
import pandas as pd
```

# 

###  Revenue Contribution by Product Category
An e-commerce company wants to know which product category contributes most revenue.

```python
df1 = pd.DataFrame({
    'Category':['Electronics','Fashion','Home','Books'],
    'Revenue':[450000,250000,150000,50000]
})
```

```python
df1
```

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'])
plt.show()
```

# 

###### Adding Labels : Displays Categories Names

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'])
plt.show()
```

# 

###### Showing Percentage 

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%')
plt.show()
```

# 

###### Changing Colors 

```python
colors=['gold', 'brown', 'gray', 'olive']
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%', colors=colors)
plt.show()
```

# 

######  Exploding a slice - Used to highlight imp categories

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%', colors=colors, explode=[0,0,0,0.4])
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%', colors=colors, explode=[0,0.1,0,0.1])
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%', colors=colors, explode=[0,0,0,0.2])
plt.show()
```

# 

###### Showing Shadows

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%', colors=colors, explode=[0,0,0,0.2],shadow=True)
plt.show()
```

# 

###### Start Angle : Controls where ploting begins

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%',startangle=90)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%',startangle=180)
plt.show()
```

# 

###### Counter-Clockwise : True - Anti clockwise  .   False : Clockwise

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%',counterclock=False)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%',counterclock=True) # Default True
plt.show()
```

# 

###### Distance of Label : moves labels away from center

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%', labeldistance=1.3)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%', labeldistance=0.8)
plt.show()
```

# 

###### Percentage Distance : Controls Percenatge text position

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%',labeldistance=0.5, pctdistance=1.2)
plt.show()
```

# 

###### Size of Pie Chart

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%', labeldistance=0.7, radius=1.6)
plt.show()
```

# 

###### Fontsize and Fontcolor

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%', textprops={'fontsize':15})
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%', colors= colors, textprops={'fontsize':15, 'color':'b'})
plt.show()
```

# 

###### Label Rotation

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%',rotatelabels=True)
plt.show()
```

# 

###### Edge Properties

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], wedgeprops={'linewidth':5, 'edgecolor':'b', 'linestyle':':'})
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%', colors=colors, explode=[0,0,0,0.2],wedgeprops={'linewidth': 1, 'edgecolor': 'black', 'antialiased': True})
plt.show()
```

# 

```python
plt.figure(figsize=(7,7))
plt.pie(df1['Revenue'], labels=df1['Category'], autopct='%0.1f%%',counterclock=True) 
plt.title('Revenue Contribution by Category',fontsize=16, weight='bold')
plt.legend(loc=4)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(df1['Revenue'], autopct='%0.1f%%',counterclock=True) 
plt.title('Revenue Contribution by Category',fontsize=16, weight='bold')
plt.legend(df1['Category'], loc=4)
plt.show()
```

# 

# Create custom labels with values or percentages

```python
custom_labels = [f'{cat}\n(${rev/1000:.0f}K)' for cat, rev in zip(df1['Category'], df1['Revenue'])]
plt.pie(df1['Revenue'], labels=custom_labels, autopct='%0.1f%%')
plt.title('Revenue Contribution by Category',fontsize=16, weight='bold')
plt.show()
```

```python
plt.pie(df1['Revenue'], autopct='%0.1f%%',counterclock=True) 
plt.title('Revenue Contribution by Category',fontsize=10, weight='bold')
plt.legend(df1['Category'],  loc="center left",bbox_to_anchor=(1, 0, 0.5, 1))  # place legend outside/
plt.show()
```

```python
total = df1['Revenue'].sum()
labels_with_pct = [f"{cat} ({rev/total*100:.1f}%)" for cat, rev in zip(df1['Category'], df1['Revenue'])]
plt.pie(df1['Revenue'])
plt.title('Revenue Contribution by Category',fontsize=10, weight='bold')
plt.legend(labels_with_pct, title="Categories", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()
```

## Basic Donut Chart

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], radius=1)
plt.pie([1], radius=0.7, colors='w')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'],radius=1)
plt.pie([1], radius=0.6, colors='w')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'],radius=1, autopct='%0.1f%%')
plt.pie([1], radius=0.6, colors='w')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'], labels=df1['Category'],radius=1, autopct='%0.1f%%', pctdistance=0.8)
plt.pie([1], radius=0.6, colors='w')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'],radius=1, autopct='%0.1f%%', pctdistance=0.8)
plt.pie([1], radius=0.6, colors='w')
plt.legend(df1['Category'],  loc="center left",bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.pie(df1['Revenue'],radius=1, autopct='%0.1f%%', pctdistance=0.8, explode=[0.01,0.02,0.02,0.01])
plt.pie([1], radius=0.6, colors='w')
plt.legend(df1['Category'],  loc="center left",bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()
```
