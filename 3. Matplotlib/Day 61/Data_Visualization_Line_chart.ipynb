#### Exercise 1: Monthly Sales Trend Analysis
The Sales Director wants to understand how sales have changed over the first six months of the year.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [120000, 135000, 128000, 145000, 160000, 170000]
```

# 

```python
plt.figure(figsize=(10,5)) # set canvas or window for draw chart 
plt.plot(months, sales, color = 'blue', linestyle='--', linewidth = 2, marker='o', markerfacecolor= 'red')
plt.title('Monthly Sales Trend')
plt.xlabel('Months')
plt.ylabel('Sales ($)')
plt.show()
```

```python
plt.figure(figsize=(10,5)) # set canvas or window for draw chart 
plt.plot(months, sales, color = 'blue', linestyle='-', linewidth = 2, marker='o', markerfacecolor= 'red')
plt.title('Monthly Sales Trend')
plt.xlabel('Months')
plt.ylabel('Sales ($)')
plt.show()
```

```python
plt.figure(figsize=(10,5)) # set canvas or window for draw chart 
plt.plot(months, sales, color = 'blue', linestyle='--', linewidth = 2, marker='o', markerfacecolor= 'red')
plt.title('Monthly Sales Trend')
plt.xlabel('Months')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.show()
```

### Exercise 3: Revenue vs Target Revenue
Management compares actual revenue with targets.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
actual = [95, 105, 110, 120, 135, 145]
target = [100, 100, 110, 115, 130, 140]
```

```python
plt.figure(figsize=(10,5),facecolor='silver')
plt.plot(months, actual,color = 'blue', marker='s', markerfacecolor='black', linestyle=':', linewidth=2)
plt.plot(months, target,color = 'red', marker='o', markerfacecolor='green', linestyle='-', linewidth=2)
plt.title('Revenue vs Target Revenue')
plt.xlabel('Months')
plt.ylabel('Revenue')
plt.grid(True, linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(10,5),facecolor='silver')
plt.plot(months, actual,color = 'blue', marker='s', markerfacecolor='black', linestyle=':', label='Actual')
plt.plot(months, target,color = 'red', marker='o', markerfacecolor='green', linestyle='-', label='Target')
plt.title('Revenue vs Target Revenue')
plt.xlabel('Months')
plt.ylabel('Revenue')
plt.grid(True, linestyle='--')
plt.legend()
plt.show()
```

### Exercise 4: Customer Retention Trend
Customer Success team tracks retention rate monthly.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug']
retention = [85, 84, 83, 86, 88, 87, 89, 91]
```

```python
plt.figure(figsize=(10,5))
plt.plot(months, retention, color='blue', linestyle='--', marker='^', markerfacecolor='red')
plt.xlabel('Months')
plt.ylabel('Percentage')
plt.title('Customer Retention Trend')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(months, retention, color='blue', linestyle='--', marker='^', markerfacecolor='red')
plt.xlabel('Months')
plt.ylabel('Percentage')
plt.title('Customer Retention Trend')
plt.grid(True)
for i , value in enumerate(retention):
    plt.annotate(value, xy=(i,value))
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(months, retention, color='blue', linestyle='--', marker='^', markerfacecolor='red')
plt.xlabel('Months')
plt.ylabel('Percentage')
plt.title('Customer Retention Trend')
plt.grid(True)
for i , value in enumerate(retention):
    plt.annotate(value, xy=(i,value), textcoords="offset points")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(months, retention, color='blue', linestyle='--', marker='^', markerfacecolor='red')
plt.xlabel('Months')
plt.ylabel('Percentage')
plt.title('Customer Retention Trend')
plt.grid(True)
for i , value in enumerate(retention):
    plt.annotate(value, xy=(i,value), textcoords="offset points",xytext=(0, 10))
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(months, retention, color='blue', linestyle='--', marker='^', markerfacecolor='red')
plt.xlabel('Months')
plt.ylabel('Percentage')
plt.title('Customer Retention Trend')
plt.grid(True)
for i , value in enumerate(retention):
    plt.annotate(value, xy=(i,value), textcoords="offset points",xytext=(0, 10), ha='center')
plt.show()
```

```python
plt.plot(weeks, resolution_time, linestyle=':', marker='o')
plt.title('Call Center Trend')
plt.xlabel('Weeks')
plt.ylabel('Resolution Time')
plt.grid(True)
for i , value in enumerate(resolution_time):
    plt.annotate(value, xy=(i,value), xytext=(1,11),textcoords='offset points',ha='center',color='Red', fontsize=10 )
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(weeks, resolution_time, linestyle=':', marker='o')
plt.title('Call Center Trend')
plt.xlabel('Weeks')
plt.ylabel('Resolution Time')
plt.grid(True)
for i , value in enumerate(resolution_time):
    plt.annotate(value, xy=(i,value), xytext=(1,11),textcoords='offset points',ha='center',color='Red', fontsize=10 )
plt.show()
```

#### Deepseek code

```python


# Data
weeks = ['W1', 'W2', 'W3', 'W4', 'W5', 'W6']
resolution_time = [12.5, 11.8, 11.2, 10.5, 9.8, 9.1]

# Create plot
plt.figure(figsize=(10, 6))
plt.plot(weeks, resolution_time, marker='o', linewidth=2, markersize=8)

# Add annotations
plt.annotate('Starting point\n12.5 min', 
             xy=('W1', 12.5), 
             xytext=('W1', 13.5),
             arrowprops={'arrowstyle': '->', 'color': 'red'})

plt.annotate('Target achieved!\n9.1 min', 
             xy=('W6', 9.1), 
             xytext=('W5', 8.5),
             arrowprops={'arrowstyle': '->', 'color': 'green'})

plt.annotate('Best improvement\nW3→W4 (-0.7 min)', 
             xy=('W4', 10.5), 
             xytext=('W3.5', 11.5),
             arrowprops={'arrowstyle': '->', 'color': 'blue'})

# Labels and title
plt.xlabel('Week')
plt.ylabel('Resolution Time (minutes)')
plt.title('Call Center Performance - Average Resolution Time Trend')
plt.grid(True, alpha=0.3)

# Add target line
plt.axhline(y=9.0, color='green', linestyle='--', alpha=0.7, label='Target: 9 min')
plt.legend()

plt.show()
```

### Exercise 6: Product Category Sales Comparison
Compare sales trends across product categories.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
electronics = [50, 55, 60, 70, 75, 80]
fashion = [40, 42, 45, 48, 50, 53]
grocery = [30, 35, 38, 42, 46, 50]
```

```python
plt.figure(figsize=(10,6))
plt.plot(months, electronics, color= 'blue', marker='o',label='Electronics')
plt.plot(months, fashion, color= 'gray', marker='s',label='Fashion')
plt.plot(months, grocery, color= 'green', marker='^',label='Grocery')
plt.title('Product Category Sales: Electronics Leads Growth')
plt.xlabel('Month')
plt.ylabel('Sales (thousands)')
plt.grid(True,alpha=0.3)
plt.legend()
plt.show()

```

### Exercise 7: Stock Inventory Monitoring
Supply Chain team monitors inventory levels.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
inventory = [5000, 4800, 4500, 4300, 4100, 3900]
```

```python
plt.figure(figsize=(10,6))
plt.plot(months, inventory, color= 'blue', marker='o')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Inventory Level (units)', fontsize=12)
plt.title('Stock Inventory Monitoring', fontsize=14, fontweight='bold')
for i, v in enumerate(inventory):
    plt.text(months[i], v + 50, str(v), ha='center', fontweight='bold')
plt.grid(True,alpha=0.3)
plt.show()

```
