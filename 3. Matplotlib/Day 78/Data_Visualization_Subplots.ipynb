## Matplotlib Subplots

```python
import matplotlib.pyplot as plt
```

### Exercise 1: Single Plot

### Tasks
Create one figure and one axes using plt.subplots().
- Plot sales as a line chart.
- Add:
- - Title: "Monthly Sales"
- - X-label: "Month"
- - Y-label: "Sales"
- - Grid

```python
# dataset
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [120, 150, 170, 160, 190]
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(months, sales, marker='o')
ax.set_title('Monthly Sales')
ax.set_xlabel('Month')
ax.set_ylabel('Sales')
ax.grid(True)
plt.show()
```

### Exercise 2: 1 Row × 2 Columns

### Tasks
Create a (1, 2) layout.
-  Plot:
- -  Sales → Line chart
- - Profit → Bar chart
- Add titles and grids.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [120, 150, 170, 160, 190]
profit = [20, 30, 35, 33, 40]
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,5))
ax[0].plot(months, sales, marker='o')
ax[0].set_title('Monthly Sales')
ax[0].set_xlabel('Month')
ax[0].set_ylabel('Sales')
ax[0].grid()

ax[1].bar(months,profit)
ax[1].set_title('Monthly Profit')
ax[1].set_xlabel('Month')
ax[1].set_ylabel('Profit')
plt.tight_layout()
plt.show()
```

### Exercise 3: 2 Rows × 1 Column

### Tasks
Create (2, 1) subplots.
- Use sharex=True.
- Add markers.
- Add grids.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
revenue = [1000, 1200, 1400, 1500, 1700]
orders = [50, 55, 60, 70, 75]
```

```python
fig, ax = plt.subplots(2,1,figsize=(10, 8),sharex=True)

ax[0].plot(months, revenue, marker='o')
ax[0].set_title('Revenue')
ax[0].grid(True)
ax[1].plot(months, orders, marker='o')
ax[1].set_title('Orders')
ax[1].grid(True)
plt.tight_layout()
plt.show()
```

### Exercise 4: 2 × 2 Dashboard

### Tasks
Create (2, 2) subplots.
- Use four different chart types:
- - Line
- - Bar
- - Scatter
- - Histogram
- Add:
- - Titles
- - Grids
- - Figure title: "Business Dashboard"

```python
days = [1, 2, 3, 4, 5]
revenue = [1000, 1200, 900, 1400, 1500]
orders = [50, 60, 45, 70, 75]
customers = [120, 140, 130, 160, 180]
returns = [5, 7, 4, 6, 8]
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Business Dashboard', fontsize=16)

ax[0, 0].plot(days, revenue, marker='o')
ax[0, 0].set_title('Revenue')
ax[0, 0].grid(True)

ax[0, 1].bar(days, orders)
ax[0, 1].set_title('Orders')
ax[0, 1].grid(True)

ax[1, 0].scatter(days, customers)
ax[1, 0].set_title('Customers')
ax[1, 0].grid(True)

ax[1, 1].hist(returns, bins=3)
ax[1, 1].set_title('Returns')
ax[1, 1].grid(True)

plt.tight_layout()
plt.show()
```

### Exercise 5: 1 × 4 Comparison Dashboard

### Tasks
Create (1, 4) subplots.
- Use sharey=True.
- Add markers and grids.
- Give each subplot a title.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']
north = [100, 120, 140, 160]
south = [90, 110, 125, 135]
east = [80, 100, 115, 130]
west = [110, 130, 145, 170]
```

```python
fig, ax = plt.subplots(1,4,figsize=(15, 4),sharey=True)
ax[0].plot(months, north, marker='o')
ax[0].set_title('North')
ax[0].grid(True)

ax[1].plot(months, south, marker='o')
ax[1].set_title('South')
ax[1].grid(True)

ax[2].plot(months, east, marker='o')
ax[2].set_title('East')
ax[2].grid(True)

ax[3].plot(months, west, marker='o')
ax[3].set_title('West')
ax[3].grid(True)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1,4,figsize=(15, 4),sharey=True)
regions = {'North': north,
           'South': south,
            'East': east,
            'West': west
}

for ax, (region, values) in zip(axes, regions.items()):
    ax.plot(months, values, marker='o')
    ax.set_title(region)
    ax.grid(True)

plt.tight_layout()
plt.show()
```
