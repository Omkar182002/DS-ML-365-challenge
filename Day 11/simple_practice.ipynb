#### Shape manipulation


```python

a = np.array([[1, 2],
              [3, 4]])  # shape: (2, 2)
np.swapaxes(a, 0, 1)
# print(np.swapaxes(a, 0, 1))
print(np.swapaxes(a,  1,0))
# print(np.swapaxes(a, 0, 2))
```

##### expand_dims()

```python
a = np.array([1, 2, 3])        # shape: (3,)
b = np.expand_dims(a, axis=0) # shape: (1, 3)
c = np.expand_dims(a, axis=1) # shape: (1, 3)
print(b)
print(c)
```
