### FUNCTION 3: np.savez()

#### Q11. Save all three arrays (names, ages, scores) into one file "student_data.npz" using np.savez().

```python
np.savez("students_data.npz",names,ages,scores)
```

```python
all_data = np.load("students_data.npz")
all_data
```

### FUNCTION 4: np.load() with .npz files
Load multiple arrays from compressed .npz file.

```python
# Create and save .npz file
x = np.array([1, 2, 3, 4])
y = np.array([5, 6, 7, 8])
z = np.array([9, 10, 11, 12])
np.savez("xyz_data.npz", first=x, second=y, third3=z)
print("Saved x, y, z to xyz_data.npz")
```

### FUNCTION 5: np.savetxt()
Save array to text file (.txt, .csv) with custom delimiter.

```python
sales = np.array([
    [1200, 1350, 1400, 1600],
    [800, 950, 1100, 1250],
    [2000, 2100, 2300, 2500],
    [500, 600, 700, 800],
    [1500, 1650, 1750, 1900]
])
print("Sales data (products × quarters):\n", sales)
```

#### Q21. Save sales array to "sales.csv" with comma delimiter using np.savetxt().

```python
np.savetxt("sales.csv",sales,delimiter=',')
```

#### Q22. Save same array to "sales.txt" with space delimiter.

```python
np.savetxt("sales.txt",sales,delimiter=' ')
```

```python
np.savetxt("sales.csv",sales,delimiter=',')
```

#### Q23. Save as integers with format "%d" (no decimal points).

```python
np.savetxt("sales.csv",sales,delimiter=',',fmt="%d")
```

#### Q24. Save only first 3 rows and first 2 columns to "subset_sales.csv".

```python
np.savetxt("subset_sales.csv",sales[:3,:2])
```

# FUNCTION 7: np.genfromtxt()
Advanced text loading with handling missing values.

#### Array (Create file with missing values)

```python
# Create data with missing values as string
data_with_missing = """ID,Age,Score
1,25,85
2,,92
3,30,
4,28,78
5,,88
"""
with open("missing_data.csv", "w") as f:
    f.write(data_with_missing)
print("Created missing_data.csv")
```

#### Q31. Load "missing_data.csv" using np.genfromtxt() with delimiter=','.

```python
miss_data = np.genfromtxt("missing_data.csv",delimiter=',')
print(miss_data)
```

#### Q32. Handle missing values by filling with 0 using filling_values=0.

```python
new_data = np.genfromtxt("missing_data.csv" ,delimiter=',',filling_values=0)
new_data
```

#### Q33. Skip header row using skip_header=1.

```python
new_data2 = np.genfromtxt("missing_data.csv" ,delimiter=',',skip_header=1)
new_data2
```

#### Q34. Use missing_values="" and filling_values=-1 to mark missing as -1.

```python
new_data3 = np.genfromtxt("missing_data.csv" ,delimiter=',',filling_values=-1,missing_values='')
new_data3
```

#### Q35. Load only numeric columns (skip ID column) using usecols=(1,2).

```python
new_data4 = np.genfromtxt(
    "missing_data.csv",
    delimiter=',',
    usecols=(1, 2),
    skip_header=1,
    filling_values=-1
)
print(new_data4)
```
