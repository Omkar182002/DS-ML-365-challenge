```python
df = df.drop_duplicates()
```

```python
df.dropna(subset='Name',inplace=True)
```

```python
df['Age'].unique()
```

```python
df['Age'] = df['Age'].replace(r'[^0-9.]','',regex=True)
```

```python
df['Age'] = pd.to_numeric(df['Age'],errors='coerce').round()
```

```python
df['Age'].head()
```

```python
df.loc[(df['Age']<0)|(df['Age']>110),'Age'] = np.nan
```

```python
df['Age'].head()
```

```python
df['Age'] = df['Age'].astype('Int64')
```

```python
df['Department'].unique()
```

```python
df['Department'].str.strip().str.title().unique()
```

```python
dept_map={
    'Mamketing' : 'Marketing', 
    'Operathons' : 'Operations',  
    'Operationp' : 'Operations',
    'Customer Supoort' : 'Customer Support', 
    'Customerrsupport': 'Customer Support',
    'Engineerinj' : 'Engineering',
    'Dezign' : 'Design',
    'Dekign' : 'Design', 
    'Fznance' : 'Finance'
}
```

```python
df['Department'].str.strip().str.title().replace(dept_map).unique()
```

```python
df['Department'] = df['Department'].str.strip().str.title().replace(dept_map)
```

```python
df['Department'].fillna(df['Department'].mode()[0]).unique()
```

```python
df['Department'] =df['Department'].fillna(df['Department'].mode()[0])
```

```python
df['Age'] = df['Age'].fillna(df.groupby('Department')['Age'].transform(lambda a : a.mode()[0]))
```
