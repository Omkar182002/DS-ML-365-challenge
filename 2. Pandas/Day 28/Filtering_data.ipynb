# Filtering Data

```python
import pandas as pd
```

```python
import numpy as np
```

## Series Filtering

```python
s = pd.Series([10, 25, 30, 45, 50, 15, 5, 60, 35, 20])
```

```python
s
```

#### 1. Get values greater than 30

```python
s[s>30]
```

#### 2. Get values less than 20

```python
s[s<20]
```

#### 3. Get values equal to 25

```python
s[s==25]
```

#### 4. Get values not equal to 10

```python
s[s!=10]
```

#### 5. Get values greater than or equal to 50. 

```python
s[s>=50]
```

#### 6. Get values between 20 and 40

```python
s[(s>20) &(s<40)]
```

#### 7. Get even numbers only

```python
s[s%2==0]
```

#### 8. Get even numbers only

```python
s[s%2!=0]
```

#### 9. Get values divisible by 5

```python
s[s%5==0]
```

#### 10. Get top 3 highest values

```python
s.nlargest(3)
```

#### 11. Get values greater than mean

```python
mean = np.mean(s)
s[s>mean]
```

```python
s[s > s.mean()]
```

#### 12. Get values less than median

```python
mdm = np.median(s)
s[s<mdm]
```

```python
s[s < s.median()]
```

#### 13. Get values in list [10, 20, 30]

```python
s[s in [10,20,30]]
```

```python
s[s.isin([10, 20, 30])]
```

#### 14. Remove values less than 15

```python
s[s >= 15]
```

#### 15. Replace values > 40 with 100 (filter + modify)

```python
s.where(s <= 40, 100)
```

## DataFrame Filtering

```python
data = pd.read_csv('col_row-dataset.csv')
```

```python
data.sample(10)
```

#### 1. Get rows where age > 22

```python
data[data['age']>22]
```

#### 2. Get rows where salary < 30000

```python
data[data["salary"]<30000].head()
```

#### 3. Get rows where city = "Delhi"

```python
data[data["city"]=='Delhi'].head()
```

#### 4. Get rows where department = "IT"

```python
data[data["department"]=='IT'].head()
```

#### 5. Get rows where experience >= 5

```python
data[data['experience']>=5].head(10)
```

#### 6. Get rows where age > 25 AND salary > 40000

```python
data[data[(data['age']>25)&(data['salary']>40000)]]
```

#### 7. Get rows where city = "Mumbai" OR "Pune"

```python
data[data['city'].isin(['Mumbai','Pune'])]
```

#### 8. Get rows where salary is between 30000 and 50000

```python
data[(data['salary']>=30000) &(data['salary']<=50000)]
```

```python
data[data['salary'].between(30000, 50000)]
```

#### 9 . Get rows where age is null 

```python
data[data['age']==None]
```

```python
data[data['age'].isnull()]
```

#### 10. Get rows where city is not null

```python
data[data['city'].notnull()]
```

#### 11. Get rows where salary > average salary

```python
data[data['salary'] > data['salary'].mean()]
```

#### 12. Get rows where experience < median experience

```python
data[data['experience'] < data['experience'].median()]
```

#### 13. Get top 10 highest salary rows

```python
data.nlargest(10, 'salary')
```

#### 14. Get rows where name starts with "A"

```python
data[data['name'].str.startswith('A')]
```

#### 15. Get rows where department is NOT "HR"

```python
data[data['department'] != "HR"]
```
