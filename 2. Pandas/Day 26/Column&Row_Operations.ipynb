```python
data1 = data.copy()
data1.head()
```

```python
data1.drop(columns={"salary"},inplace=True)
data1.head()
```

#### 6. Create a new column salary_in_k (salary / 1000)

```python
data1["salary_in_k"] = data["salary"]/1000
```

#### 7. Convert all column names to lowercase

```python
data1.columns = data1.columns.str.lower()
```

```python
data1.columns
```

```python
data1 = data1[['city','department','experience','age','salary_in_k','full_name','id','age_plus']]
```

```python
data1.head()
```

#### 9. Replace column city with uppercase values

```python
data1['city'] = data1['city'].str.upper()
```

```python
data1.head()
```

#### 10. Create a column is_adult (age >= 18 → True/False)

```python
data1["Is_adult"] = data1["age"]>=18
```

```python
data1.head(10)
```

```python
data1.sample(5)
```

#### 11. Create column salary_category:
<30000 → Low
30000–40000 → Medium

40000 → High

```python
def category(a):
    if a < 30000:
        return 'low'
    elif a < 40000 :
        return 'Medium'
    else:
        return "High"
data1["Salary_category"] = data["salary"].apply(category)
```

```python
data1.sample(10)
```

#### 12. Drop multiple columns at once (age, salary)
