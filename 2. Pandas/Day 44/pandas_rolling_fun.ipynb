# Pandas Rolling Function

```python
import pandas as pd
```

## 1 — Rolling Mean

```python
sales_df = pd.DataFrame({'Sales': [100, 200, 300, 400, 500, 600]})
```

```python
sales_df
```

### 1. Create a 2-row rolling mean.

```python
sales_df.rolling(window=2).mean()
```

```python
sales_df['Sales'].rolling(window=2).mean()
```

```python
sales_df['Sales'].rolling(2).mean()
```

# 

### 2. Create a 3-row rolling mean.

```python
sales_df['Sales'].rolling(window=3).mean()
```

```python
sales_df.rolling(3).mean()
```

### 3. Create a 4-row rolling mean.

```python
sales_df.rolling(4).mean()
```

### 4. Store result in a new column named Rolling_Avg.

```python
sales_df['Rolling_Avg'] = sales_df['Sales'].rolling(2).mean()
```

```python
sales_df
```

### 5. Find which row gets the first non-NaN value.

```python
print(sales_df['Rolling_Avg'].first_valid_index())
```

### 6. Explain why first rows become NaN.

```python
# rolling(3)
# Pandas needs 3 values to calculate the mean.
# Row 0 has only [10] → not enough → NaN
# Row 1 has only [10,20] → not enough → NaN
# Row 2 has [10,20,30] → enough → mean = 20
```

```python
#WE can  avoid initial NaN: by using parameter  ----- min_periods. 
# min_periods=1 means calculation can start even with 1 value.
```

### 7. Create rolling mean using min_periods=1.

```python
sales_df.rolling(window= 3,min_periods=1).mean()
```

```python
sales_df['Sales'].rolling(window= 3,min_periods=1).mean()
```

### 8. Compare normal rolling mean vs min_periods=1.

```python
sales_df['Min_Periods'] = sales_df['Sales'].rolling(2, min_periods=1).mean()
```

```python
print(sales_df[['Rolling_Avg', 'Min_Periods']])
```

### 9. Find overall difference between original sales and rolling average.

```python
sales_df['Difference'] = sales_df['Sales'] - sales_df['Rolling_Avg']
```

```python
sales_df
```

### 10. Create centered rolling mean using center=True.

```python
sales_df['Sales'].rolling(window= 3,min_periods=1, center=True).mean()
```

```python
sales_df['Sales'].rolling(window= 3,min_periods=1).mean()
```

## 9 — GroupBy + Rolling

```python
store_df = pd.DataFrame({
    'Store': ['A', 'A', 'A', 'B', 'B', 'B'],
    'Sales': [100, 200, 300, 400, 500, 600]
})
```

```python
store_df
```

### 1. Create rolling mean separately for each store.

```python
store_df['Mean'] = (store_df.groupby('Store')['Sales'].rolling(2).mean().reset_index(level=0, drop=True))
store_df
```

### 2. Create rolling sum separately for each store.

```python
store_df['Sum'] = (store_df.groupby('Store')['Sales'].rolling(2).sum().reset_index(level=0, drop=True))
store_df
```

### 3. Create rolling max separately for each store.

```python
store_df['Max'] = (store_df.groupby('Store')['Sales'].rolling(2).max().reset_index(level=0, drop=True))
```

```python
store_df
```

### 4. Create rolling min separately for each store.

```python
store_df['Min'] = (store_df.groupby('Store')['Sales'].rolling(2).min().reset_index(level=0, drop=True))
store_df
```

### 5. Create rolling std separately for each store.

```python
store_df['Std'] = (store_df.groupby('Store')['Sales'].rolling(2).std().reset_index(level=0, drop=True))
```

```python
store_df
```
