# 

```python
df.drop_duplicates(inplace=True)
```

```python
df['Employee_ID'] = df['Employee_ID'].astype('Int64')
```

```python
df['Employee_ID'] = df['Employee_ID'].astype('str')
```

```python
df['Employee_ID'].head()
```

# 

```python
df['Employee_Name'] = df['Employee_Name'].astype(str).str.strip().str.title()
```

```python
df['Employee_Name'].unique()
```

```python
df['Employee_Name'] = df['Employee_Name'].replace({'':'Unknown','None':'Unknown'})
```

```python
df['Employee_Name'].unique()
```

```python
df['Employee_Name'] = df['Employee_Name'].fillna('Unknown')
```

```python
df['Attendance_Status'].unique()
```

```python
df['Attendance_Status'] = df['Attendance_Status'].replace({
    'present':'Present',
    'PRESENT':'Present',
    'P':'Present',
    'absent':'Absent'
})
```

```python
df['Attendance_Status'] = df['Attendance_Status'].fillna('Unknown')
```

```python
df['Attendance_Status'].unique()
```

```python
df['Work_Location'].unique()
```

```python
df['Work_Location'] = df['Work_Location'].replace({'bengaluru':'Bangalore','BLR':'Bangalore','Bangalore City':'Bangalore'})
```

```python
df['Work_Location'].unique()
```

```python
df['Work_Location'].isnull().sum()
```

```python
df['Work_Location'] = df['Work_Location'].fillna('Unknown')
```

```python
df['Work_Location'].isnull().sum()
```

# 

```python
df['Hours_Worked'] = pd.to_numeric(df['Hours_Worked'], errors='coerce')
```

```python
df['Hours_Worked'].head()
```

```python
df.loc[(df['Hours_Worked'] < 0) | (df['Hours_Worked'] > 24), 'Hours_Worked'] = np.nan
```

```python
df['Hours_Worked'] = df['Hours_Worked'].fillna(df['Hours_Worked'].median())
```

```python
df['Hours_Worked'].head()
```
