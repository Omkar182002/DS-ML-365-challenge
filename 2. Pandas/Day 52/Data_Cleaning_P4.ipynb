### Data Validation

```python
import re
```

```python
pattern = r'^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$'
```

```python
valid_email = df['Email'].str.match(pattern,na=False)
```

```python
df.loc[~valid_email, 'Email'] = pd.NA
```

```python
mode_values = df.groupby('Order_ID').agg({
    'City': lambda x: x.mode().iloc[0] if not x.mode().empty else pd.NA,
    'Order_Date': lambda x: x.mode().iloc[0] if not x.mode().empty else pd.NA
}).reset_index()
```

```python
mode_values
```

```python
mode_values.rename(columns={
    'City': 'Valid_City',
    'Order_Date': 'Valid_Date'
}, inplace=True)
```

```python
df = df.merge(mode_values, on='Order_ID')
```

```python
df.head()
```

```python
df = df[(df['City'] == df['Valid_City']) &(df['Order_Date'] == df['Valid_Date'])]
```

```python
df.drop(columns=['Valid_City', 'Valid_Date'], inplace=True)
```
