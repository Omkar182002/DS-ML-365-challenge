### 20. Classify all values into:

"Low" for marks below 40
"Medium" for marks between 40 and 75
"High" for marks above 75

```python
def classify(m):
    if m < 40:
        return "Low"
    elif m <= 75:
        return "Medium"
    else:
        return "High"
```

```python
df1.applymap(classify)
```
