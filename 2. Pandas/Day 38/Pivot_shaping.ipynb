### A.  In Student Data  : Convert Subjects should become columns.

```python
dataA = {
    'Student': ['A','A','B','B'],
    'Subject': ['Math','Science','Math','Science'],
    'Marks': [90,85,78,88]
}
dfA = pd.DataFrame(dataA)
print('-------DataFrame ---------')
print(dfA)
pvA = dfA.pivot(index='Student',columns='Subject',values='Marks')
print()
print('----After Pivot---')
print(pvA)
```

### B. In Hospital Daily Patients data :  Departments become columns.

```python
hospital_data = {
    'Date': ['1-Jan','1-Jan','2-Jan','2-Jan'],
    'Department': ['Cardio','ENT','Cardio','ENT'],
    'Patients': [45,30,50,28]
}

dfB = pd.DataFrame(hospital_data)
print('-------DataFrame ---------')
print(dfB)
pvB = dfB.pivot(
    index='Date',
    columns='Department',
    values='Patients'
)
print()
print('--Pivote conversion -------------')
print(pvB)
```

```python
C. In  Website Traffic Data : Sources become columns.
```

```python
web_data = {
    'Date': ['Mon','Mon','Mon'],
    'Source': ['Google','Instagram','YouTube'],
    'Visitors': [5000,3200,4100]
}

dfC = pd.DataFrame(web_data)

pvC = dfC.pivot(
    index='Date',
    columns='Source',
    values='Visitors'
)

print("---DF---")
print(dfC)
print()
print(pvC)
```

### D. Employee data : Employees become columns.

```python
emp_data = {
    'Date': ['Mon','Mon','Tue','Tue'],
    'Employee': ['Rahul','Amit','Rahul','Amit'],
    'Status': ['P','A','P','P']
}

dfD = pd.DataFrame(emp_data)

pvD = dfD.pivot(index='Date',columns='Employee',values='Status')
print(dfD)
print()
print(pvD)
```

### E. Cricket Data :  Players become columns.

```python
cr_data = {
    'Match': ['Match1','Match1','Match2','Match2'],
    'Player': ['Virat','Rohit','Virat','Rohit'],
    'Runs': [80,45,70,95]
}

dfE = pd.DataFrame(cr_data)

pvE = dfE.pivot(index='Match',columns='Player',values='Runs')
print(dfE)
print()
print(pvE)
```
