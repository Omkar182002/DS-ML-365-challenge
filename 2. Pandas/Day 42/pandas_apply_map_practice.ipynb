## 1. map()

```python
df1 = pd.DataFrame({
    "Name": ["Omkar", "Rahul", "Sneha", "Amit", "Pooja"],
    "Gender": ["M", "M", "F", "M", "F"],
    "Marks": [78, 56, 91, 45, 88],
    "City": ["Belgaum", "Delhi", "Mumbai", "Pune", "Bangalore"]
})

print(df1)
```

```python
df1.info()
```

### 1.Convert Gender: M → Male,  F → Female

```python
 df1.map({'M':'Male', 'F':'Female'})
```

### 2. Convert Marks into grades:
##### 90+ → A
##### 70-89 → B
##### 50-69 → C
##### Below 50 → Fail

```python
df1.columns
```

```python
dfc2 = df1.copy()
```

```python
def grade(x):
    if x >= 90:
        return "A"
    elif x >= 70:
        return "B"
    elif x >= 50:
        return "C"
    else:
        return "Fail"

dfc2["Grade"] = dfc2["Marks"].map(grade)
```

## 2. apply()

```python
df2 = pd.DataFrame({
    "Math": [78, 56, 90, 45, 88],
    "Science": [67, 72, 95, 40, 91],
    "English": [80, 60, 89, 50, 85]
})
df2
```

```python
df2.info()
```

```python
marks_df = df2.copy()
```

### 1. Find total marks row-wise using apply().

```python
marks_df["Total"] = marks_df.apply(sum, axis=1)
```

```python
marks_df
```

### 2. Find average marks row-wise.

```python
marks_df["Avg"] = marks_df.apply(mean, axis=1)
```

```python
marks_df["Avg"] = marks_df[["Math", "Science", "English"]].apply(lambda row: row.mean(), axis=1)
```

```python
marks_df
```

### 18. Create custom performance labels:

Excellent
Good
Average
Poor

```python
def performance(avg):
    if avg >= 85:
        return "Excellent"
    elif avg >= 70:
        return "Good"
    elif avg >= 50:
        return "Average"
    else:
        return "Poor"

marks_df["Performance"] = marks_df[["Math", "Science", "English"]].apply(lambda row: performance(row.mean()),axis=1)

print(marks_df)
```

### 19. Find topper subject for each student.
