# Extra Spaces

```python
df.head()
```

```python
ob_col = df.select_dtypes(include='object').columns
```

```python
ob_col
```

```python
for col in ob_col:
    df[col] = df[col].str.strip()
```

```python
df["Name"] = df["Name"].str.title()
```

```python
df.loc[:,"Name"] = df["Name"].str.title()
```

```python
df['Name'] = df['Name'].str.replace(r'[^A-Za-z]','',regex=True)
```

```python
df['Name'] = df['Name'].str.replace(r'^\s*$','',regex=True)
```

# 

```python
df['Department'].unique()
```

```python
df.loc[:,"Email"] = df["Email"].str.lower()
```

```python
df['Email'] = df['Email'].str.replace(r'^\s*$','',regex=True)
```

# Missing Values

```python
df.isna().sum()
```

```python
(df.isna().mean())*100
```

```python
df['Age'].unique()
```

```python
df['Department'].unique()
```

```python
df['Department'] = df['Department'].fillna(df['Department'].mode()[0])
```

```python
df['Department'].unique()
```

```python
df['Department'].isna().sum()
```

```python
df['Age'] = df.groupby('Department')['Age'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isna().sum()
```

```python
df['Salary'] = df.groupby('Department')['Salary'].transform(lambda x: x.fillna(x.median()))
```

```python
df['Salary'].isna().sum()
```

```python
df['City'] = df['City'].fillna(df['City'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['Experience'] = df.groupby('Department')['Experience'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isna().sum()
```
