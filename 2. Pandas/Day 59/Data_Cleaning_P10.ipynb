# Cleaning 

```python
df.dropna(subset='Customer_ID',inplace=True)
```

```python
df.drop_duplicates(subset='Customer_ID',keep='first',inplace=True)
```

```python
df['Full_Name']=df['Full_Name'].fillna('Unknown')
```

```python
df['Full_Name']=df['Full_Name'].str.strip().str.title()
```

```python
mask = df['Email'].str.contains(
    r'^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$',
    regex=True,
    na=False
)
```

```python
df.loc[~mask, 'Email'] = np.nan
```

```python
df['Purchase_Date'].unique()[:23]
```

```python
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'],format = 'mixed',yearfirst=True,errors='coerce')
```

```python
df['Purchase_Date'].isnull().sum()
```

```python
df['Amount_Spent'].unique()[:20]
```

```python
df['Amount_Spent'] = pd.to_numeric(df['Amount_Spent'],errors='coerce')
```

```python
df.loc[df['Amount_Spent']<0,'Amount_Spent'] = np.nan
```

```python
len(df['City'].unique())
```

```python
len(df['City'].str.strip().str.title().unique())
```

```python
df['City'].str.strip().str.title().unique()
```

```python
city_map = {
    'Albuquerqe': 'Albuquerque',
    'Dalas': 'Dallas',
    'Indinapolis': 'Indianapolis',
    'Denvr': 'Denver',
    'San Anonio': 'San Antonio',
    'Sanantonio': 'San Antonio',
    'Scramento': 'Sacramento',
    'Oakand': 'Oakland',
    'Ls Angeles': 'Los Angeles'
}
```

```python
df['City'] = df['City'].replace(city_map)
```

```python
df['Amount_Spent'] = df['Amount_Spent'].fillna(df.groupby('City')['Amount_Spent'].transform(lambda a : a.median()))
```

```python
df['City'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['City'] = df['City'].fillna('Unknown')
```

```python
df['Purchase_Date'] = df['Amount_Spent'].fillna(df.groupby('City')['Purchase_Date'].transform(lambda a : a.mode()[0] if not a.mode().empty else np.nan))
```

```python
df.isnull().sum()
```

```python
df['Email'] = df['Email'].fillna('not available')
```

```python
df.dropna(subset='Amount_Spent',inplace=True)
```
