# Pandas pivot_table()

```python
import pandas as pd
```

```python
data = {
    'city': ['Delhi','Delhi','Mumbai','Mumbai','Pune','Pune','Delhi','Mumbai'],
    'product': ['Laptop','Mobile','Laptop','Mobile','Laptop','Mobile','Tablet','Tablet'],
    'sales': [50000,20000,45000,25000,40000,18000,15000,22000],
    'quantity': [5,10,4,12,3,8,6,5],  
    'payment_method': ['UPI','Card','Cash','UPI','Card','Cash','UPI','Card']
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.info()
```

### 1. Create pivot table using only sales column.

```python
df.pivot_table(values='sales', index='city')
```

### 2. Create pivot table using only quantity column.

```python
df.pivot_table(values='quantity',index='product',aggfunc='sum')
```

```python
df.pivot_table(values='quantity',index='city',aggfunc='sum')
```

### 3. Create pivot table using both sales and quantity.

```python
df.pivot_table( values=['sales','quantity'], index='city')
```

### 4.  Create pivot table with city as index.

```python
df.head()
```

```python
df.pivot_table(index='city', values = 'quantity',aggfunc='count')
```

```python
df.pivot_table(index='city', values = 'quantity',columns='product',aggfunc='count')
```

### 5. Create pivot table with product as index.

```python
df.pivot_table(index='product', values = 'sales')
```

```python
df.pivot_table(index='product', values = 'sales',columns='city')
```

### 6.  Create pivot table with : city , payment_method as multi-index.

```python
df.pivot_table(values='sales',index=['city','payment_method'])
```

### 7.  Create pivot table with: city, product, payment_method as multi-index.

```python
df.pivot_table( values='sales',   index=['city','product','payment_method'])
```

### 8. Create pivot table with:  product as columns.

```python
df.pivot_table(values='sales', columns='product')
```

### 9. Create pivot table with: payment_method  as columns.

```python
df.pivot_table( values='sales',columns='payment_method')
```

### 10. Create pivot table with: product, payment_method  as multi-columns.

```python
df.pivot_table( values='sales', columns=['product','payment_method'])
```

### 11. Create pivot table using: aggfunc='sum'

```python
df.pivot_table(values='sales',index='city',aggfunc='sum')
```

```python
df.pivot_table(values='sales',index='product',aggfunc='sum')
```

### 12. Create pivot table using:aggfunc='mean'

```python
df.pivot_table(values='sales',index='city',aggfunc='mean')
```

```python
df.pivot_table(values='sales',index='payment_method',aggfunc='mean')
```

### 13. Create pivot table using: aggfunc='max'

```python
df.pivot_table(values='sales',index='payment_method',aggfunc='max')
```

### 14. Create pivot table using: aggfunc='min'

```python
df.pivot_table(values='sales',index='payment_method',aggfunc='min')
```

```python
df.pivot_table(values='sales',index='city',aggfunc='min')
```

### 15. Create pivot table using:aggfunc='count'

```python
df.pivot_table(values='sales',index='city',aggfunc='count')
```

### 16. Create pivot table using: aggfunc=['sum','mean']

```python
df.pivot_table(values='sales',index='city',aggfunc=['sum','mean'])
```

### 17. Create pivot table and replace all NaN with 0.

```python
df.pivot_table(index='city', values = 'quantity',columns='product',aggfunc='count')
```

```python
df.pivot_table(index='city', values = 'quantity',columns='product',aggfunc='count',fill_value=0)
```

```python
df.pivot_table(values='sales',index='city',columns='product')
```

```python
df.pivot_table(values='sales',index='city',columns='product',fill_value=0)
```

### 18. Create pivot table: 
#### city as index
#### product as columns
#### fill missing values with 100

```python
df.pivot_table(values='sales',index='city',columns='product',fill_value=100)
```

### 19. Create pivot table with grand totals.

```python
df.pivot_table(values='sales',index='city',margins=True)
```

### 20. Create pivot table and rename total column to: 'Grand Total'

```python
df.pivot_table(values='sales',index='city', columns='product', margins=True, margins_name='Grand Total')
```

### 21. Create pivot table and rename totals to:'Final Report'

```python
df.pivot_table(
     values='sales',
    index='city',
    columns='product',
    margins=True,
    margins_name='Final Report'
)
```

### 22. Create pivot table using:
aggfunc=['sum','mean','max']

```python
df.pivot_table( values='sales',
     index='city',
    aggfunc=['sum','mean','max']
)
```

### 23. Create pivot table with:
multi-index
multi-columns

```python
df.pivot_table(
    values='sales',
    index=['city','product','payment_method'],
    columns=['product','payment_method'],
    aggfunc='sum'
)
```
