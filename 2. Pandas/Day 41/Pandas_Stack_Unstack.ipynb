## Pandas Stack

```python
import pandas as pd
```

```python
df1.stack()
```

## Dataset 2

```python
data2 = {
    'Product': ['Laptop','Mouse','Keyboard'],
    'Jan': [50000,12000,15000],
    'Feb': [52000,14000,16000],
    'Mar': [51000,13000,17000]
}

sales_df = pd.DataFrame(data2)
sales_df
```

### 1. Set Product as index and apply stack().

```python
sales_df.reset_index('Product').stack()
```

```python
sales_df.set_index('Product').stack()
```

### 2. Convert month columns into rows.

```python
sales_df.set_index('Product').stack().reset_index()
```

### 3. Store stacked output in sales_stack.

```python
sales_stack = sales_df.set_index('Product').stack()
sales_stack
```

### 4. Reset index after stacking.

```python
sales_stack.reset_index()
```

### 5. Rename columns:
level_1 → Month
0 → Revenue

```python
sales_stack.reset_index().rename(columns={'level_1': 'Month',0: 'Revenue'})
```

### 6. Find Laptop sales after stacking.

```python
sales_stack['Laptop']
```

### 7. Access February revenue only.

```python
sales_stack[:,'Feb']
```

```python
sales_stack.xs('Feb', level=1)
```

### 8. Sort stacked revenue values.

```python
sales_stack.sort_values()
```

### 9. Convert stacked output back using unstack().

```python
sales_stack.unstack()
```

### 10. Compare original and unstacked output.

```python
sales_df.set_index('Product').equals(sales_stack.unstack())
```

## Dataset 3

```python
data3 = {
    'Website': ['Amazon','Flipkart','Myntra'],
    'Monday': [1200,900,700],
    'Tuesday': [1300,850,750],
    'Wednesday': [1250,920,780]
}

traffic_df = pd.DataFrame(data3)
traffic_df
```

### 1. Set Website as index and stack weekdays.

```python
traffic_df.set_index('Website').stack()
```

### 2. Convert traffic data into long format using stack().

```python
traffic_df.set_index('Website').stack().reset_index()
```

### 3. Reset index after stacking.

```python
traffic_df.set_index('Website').stack().reset_index()
```

### 4. Rename columns:

level_1 → Day
0 → Visitors

```python
traffic_df.set_index('Website').stack().reset_index().rename(
    columns={'level_1': 'Day',0: 'Visitors'})
```

### 5.  Find Amazon Tuesday visitors.

```python
traffic_df.set_index('Website').stack()['Amazon', 'Tuesday']
```

### 6.   Sort visitors in descending order.

```python
traffic_df.set_index('Website').stack().sort_values(ascending=False)
```

### 7. Convert stacked data back using unstack().

```python
stack_data = traffic_df.set_index('Website').stack()
stack_data
```

```python
stack_data.unstack()
```

### 9. Access only Wednesday data after stacking.

```python
traffic_df.set_index('Website').stack().xs('Wednesday', level=1)
```

### 10. Convert stacked Series into DataFrame.

```python
traffic_df.set_index('Website').stack().xs('Wednesday', level=1)
```

## Unstack Practice Dataset

```python
data = {
    'Student': ['Amit','Amit','Neha','Neha'],
    'Subject': ['Math','Science','Math','Science'],
    'Marks': [78,82,85,88]
}
```

```python
long_df = pd.DataFrame(data)
```

```python
long_df = long_df.set_index(['Student','Subject'])
```

```python
print(long_df)
```

### 1. Apply unstack() on Subject level.

```python
long_df.unstack(level='Subject')
```

### 2. Convert Subject rows into columns.

```python
long_df.unstack(level=1)
```

```python
long_df.unstack(level='Subject')
```

### 3. Store unstacked output into wide_df.

```python
wide_df = long_df.unstack()
print(wide_df)
```

### 4. Reset index after unstacking.

```python
wide_df.reset_index()
```

### 5. Access Amit's Math marks after unstacking.

```python
wide_df.loc['Amit', ('Marks', 'Math')]
```

### 6. Fill missing values using fill_value.

```python
long_df.unstack(fill_value=0)
```

### 7. Apply stack() again on unstacked data.

```python
wide_df.stack()
```

### 8. Compare original and reconstructed data.

```python
long_df.equals(wide_df.stack())
```

### 9. Use unstack(level=0).

```python
long_df.unstack(level=0)
```
