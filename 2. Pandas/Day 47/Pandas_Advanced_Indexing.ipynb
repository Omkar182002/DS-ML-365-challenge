# 4. Index Alignment

```python
s1 = pd.Series([100, 200, 300], index=['A', 'B', 'C'])
```

```python
s1
```

```python
s2 = pd.Series([10, 20, 30], index=['B', 'C', 'D'])
```

```python
s2
```

### 1. Add two Series objects using index alignment.

```python
s1 + s2
```

### 2. Identify positions containing NaN after alignment.

```python
(s1 + s2).isna()
```

### 3. Add two Series and replace missing values with 0.

```python
s1.add(s2, fill_value=0)
```

```python
(s1+s2).fillna(0)
```

### 4. Subtract second Series from first Series.

```python
s1 - s2
```

### 5. Multiply two aligned Series.

```python
s1 * s2
```

### 6. Divide first Series by second Series.

```python
(s1 / s2)
```

### 7. Add two DataFrames with different indexes.

```python
df1 = pd.DataFrame({'A': [1, 2]}, index=['X', 'Y'])
```

```python
df2 = pd.DataFrame({'A': [3, 4]}, index=['Y', 'Z'])
```

```python
df1+df2
```

### 8. Add two DataFrames and fill missing aligned values with 0.

```python
df1.add(df2, fill_value=0)
```

```python
(df1+df2).fillna(0)
```

### 9. Check whether indexes of both Series are equal.

```python
s1.index.equals(s2.index)
```

### 10. Reorder Series index using reindexing.

```python
s1.reindex(['C', 'B', 'A'])
```

# 5. Reindexing

```python
df3 = pd.DataFrame({
    'Sales': [100, 200, 300]
}, index=['A', 'B', 'C'])
```

```python
df3
```

### 1. Reindex the DataFrame by adding a new row D.

```python
df3.reindex(['A', 'B', 'C', 'D'])
```

### 2. Reindex the DataFrame and fill missing values with 0.

```python
df3.reindex(['A', 'B', 'C', 'D'], fill_value=0)
```

### 3. Change the order of indexes to C, A, B.

```python
df3.reindex(['C', 'A', 'B'])
```

### 4. Reindex columns by adding a Profit column.

```python
df3.reindex(columns=['Sales', 'Profit'])
```

### 5. Apply forward fill while reindexing.

```python
df3.reindex(['A', 'B', 'C', 'D'], method='ffill')
```

### 6. Apply backward fill while reindexing.

```python
df3.reindex(['A', 'B', 'C', 'D'], method='bfill')
```

### 7. Add multiple new indexes D and E using reindex.

```python
df3.reindex(['A', 'B', 'C', 'D', 'E'])
```

### 8. Detect rows containing missing values after reindexing.

```python
new_df = df3.reindex(['A', 'B', 'C', 'D'])
```

```python
new_df[new_df['Sales'].isna()]
```

### 9. Replace old index labels with new labels X, Y, Z.

```python
df3.set_axis(['X', 'Y', 'Z'])
```

### 10. Reindex DataFrame and replace missing values using fillna().

```python
df3.reindex(['A', 'B', 'C', 'D']).fillna(999)
```
