# -----------------------METHOD: .merge()-----------------------

#### 1. Merge customers and orders using common column customer_id.

```python
pd.merge(customers, orders, on='customer_id')
```

#### 2. Show only matching customer records who placed orders.

```python
pd.merge(customers, orders, on='customer_id',how='inner')
```

#### 3. Show all customers even if they never ordered.

```python
pd.merge(customers, orders, on='customer_id',how='left')
```

#### 4. Show all orders even if customer details are missing.

```python
pd.merge(customers,orders,on='customer_id',how='right')
```

#### 5. Show all customers and all orders including unmatched data.

```python
pd.merge(customers,orders,on='customer_id',how='outer')
```

# -----------------------METHOD: DataFrame.join()-----------------------

```python
employees = pd.DataFrame({'emp_id':[1,2,3,4],'name':['Om','Raj','Neha','Amit']}).set_index('emp_id')
```

```python
employees
```

```python
salaries = pd.DataFrame({'salary':[50000,60000,70000]}, index=[1,2,5])
```

```python
salaries
```

# 

#### 1. Join employees and salaries.

```python
employees.join(salaries)
```

#### 2. Show only matching indexes.

```python
employees.join(salaries,how='inner')
```

#### 3. Keep all employee records.

```python
employees.join(salaries,how='left')
```

#### 4. Keep all salary records.

```python
employees.join(salaries,how='right')
```

#### 5. Keep all indexes from both DataFrames.

```python
employees.join(salaries,how='outer')
```

#### 

#### 6. Join using column instead of index.

```python
departments = pd.DataFrame({
    'emp_id':[1,2,3],
    'dept':['HR','IT','Sales']
})
```

```python
departments
```

```python
employees.join(departments.set_index('emp_id'),on='emp_id')
```

# 

#### 7. Join employees + salaries + departments.

```python
employees.join(salaries).join(departments.set_index('emp_id'))
```

#### 8. Reset Index After Join

```python
employees.join(salaries)
```

```python
employees.join(salaries).reset_index()
```

## join() — Overlapping Columns

```python
emp = pd.DataFrame({
    'emp_id':[1,2,3,4],
    'name':['Om','Raj','Neha','Amit'],
    'salary':[50000,60000,70000,80000]
}).set_index('emp_id')
```

```python
emp
```

```python
bonus1 = pd.DataFrame({
    'salary':[5000,7000,9000],
    'bonus':[1000,2000,3000]
}, index=[1,2,5])
```

```python
bonus1
```

#### 1 . OVERLAPPING COLUMN ERROR
Join employees and bonus directly.

```python
emp.join(bonus1)
```

#### 2. HANDLE OVERLAP USING SUFFIXES

```python
emp.join(bonus1,lsuffix='_emp',rsuffix='_bonus')
```

#### 3. INNER JOIN + OVERLAPPING COLUMNS
Show only matching indexes.

```python
emp.join(bonus1,how='inner',lsuffix='_emp',rsuffix='_bonus')
```

#### 4 . LEFT JOIN + OVERLAPPING
Keep all employee records.

```python
emp.join(bonus1,how='left',lsuffix='_emp',rsuffix='_bonus')
```

#### 5 . RIGHT JOIN + OVERLAPPING
Keep all bonus records.

```python
emp.join(bonus1,how='right',lsuffix='_emp',rsuffix='_bonus')
```

#### 6. SELECT ONLY NEEDED COLUMNS
Join only bonus column.

```python
emp.join(bonus1[['bonus']])
```

#### 7. RENAME COLUMN BEFORE JOIn
Rename salary column before join.

```python
bonus3 = bonus1.rename(columns={'salary':'bonus_salary'})
emp.join(bonus3)
```
