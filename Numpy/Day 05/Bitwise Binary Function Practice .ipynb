## Bitwise Binary Function Practice 

```python
import numpy as np
```

##### 1.	Perform np.bitwise_and() on a = [1, 2, 3], b = [3, 2, 1].

```python
a1 = np.array([1,2,3])
b1 = np.array([3,2,1])
print(a1&b1)
print(np.bitwise_and(a1,b1))
```

##### 2.	Perform np.bitwise_or() on a = [5, 10, 15], b = [1, 2, 3].

```python
a2 = np.array([5,10,15])
b2 = np.array([1,2,3])
print(a2|b2)
print(np.bitwise_or(a2,b2))  
```

##### 3.	Use np.bitwise_xor() on a = [12, 25, 37], b = [15, 30, 42].

```python
a3 = np.array([12,25,37])
b3 = np.array([15,30,42])
print(a3^b3)
print(np.bitwise_xor(a3,b3))
```

#### 4.	Compare the result of np.bitwise_and(6, 3) manually and verify with code.

```python
print(np.bitwise_and(6, 3) )
print(6&3)
```

##### 5.	Perform bitwise AND between two boolean arrays:
a = [True, False, True], b = [False, False, True].


```python
a4 = np.array([True, False, True])
b4 = np.array([False, False, True])
print(a4&b4)
print(np.bitwise_and(a4,b4))
```
