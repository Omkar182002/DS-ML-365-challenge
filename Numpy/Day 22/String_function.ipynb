## String Function

```python
import numpy as np
```

```python
st = "Hello boss"
```

```python
print(type(st))
```

```python
np.char.lower(st)
```

```python
np.char.upper(st)
```

```python
np.char.capitalize(st)
```

```python
np.char.title(st)
```

```python
np.char.join(st,"HI")
```

```python
np.char.split(st)
```

```python
np.char.splitlines(st)
```

```python
np.char.strip(st)
```

```python
np.char.lstrip(st)
```

```python
np.char.startswith(st,'a')
```

```python
np.char.startswith(st,"H")
```

```python
np.char.endswith(st, 'b')
```

```python
np.char.endswith(st, 's')
```

```python
st2 = np.array(['hello', 'numpy'])
print(np.char.str_len(st2))
```
