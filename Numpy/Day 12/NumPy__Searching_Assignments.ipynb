# NumPy Searching Functions – Simple Assignments


### 1: np.where()
Returns indices where condition is True.

```python
import numpy as np
```

```python
scores = np.array([85, 42, 97, 63, 78, 51, 89, 34, 76, 91, 42, 88, 42, 95, 67])
indeces = np.arange(15)
print("Scores:", scores)
print("Indices:", indeces)
```

#### Q1. Find all indices where score is exactly 42 using np.where().

```python
result_index = np.where(scores==42,indeces,np.nan)
result_index
```

```python
result_index[~np.isnan(result_index)]
```

```python
np.where(scores==42)
```

#### Q2. Find all indices where score is greater than 80.

```python
result_index2 = np.where(scores>80,indeces,np.nan)
result_index2[~np.isnan(result_index2)]
```

```python
np.where(scores>80)
```

#### Q3. Find all indices where score is between 70 and 90 (inclusive)

```python
condition = np.logical_and(scores>70 ,scores<90)
re_index3 = np.where(condition,indeces,np.nan)
```

```python
re_index3[~np.isnan(re_index3)]
```

```python
np.where((scores>70) &(scores<90))
```

#### Q4. Find all indices where score is less than 50 or greater than 90.

```python
condition1 = np.logical_or(scores<50 ,scores>90)
re_index4 = np.where(condition1,indeces,np.nan)
```

```python
re_index4[~np.isnan(re_index4)]
```

```python
np.where((scores<50)|(scores>90))
```

#### Q5. Replace all scores below 60 with 60 using np.where().

```python
scores2 = scores.copy()
```

```python
np.where(scores2<60,60,scores2)
```

### 5: np.extract()
Extracts elements that satisfy a condition.



```python
expenses = np.array([120, 45, 200, 80, 350, 65, 180, 95, 420, 150, 75, 290, 110, 55, 400])
print("Expenses:", expenses)
```

#### Q21. Extract all expenses greater than 200 using np.extract().

```python
np.extract(expenses>200,expenses)
```

#### Q22. Extract all expenses between 100 and 250 (inclusive).

```python
np.extract(((expenses>100 ) & (expenses<250)),expenses)
```

#### 23. Extract all expenses that are even numbers.

```python
np.extract(expenses%2== 0,expenses)
```

#### Q24. Extract expenses that are less than 50 or greater than 300.

```python
np.extract(((expenses<50 )  |(expenses>300)),expenses)
```

#### Q25. Extract expenses and their indices using np.where() (not extract).

```python
indices = np.where(expenses>200)
print("Indices  : ",indices)
print("Expenses : ",expenses[indices])
```

###  REAL-WORLD SCENARIO: Customer Data Search

```python
# 1000 customers with [CustomerID, Age, PurchaseAmount, Satisfaction_1-5]
np.random.seed(42)
customers = np.random.randint(1000, 5000, size=(1000, 4))
customers[:, 1] = np.random.randint(18, 80, 1000)  # Age
customers[:, 2] = np.random.randint(10, 1000, 1000)  # Purchase amount
customers[:, 3] = np.random.randint(1, 6, 1000)  # Satisfaction

print("Customers shape:", customers.shape)
print("First 5 customers [ID, Age, Purchase, Satisfaction]:\n", customers[:5])
```

#### Q31. Find all customers with satisfaction score of 5 using np.where().

```python
satisfy_cust = np.where(customers[:,3]==5) # return row indices
satisfy_cust
```

```python
customers[17,3]
```

```python
customers[satisfy_cust,3]
```

#### Q32. Find customers aged 25-35 with purchase amount > 500.

```python
cond1 = ((customers[:,1]>25 )& (customers[:,1]<35))
cond2 = customers[:,2]>500
result = np.where(np.logical_and(cond1,cond2)) #return row indices
result
```

```python
customers[result,0]# customers id 
```

#### Q33. Find the customer with maximum purchase amount (return full row).

```python
max_cust  =  np.argmax(customers[:,2])
```

```python
print(customers[max_cust,:])
```

#### Q34. Find the youngest customer (minimum age) using np.argmin().

```python
min_age = np.argmin(customers[:,1])
```

```python
print(customers[min_age,:])
```

#### Q35. Find how many customers have age > 60 and purchase < 200.
