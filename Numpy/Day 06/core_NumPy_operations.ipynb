### 3. LOGICAL OPERATIONS

#### 1. Check (a > 10) & (b > 10)

```python
(a>10)&(b>10)
```

```python
np.logical_and(a>b,b>10)
```

#### 2. Check (a > 20) | (b > 20)

```python
np.logical_or(a>20,b>20)
```

```python
(a>20)|(b>20)
```

#### 3. Check ~(a > 20)

```python
np.logical_not(a>20)
```

```python
~(a>30)
```

#### 4. Find values where a > 15 AND b < 20

```python
np.select([a>15,b<20],[a,b])
```

```python
result = a[(a > 15) & (b < 20)]
print(result)
```

```python
a
```

```python
b
```

#### 5. Find values where a > 20 OR b > 20

```python
result = a[(a > 20) | (b > 20)]
print(result)
```

#### 6. Find values where NOT (a < 20)

```python
a[~(a<20)]
```

```python
a[np.logical_not(a<20)]
```

#### 7. Combine condition: (a > 15) & (b == 15)

```python
np.logical_and(a>15,b==5)
```

#### 8. Filter values where both arrays are equal

```python
a[a==b]
```

```python
b[a==b]
```

#### 9. Filter values where either is > 25

```python
a[(a>25)|(b>25)]
```

#### 10. Find elements where neither condition is true

```python
a[~((a>25)|(b>25))]
```

```python
result = a[~((a > 25) | (b > 25))]
print(result)
```

#### 11. Count how many satisfy a > 20 & b < 20

```python
np.logical_and(a>20,b<20).sum()
```

```python
np.sum((a > 20) & (b < 20))
```

#### 12. Replace values where (a < 20) with 0

```python
np.where(a<20,0,a)
```

#### 13.Replace values where (b > 20) with -1

```python
np.where(b>20,-1,a)
```

#### 14. Create boolean mask for a in range 10–25

```python
mask = ((a>=10)&(a<=25))
mask
```

#### 15. Create mask for b not equal to 15

```python
mask = b!=15
mask
```

#### 16. Find valid entries where both > 10

```python
result = a[(a > 10) & (b > 10)]
print(result)
```
