##### Boolean Indexing

```python
a7 = np.array([10,11,15,18,17])
a7[a7>15]
```

```python
marks = np.array([60,98,69,84,49,51])
marks[marks>55]
```

```python
##### multiple condition
```

```python
np.random.seed(12)
age = np.random.randint(1,100,20)
adults = age[(age>30)&(age<50)]
adults
```

```python
a8 = np.array([1,2,3,4,2,4,5,6])
ar= a8[(a8==2) |(a8==4)]
ar.size
```

##### np.where()

```python
a9 = np.array([10,20,30,40])
b9 = np.where(a9>20,1,0)
b9
```

```python
c9 = np.argwhere(a9>10) #return index
c9
```

```python
np.where(a9 > 10)
```

##### np.ix_() imp

```python
a10 = np.arange(20).reshape(5,4)
a10
```

```python
a10[np.ix_(1)]
```

```python
a10[np.ix_([1])]
```

```python
a10[np.ix_([1:])]
```

```python
a10[np.ix_([1,2])]
```

```python
# without np.ix_()
a10[[1,2]]
```

```python
a10[np.ix_([1,2],[2,3])]
```

```python
# without np.ix_()
a10[[1,2],[2,3]]
```

#####  now np.take() - same like fancy indexing but fatser + flexible
