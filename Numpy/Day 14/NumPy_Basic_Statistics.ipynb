## REAL-WORLD SCENARIO 1: Student Performance

```python
np.random.seed(42)
# 30 students, 5 subjects
students = np.random.randint(50, 100, size=(30, 5))
subjects = ["Math", "Science", "English", "History", "Art"]
print("Students shape:", students.shape)
print("First 3 students:\n", students[:3])
```

#### Q41. Calculate mean score for each student.

```python
print(np.mean(students,axis=1))
```

#### Q42. Calculate mean score for each subject.

```python
sub_mean = np.mean(students,axis=0)
for s, m in zip(subjects,sub_mean):
    print(f"{s} : {m:.2f}")
```

#### Q43. Find student with highest average.

```python
student_mean = np.mean(students,axis=1)
print("Student Index No : ",np.argmax(student_mean))
print("mean : ",np.max(student_mean))
```

#### Q44. Find subject with lowest average.

```python
sub_mean
```

```python
low_sub_ind = np.argmin(sub_mean)
print("Subject : ",subjects[low_sub_ind])
print("Mean :  ",np.min(np.round(sub_mean,2)))
```

#### Q45. Calculate standard deviation for each subject (which subject has most variation?).

```python
std_dev = np.std(students,axis=0)
max_var_sub = subjects[np.argmax(std_dev)]
print("Standard Deviation : ",std_dev)
print("Max variation sub  : ",max_var_sub)
```

### 🔹 REAL-WORLD SCENARIO 2: Sales Data


```python
# 12 months, 4 products
sales = np.array([
    [120, 145, 130, 155],  
    [85, 92, 88, 105],     
    [200, 210, 105, 130],  
    [45, 52, 48, 60],      
    [160, 175, 168, 190],  
    [95, 88, 102, 110],    
    [180, 195, 188, 210],  
    [70, 78, 72, 85],      
    [150, 160, 155, 175],  
    [110, 118, 115, 125],  
    [230, 145, 138, 160],  
    [65, 70, 68, 75]       
])
months = np.array([ "Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
print("Sales shape (12 months × 4 products):", sales.shape)
print("First 3 months of sales:\n", sales[:3])
```

```python
months
```

#### Q46. Calculate total sales per month (sum across products).

```python
total_sales = np.sum(sales , axis=1)
for m , s in zip(months,total_sales):
    print(f"{m} : {s}")
```

#### Q47. Calculate average sales per product (across 12 months).

```python
print(np.mean(sales,axis=0))
```

#### Q48. Find best month for each product.

```python
bst_m = np.argmax(sales,axis=0)
print("Best month for peoduct : ")
product = ["Product A","Product B","Product C","Product D"]
for p, m in zip(product,bst_m):
    print(f"{p} : {months[m]}")
```
