## 1. Seaborn Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("tips")
data.head()
```

```python
data.info()
```

# 

### Basic Histogram : 

```python
sns.histplot(x=data.total_bill)
plt.show()
```

```python
sns.histplot(data= data,x='total_bill')
plt.show()
```

#### bins parameter:

```python
sns.histplot(data= data,x='total_bill', bins=5)
plt.show()
```

```python
sns.histplot(data= data,x='total_bill', bins=[10,25,50])
plt.show()
```

#### hue parameter :

```python
data.head(2)
```

```python
sns.histplot(data= data,x='total_bill', bins=5,hue="sex")
plt.show()
```

```python
sns.histplot(data= data,x='total_bill', bins=5,hue="sex",hue_order=['Female','Male'] ,palette="Set2")
plt.show()
```

```python
sns.histplot(data= data,x='total_bill', bins=5,hue='smoker')
plt.show()
```

```python
sns.histplot(data= data,x='total_bill', bins=5,hue='smoker', hue_order=['No','Yes'])
plt.show()
```

```python
sns.histplot(data= data,x='total_bill', bins=5,  multiple="layer")
plt.show()
```

```python
sns.histplot(data= data,x='total_bill', bins=5, hue="sex", multiple="dodge")
plt.show()
```

```python
sns.histplot(data= data,x='total_bill', bins=5, hue="sex", multiple="fill")
plt.show()
```

#### element parameter ;

```python
sns.histplot(data=data,x="total_bill",element="bars")
plt.show()
```

## 3. Seaborn Heatmap

```python
corr = tips.corr(numeric_only=True)
corr
```

```python
import numpy as np 
```

#### Basic Heatmap :

```python
data1 = np.linspace(1,10,20).reshape(4,5)
```

```python
data1
```

```python
sns.heatmap(data1)
plt.show()
```

#### annot parameter : 

```python
sns.heatmap(data1,annot=True)
plt.show()
```

#### cmap parameter : 

```python
sns.heatmap(data1,annot=True, cmap="coolwarm")
plt.show()
```

```python
np.random.seed(42)
data2 = np.random.randint(1,10,(4,5))
```

```python
data2
```

```python
sns.heatmap(data2,annot=True, cmap="coolwarm")
plt.show()
```

#### linewidths paramters  : 

```python
sns.heatmap(data2,annot=True, cmap="coolwarm", linewidths=3)
plt.show()
```

```python
sns.heatmap(data2,annot=True, cmap="coolwarm", linewidths=0.5)
plt.show()
```

#### linecolor paramters ; 

```python
sns.heatmap(data2,annot=True, cmap="coolwarm", linewidths=1, linecolor='blue')
plt.show()
```

```python
sns.heatmap(data2,annot=True, cmap="coolwarm", linewidths=0.2, linecolor='k')
plt.show()
```

#### fmt paramter : 

```python
sns.heatmap(data1,annot=True, cmap="coolwarm", linewidths=0.2, fmt='.2f')
plt.show()
```

```python
sns.heatmap(data2,annot=True, cmap="coolwarm", linewidths=0.2, fmt='.2f')
plt.show()
```

#### vmin and vmax parameters : 

```python
sns.heatmap(data1,annot=True, cmap="coolwarm", vmin=1,vmax=6)
plt.show()
```

```python
sns.heatmap(data1,annot=True, cmap="coolwarm", vmin=0,vmax=20)
plt.show()
```

#### square paramter : 

```python
sns.heatmap(data1,annot=True, cmap="coolwarm", square=True)
plt.show()
```

```python
sns.heatmap(data1,annot=True, cmap="coolwarm", square=False)
plt.show()
```

#### cbar paramter : 

```python
sns.heatmap(data1,annot=True, cmap="coolwarm",cbar=False)
plt.show()
```
