###### Visualzation : 

#### 1. How are Math Scores distributed?

```python
plt.figure(figsize=(8,5))
sns.histplot(data=data,x="math score",bins=20,kde=True)
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Number of Students")
plt.show()
```

###### Observation
- Most students scored between 60 and 80.
- Very few students scored below 30.

#### 2. How are Reading Scores distributed?

```python
plt.figure(figsize=(8,5))
sns.histplot(data=data,x="reading score",bins=20,kde=True)
plt.title("Distribution of Reading Scores")
plt.xlabel("Reading Score")
plt.ylabel("Number of Students")
plt.show()
```

###### Observation : Most students scored between 55 and 85.

#### 3. How are Writing Scores distributed?

```python
plt.figure(figsize=(8,5))
sns.histplot(data=data,x="writing score",bins=20,kde=True)
plt.title("Distribution of Writing Scores")
plt.xlabel("Writing Score")
plt.ylabel("Number of Students")
plt.show()
```

###### Observation
- Most students scored between 50 and 85.
- Very few students achieved extremely low or high scores.

#### 4. Are there outliers in Math Scores?

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=data,y="math score")
plt.title("Math Score Distribution")
plt.show()
```

###### Observation :  
- A few students have low math scores .
- No major high-score outliers are observed.

#### 5. Are there outliers in Reading Scores?

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=data,y="reading score")
plt.title("Reading Score Distribution")
plt.show()
```

###### Observation
- Reading scores have a few low-score.
- Most students perform within a consistent range.

#### 6. Are there outliers in Writing Scores?

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=data,y="writing score")
plt.title("Writing Score Distribution")
plt.show()
```

###### Observation

- Most scores fall within the expected range.

#### 7. How many Male and Female Students are there?

#### 10. Which Lunch Type is most common?

```python
plt.figure(figsize=(6,4))
sns.countplot(data=data,x="lunch")
plt.title("Lunch Type Distribution")
plt.xlabel("Lunch Type")
plt.ylabel("Number of Students")
plt.show()
```

#### 11. Does Test Preparation improve Math Scores?

```python
data.columns
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=data,x="test preparation course",y="math score")
plt.title("Math Score by Test Preparation")
plt.xlabel("Test Preparation")
plt.ylabel("Math Score")
plt.show()
```

- Students who completed the test preparation course generally have higher math scores.
- Students who did not complete the course show lower median scores.

#### 12. Does Test Preparation improve Reading Scores?

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=data,x="test preparation course",y="reading score")
plt.title("Reading Score by Test Preparation")
plt.xlabel("Test Preparation")
plt.ylabel("Reading Score")

plt.show()
```

 - Students who completed the course generally achieved higher reading scores.
- The median reading score is higher for students who completed the preparation course.

#### 13 . Does Test Preparation improve Writing Scores?

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=data,x="test preparation course", y="writing score")
plt.title("Writing Score by Test Preparation")
plt.xlabel("Test Preparation")
plt.ylabel("Writing Score")
plt.show()
```

 - Students who completed the preparation course have higher writing scores.
- Students who did not complete the course show greater variation in writing scores.
