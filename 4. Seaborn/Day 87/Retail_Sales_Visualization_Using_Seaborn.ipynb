### 1. What is the distribution of Sales?
- Are most orders low-value or high-value?

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df,x="Sales",bins=10)
plt.title("Sales Distribution")
plt.xticks([1000, 3000,5000,  7000,  9000, 12000])
plt.show()
```

#### 2. Are there any outliers in Sales?
- Are there unusually large orders?

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df,y="Sales")
plt.title("Sales Outliers")
plt.show()
```

#### 3. Which Product Category has the highest Average Sales?
- Which category generates the highest average sales per order?

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df,x="Category",y="Sales",errorbar=None)
plt.title("Average Sales by Category")
plt.show()
```

#### 4. Which Category receives the highest number of Orders?
- Which category is purchased most frequently?

```python
plt.figure(figsize=(8,5))
sns.countplot( data=df,x="Category")
plt.title("Number of Orders by Category")
plt.show()
```

#### 13. Which State has the highest Sales?
- Which states contribute the highest sales?

```python
top_states = (df.groupby("State", as_index=False)["Sales"].sum().sort_values("Sales", ascending=False).head(10))
```

```python
plt.figure(figsize=(12,5))
sns.barplot(data=top_states,x="State",y="Sales")
plt.title("Top 10 States by Sales")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(data=top_states,y="State",x="Sales",orient='h')
plt.title("Top 10 States by Sales")
plt.show()
```

#### 14. Which Sub-Category earns the highest Profit?
- Which products are the most profitable?

```python
plt.figure(figsize=(12,5))
sns.barplot(data=df,x="Sub-Category",y="Profit",errorbar=None)
plt.xticks(rotation=60)
plt.title("Profit by Sub-Category")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(data=df,y="Sub-Category",x="Profit",errorbar=None, orient='h')
plt.xticks(rotation=60)
plt.title("Profit by Sub-Category")
plt.show()
```

#### 15. How do Monthly Sales change over time?
- Is sales increasing or decreasing month by month?

```python
monthly_sales = (df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum().reset_index())
```

```python
monthly_sales
```

```python
monthly_sales["Order Date"] = monthly_sales["Order Date"].astype(str)
```

```python
plt.figure(figsize=(14,5))
sns.lineplot(data=monthly_sales,x="Order Date",y="Sales",marker="o")
plt.xticks(rotation=90)
plt.title("Monthly Sales Trend")
plt.show()
```
