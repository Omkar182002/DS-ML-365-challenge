###### Seaborn :
- catplot() 
- relplot() 

```python
import  pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({
    "Customer_ID":[101,102,103,104,105,106,107,108,109,110],
    "Category":["Electronics","Furniture","Clothing","Electronics","Furniture",
                "Clothing","Electronics","Furniture","Clothing","Electronics"],
    "Region":["North","South","East","West","North",
              "South","East","West","North","South"],
    "Gender":["Male","Female","Female","Male","Female",
              "Male","Female","Male","Female","Male"],
    "Customer_Type":["New","Returning","Returning","New","New",
                     "Returning","New","Returning","New","Returning"],
    "Sales":[500,800,300,700,900,450,650,750,350,1000],
    "Profit":[100,120,50,150,180,80,130,140,60,220],
    "Discount":[10,20,5,15,25,10,12,18,8,30],
    "Quantity":[2,4,1,3,5,2,3,4,1,6],
    "Returned":["No","Yes","No","No","Yes","No","No","Yes","No","Yes"]
})
```

```python
df.head()
```

###### Business Questions : 

#### 1. Which Product Category Has the Highest Average Sales?
The Sales Manager wants to know which product category generates the highest revenue.


```python
sns.catplot(data=df,x="Category",y="Sales",kind="bar")
plt.show()
```

#### 2. Which Region Has the Most Orders?
Management wants to know which region receives the highest number of orders.

```python
sns.catplot(data=df,x="Region",kind="count")
plt.show()
```

```python
# Also we can use  countplot() to check region has the most orders.
sns.countplot(data=df , x='Region')
```

#### Show order counts by region separately for New and Returning customers

```python
sns.catplot(data=df,x="Region",kind="count", col="Customer_Type")
plt.show()
```

#### 3. Which Region Has the Highest Profit Variation?
To detect variability and outliers.

```python
sns.catplot(data=df,x="Region",y="Profit",kind="box")
plt.show()
```

#### 4. Which Category Gets Higher Discounts?
To compare distributions, not just averages.

```python
sns.catplot(data=df,x="Category",y="Discount",kind="violin")
plt.show()
```

#### 5. Do Returning Customers Spend More?
Compare average sales across customer groups.

```python
sns.catplot(data=df,x="Customer_Type",y="Sales",kind="bar")
plt.show()
```

#### 6. Does Higher Sales Mean Higher Profit?

```python
sns.relplot(data=df,x="Sales",y="Profit",kind="scatter")
plt.show()
```

#### 7. Does Discount Reduce Profit?

```python
sns.relplot(data=df,x="Discount",y="Profit",kind="scatter")
plt.show()
```

#### 8. Do Customers Buying More Items Spend More?

```python
sns.relplot(data=df,x="Quantity",y="Sales",kind="scatter")
plt.show()
```

#### 9. Does the Relationship Change by Category?

```python
sns.relplot(data=df,x="Sales",y="Profit",hue="Category",kind="scatter")
plt.show()
```

#### 10. Compare Sales and Profit by Customer Type

```python
sns.relplot(data=df,x="Sales",y="Profit",col="Customer_Type",kind="scatter")
plt.show()
```
