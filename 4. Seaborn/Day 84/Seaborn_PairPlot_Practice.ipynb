## Seaborn Pair Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "Age":[22,25,28,30,35,40,45,50,27,33,38,42],
    "Salary":[30000,35000,42000,50000,65000,70000,80000,90000,45000,55000,68000,75000],
    "Experience":[1,2,3,5,8,10,15,20,4,6,9,12],
    "Performance":[60,65,70,75,82,85,90,95,72,78,88,91],
    "Department":[
        "Sales","HR","IT","Finance",
        "Sales","HR","IT","Finance",
        "Sales","HR","IT","Finance"
    ],
    "Gender":[
        "Male","Female","Male","Female",
        "Female","Male","Female","Male",
        "Male","Female","Male","Female"
    ]
})
```

```python
df.head()
```

```python
df.info()
```

# 

#### Basic Pair Plot :

```python
sns.pairplot(df)
plt.show()
```

#### hue parameter : 

```python
sns.pairplot(df,hue="Department")
plt.show()
```

```python
sns.pairplot(df,hue="Gender")
plt.show()
```

#### palette parameter : 

```python
sns.pairplot(df,hue="Department",palette="Set2")
plt.show()
```

```python
sns.pairplot(df,hue="Department",palette="coolwarm")
plt.show()
```

#### diag_kind parameter : 

```python
sns.pairplot(df,diag_kind="hist")
plt.show()
```

```python
sns.pairplot(df,diag_kind="kde")
plt.show()
```

#### kind parameter : 

```python
sns.pairplot(df,kind="scatter")
plt.show()
```

```python
sns.pairplot(df,kind="kde")
plt.show()
```

```python
sns.pairplot(df,kind="hist")
plt.show()
```

#### corner parameter : 

```python
sns.pairplot(df,corner=True)
plt.show()
```

#### markers parameter : 

```python
sns.pairplot(df,hue="Gender",markers=["o", "s"])
plt.show()
```

#### height parameter : 

```python
sns.pairplot(df,height=2)
plt.show()
```

#### plot_kws parameter : 

```python
sns.pairplot(df,plot_kws={"alpha":0.7})
plt.show()
```

```python
sns.pairplot(df,plot_kws={"alpha":0.7,"s":100})
plt.show()
```

```python
sns.pairplot(df,plot_kws={"alpha":0.7,"s":100, "edgecolor":"red"})
plt.show()
```

```python
sns.pairplot(df,hue="Department",palette="Set2",diag_kind="hist",height=3,markers=["o","s","D","^"])
plt.show()
```
