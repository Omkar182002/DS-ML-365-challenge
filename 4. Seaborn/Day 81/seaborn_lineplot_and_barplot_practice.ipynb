## Line Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

###  Monthly Sales Trend
Create a line plot for monthly sales.

```python
Month = ['Jan','Feb','Mar','Apr','May','Jun']
Sales= [12000,15000,18000,17000,22000,25000]
```

```python
sns.lineplot(x=Month, y=Sales)
plt.show()
```

```python
sns.lineplot(x=Month, y=Sales)
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales Trend')
plt.grid()
plt.show()
```

##### Using Pandas DataFrame

```python
data= pd.DataFrame({
    'Month':Month,
    'Sales':Sales
})
data
```

```python
sns.lineplot(data= data, x='Month', y='Sales')

plt.title('Monthly Sales Trend')
plt.grid()
plt.show()
```

```python
# Another way
sns.lineplot(x=data.Month, y=data.Sales)

plt.grid()
plt.show()
```

```python
# Some other paramters
```

```python
sns.lineplot(data= data, x='Month', y='Sales', marker='o', color='k', linewidth=3)
plt.title('Monthly Sales Trend')
plt.grid()
plt.show()
```

# 

```python
df1 = pd.DataFrame({
    'Month':['Jan','Feb','Mar','Apr','May','Jun',
             'Jan','Feb','Mar','Apr','May','Jun'],
    'Sales':[100,120,140,130,160,180,
             80,100,110,125,140,150],
    'Region':['North','North','North','North','North','North',
              'South','South','South','South','South','South']
})
```

```python
df1
```

```python
sns.lineplot(data= df1, x='Month', y='Sales', marker='o')
plt.title('Monthly Sales Trend')
plt.grid()
plt.show()
```

##### hue Parameter 

```python
sns.lineplot(data= df1, x='Month', y='Sales', marker='o', hue='Region')
plt.title('Monthly Sales Trend')
plt.grid()
plt.show()
```

#### style Parameter : 

```python
sns.lineplot(data= df1, x='Month', y='Sales', marker='o', hue='Region',style='Region')
plt.title('Monthly Sales Trend')
plt.grid()
plt.show()
```

#### dashes Parameter :

```python
sns.lineplot(data= df1, x='Month', y='Sales', marker='o', hue='Region',style='Region', dashes=False)
plt.title('Monthly Sales Trend')
plt.grid()
plt.show()
```

#### palette Parameter : 

```python
sns.lineplot(data= df1, x='Month', y='Sales', marker='o', hue='Region',style='Region', dashes=False,palette='Set2')
plt.title('Monthly Sales Trend')
plt.grid()
plt.show()
```

```python
sns.lineplot(data= df1, x='Month', y='Sales', marker='o', hue='Region',style='Region', dashes=False,palette="Accent")
plt.title('Monthly Sales Trend')
plt.grid()
plt.show()
```

#### Legend Parameter :

```python
sns.lineplot(data= df1, x='Month', y='Sales', marker='o', hue='Region',style='Region', dashes=False,palette='Set2',legend=False)
plt.title('Monthly Sales Trend')
plt.grid()
plt.show()
```

## Bar Plot

##### Basic plot

```python
Department = ['Sales','HR','IT','Finance','Sales','HR','IT']
Salary = [50000,40000,70000,60000,55000,42000,75000]
```

```python
sns.barplot(x=Department,y=Salary)
plt.show()
```

```python
df2 = pd.DataFrame({
    'Department':['Sales','HR','IT','Finance','Sales','HR','IT','Finance'],
    'Salary':[50000,40000,70000,60000,55000,42000,75000,62000],
    'Gender':['Male','Female','Male','Female',
              'Female','Male','Female','Male'],
    'Experience':['Junior','Junior','Junior','Junior',
                  'Senior','Senior','Senior','Senior']
})
```

```python
df2
```

```python
sns.barplot(data=df2,x='Department',y='Salary')
plt.show()
```

#### hue Parameter  :

```python
sns.barplot(data=df2,x='Department',y='Salary', hue='Gender')
plt.show()
```

#### palette Parameter : 

```python
sns.barplot(data=df2,x='Department',y='Salary', palette='Set2')
plt.show()
```

```python
sns.barplot(data=df2,x='Department',y='Salary', hue='Gender', palette='Set2')
plt.show()
```

```python
sns.barplot(data=df2,x='Department',y='Salary', hue='Gender', palette='coolwarm')
plt.show()
```

#### estimator Parameter : 

```python
import numpy as np
```

```python
sns.barplot(data=df2,x='Department',y='Salary',estimator=np.max)
plt.show()
```

```python
sns.barplot(data=df2,x='Department',y='Salary', hue='Gender', estimator=np.max)
plt.show()
```

```python
sns.barplot(data=df2,x='Department',y='Salary',estimator=np.min)
plt.show()
```

```python
sns.barplot(data=df2,x='Department',y='Salary',estimator=np.mean)
plt.show()
```

#### errorbar Parameter :

```python
sns.barplot(data=df2,x='Department',y='Salary', errorbar='sd')
plt.show()
```

```python
sns.barplot(data=df2,x='Department',y='Salary', errorbar=None)
plt.show()
```

#### order Parameter : 

```python
sns.barplot(data=df2,x='Department',y='Salary',order=['IT','Finance','Sales','HR'])
plt.show()
```

#### width Parameter : 

```python
sns.barplot(data=df2,x='Department',y='Salary',width=0.3)
plt.show()
```

```python
sns.barplot(data=df2,x='Department',y='Salary',width=0.6)
plt.show()
```

#### saturation Parameter

```python
sns.barplot(data=df2,x='Department',y='Salary',saturation=1)
plt.show()
```

```python
sns.barplot(data=df2,x='Department',y='Salary',saturation=0.5)
plt.show()
```

```python
sns.barplot(data=df2,x='Department',y='Salary',saturation=0.2)
plt.show()
```
