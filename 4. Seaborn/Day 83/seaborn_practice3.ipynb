## Seaborn
- Count Plot
- Violin Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "Department": [
        "Sales","Sales","Sales","HR","HR","HR",
        "IT","IT","IT","Finance","Finance","Finance",
        "Marketing","Marketing","Marketing","Support","Support","Support"
    ],

    "Gender": [
        "Male","Female","Male",
        "Female","Male","Female",
        "Male","Female","Male",
        "Female","Male","Female",
        "Male","Female","Male",
        "Female","Male","Female"
    ],

    "Experience": [
        "Junior","Senior","Junior",
        "Senior","Junior","Senior",
        "Junior","Senior","Senior",
        "Junior","Senior","Junior",
        "Senior","Junior","Senior",
        "Junior","Senior","Junior"
    ],

    "Salary": [
        50000,65000,52000,
        45000,43000,47000,
        70000,80000,75000,
        62000,68000,64000,
        55000,58000,60000,
        40000,45000,42000
    ]
})
```

# 

###  || Seaborn Count Plot ||

#### Basic Count Plot : 

```python
sns.countplot(data=df,x="Department")
plt.show()
```

```python
sns.countplot(df.Department)
plt.show()
```

```python
sns.countplot(x= df.Department)
plt.show()
```

#### hue parameter : 

```python
sns.countplot(data=df,x="Department",hue="Gender")
plt.show()
```

```python
sns.countplot(data=df,x="Department",hue="Experience")
plt.show()
```

#### palette parameter : 

```python
sns.countplot(data=df,x="Department",hue="Gender",palette="Set2")
plt.show()
```

```python
sns.countplot(data=df,x="Department",hue="Gender",palette="coolwarm")
plt.show()
```

#### order parameter : 

```python
sns.countplot(data=df,x="Department",order=["IT","Finance","Sales","Marketing","HR","Support"])
plt.show()
```

```python
sns.countplot(data=df,x="Department",order=["Sales","Marketing","HR","Support","IT","Finance"])
plt.show()
```

#### color parameter : 

```python
sns.countplot(data=df,x="Department",color="green")
plt.show()
```

```python
sns.countplot(data=df,x="Department",color="red")
plt.show()
```

#### width parameter : 

```python
sns.countplot(data=df,x="Department",width=0.5)
plt.show()
```

```python
sns.countplot(data=df,x="Department",width=0.9)
plt.show()
```

## || Violin Plot ||

#### Basic Violin Plot : 

```python
sns.violinplot(data=df,x="Department",y="Salary")
plt.grid()
plt.show()
```

#### hue parameter : 

```python
sns.violinplot(data=df,x="Department",y="Salary",hue="Gender")
plt.show()
```

#### palette parameter : 

```python
sns.violinplot(data=df,x="Department",y="Salary",hue="Gender",palette="Set2")
plt.show()
```

#### split parameter : 

```python
sns.violinplot(data=df,x="Department",y="Salary",hue="Gender",split=True)
plt.show()
```

#### inner parameter : 

```python
sns.violinplot(data=df,x="Department",y="Salary",inner="box")
plt.show()
```

```python
sns.violinplot(data=df,x="Department",y="Salary",inner="stick")
plt.show()
```

```python
sns.violinplot(data=df,x="Department",y="Salary",inner="point")
plt.show()
```

#### linewidth parameter : 

```python
sns.violinplot(data=df,x="Department",y="Salary",linewidth=3)
plt.show()
```

#### density_norm parameter : 

```python
sns.violinplot(data=df,x="Department", y="Salary", density_norm="area")
plt.show()
```

#### cut parameter : 

```python
sns.violinplot(data=df,x="Department",y="Salary",cut=0)
plt.show()
```

```python
sns.violinplot(data=df,x="Department",y="Salary",cut=1)
plt.show()
```

#### fill parameter : 

```python
sns.violinplot(data=df,x="Department",y="Salary",fill=False)
plt.show()
```
