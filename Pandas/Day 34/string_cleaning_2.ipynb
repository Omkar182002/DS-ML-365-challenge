# 

#### 1. Remove leading and trailing spaces from all text columns.

```python
text_col = df.select_dtypes(include=object).columns
```

```python
df[text_col] = df[text_col].apply(lambda col : col.str.strip())
```

```python
df[text_col].head()
```

#### 2. Convert all news titles to lowercase.

```python
df['title'] = df['title'].str.lower()
df.head()
```

#### 3. Remove duplicate spaces from article text.

```python
df['text'] = df['text'].str.replace(r'\s+',' ',regex=True)
```

```python
df.head()
```

###### BREAKING NEWS

###### SHOCKING

#### 5. Remove newline (\n) characters from text column.

```python
df['text'] = df['text'].str.replace(r'\n','',regex = True)
```

```python
df['text'] 
```

#### 6. Remove tab (\t) spaces from text.

```python
df['text'] = df['text'].str.replace(r'\t','',regex = True)
```

```python
df['word_count'] = df['text'].str.split().str.len()
```
