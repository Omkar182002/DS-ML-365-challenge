# Data Cleaning 

```python
df['Customer_Name'] = df['Customer_Name'].fillna('Unknown')
```

```python
df['Customer_Name'] = df['Customer_Name'].replace('', 'Unknown')
```

```python
df['Email'] = df['Email'].astype(str)
```

```python
mask = df['Email'].str.contains(r'^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$', regex=True, na=False)
```

```python
df.loc[~mask, 'Email'] = np.nan
```

# 

```python
df.columns
```

```python
df['City'].unique()
```

```python
df['City'] = df['City'].astype(str).str.strip().str.title()
```

```python
city_map = {
    'Blr':'Bangalore',
    'Bengaluru':'Bangalore',
    'Belagavi':'Belgaum'
}
```

```python
df['City'] = df['City'].replace(city_map)
```

# 

```python
df['Product'].unique()
```

# 

```python
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
```

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
df['Customer_Rating'] = pd.to_numeric(df['Customer_Rating'], errors='coerce')
```

# 

```python
df.loc[df['Quantity'] <= 0, 'Quantity'] = np.nan
```

```python
df.loc[df['Price'] <= 0, 'Price'] = np.nan
```

```python
df.loc[~df['Customer_Rating'].between(1,5), 'Customer_Rating'] = np.nan
```

# 

```python
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
```

```python
df['Price'] = df['Price'].fillna(df['Price'].median())
```

```python
df['Customer_Rating'] = df['Customer_Rating'].fillna(df['Customer_Rating'].median())
```

# 

```python
df['Payment_Method'] = df['Payment_Method'].replace('', np.nan)
```

```python
df['Payment_Method'] = df['Payment_Method'].fillna('Unknown')
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce', format='mixed')
```
