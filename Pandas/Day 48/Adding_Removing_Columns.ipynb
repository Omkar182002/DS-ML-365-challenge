# 4 — Intermediate

### 21. Create a new column SalaryCategory.
Condition:  
Salary > 60000 → High  
Salary between 50000 and 60000 → Medium  
Otherwise → Low

```python
df6 = df.copy()
```

```python
conditions = [df6["Salary"] > 60000, (df6["Salary"] >= 50000) & (df6["Salary"] <= 60000)]
choices = ["High", "Medium"]
df6["SalaryCategory"] = np.select(conditions, choices, default="Low")
```

```python
df6
```

### 22. Create a column NetIncome.
