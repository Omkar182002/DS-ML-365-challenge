## B. filling missing values (especially with replace()--

```python
dataB = {
    'timestamp': [
        '2024-01-01 10:00', '2024-01-01 11:00', '2024-01-01 12:00',
        '2024-01-01 13:00', '2024-01-01 14:00'
    ],
    'temperature': [22.5, np.nan, 24.0, 'NA', 25.5],
    'humidity': [45, 50, np.nan, '', 55],
    'salary': [30000, 35000, np.nan, 40000, 'null'],
    'marks': [80, np.nan, 85, 90, '?'],
    'price': [100, 200, '-', np.nan, 300],
    'category': ['A', 'A', 'B', 'B', 'A'],
    'age': [25, np.nan, 22, '', 30],
    'status': ['student', 'employee', 'student', 'employee', 'student']
}

dfB = pd.DataFrame(dataB)
```

```python
dfB
```

#### 1. Replace 'NA' in temperature with np.nan.

```python
d1= dfB.copy()
d1['temperature'] = d1['temperature'].replace('NA', np.nan)
d1
```

#### 2. Replace '' in humidity with np.nan.

```python
d1['humidity'] = d1['humidity'].replace('', np.nan)
```

```python
pd.set_option('future.no_silent_downcasting', True)
```

#### 3. Replace 'null' in salary with np.nan.

```python
d1['salary'] = d1['salary'].replace('null', np.nan)
d1
```

#### 4. Replace '?' in marks with np.nan.

```python
d1['marks'] = d1['marks'].replace('?', np.nan)
d1
```

#### 5. Replace '-' in price with np.nan.

```python
d1['price'] = d1['price'].replace('-', np.nan)
d1
```

#### 6. Replace ['NA', 'null'] only in numeric columns.

```python
num_cols = d1.select_dtypes(include=['number']).columns
d1[num_cols] = d1[num_cols].replace(['NA','null'], np.nan)
d1
```

#### 7. Replace missing indicators only in temperature and humidity.

```python
d2 = dfB.copy()
d2[['temperature','humidity']] = d2[['temperature','humidity']].replace(['NA',''], np.nan)
d2
```

#### 8. Replace all NaN values in temperature with 0.

```python
d2['temperature'] = d2['temperature'].replace(np.nan, 0)
```

```python
d2
```

#### 9. Replace NaN in each column with its median using loop.

```python
num_cols = d2.select_dtypes(include=['number']).columns
for col in num_cols:
    d2[col] = d2[col].replace(np.nan, d2[col].median())
```

```python
d2
```

#### 11. Replace missing temperature values with the column’s average temperature.

```python
d4 = dfB.copy()
```

```python
d4['temperature'] = d4['temperature'].replace(['Na','None','Null','_',''], np.nan)
d4['temperature'] = pd.to_numeric(d4['temperature'], errors='coerce')
mean_d = np.mean(d4['temperature'])
d4['temperature'] = d4['temperature'].replace(np.nan,mean_d)
d4
```

#### 12. Replace missing age with 18 if status = student.

```python
d5 = dfB.copy()
d5['age'] = d5.apply(lambda r: 18 if pd.isna(r['age']) and r['status']=='student' else r['age'], axis=1)
```

```python
d5
```

#### 13. Replace ['?', ''] only in marks and age.

#### 14. Replace missing temperature with 25 only when humidity > 50.

```python
d6 = dfB.copy()
d6['temperature'] = d6.apply(lambda r: 25 if pd.isna(r['temperature']) and r['humidity']>50 else r['temperature'], axis=1)
```
