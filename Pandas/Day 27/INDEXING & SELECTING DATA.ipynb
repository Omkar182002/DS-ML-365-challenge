## Pandas Slicing and Selecting 

```python
import pandas as pd
import numpy as np
 
```

```python
df = pd.DataFrame({"Name":["Ram","Sham","Pramod","Raj","Kallapa"],"age":[22,23,25,26,20],"city":list("ABCDE"), "Code":[101,104,103,105,111]}, index = ["row1",'row2','row3','row4','row5'])
df
```

#### 1: Select a slice of rows and all columns using .loc


```python
df.loc['row3':'row5',:]
```

#### 2: Select a slice of rows and a slice of columns using .loc

```python
df.loc['row3':'row5','Name':'city']
```

#### 3: Select the last row using .loc

```python
df.loc[['row4']]  
```

```python
df.loc['row4']
```

```python
df.loc[[df.index[-1]]]
```


#### 4.  Select first three rows using .loc

```python
df.loc[df.index[:3]]
```

```python
df.loc[df['age']>22]
```

#### 36 Select rows by matching index string pattern

```python
df4 = pd.DataFrame({'A': [1, 2, 3],'B':list("TUR")}, index=['item-1', 'item-2', 'item-3'])
mask = df4.index.str.contains('item')
print(df4[mask])
```
