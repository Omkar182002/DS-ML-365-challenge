# Exponential Weighted Moving

```python
import pandas as pd

sales_df = pd.DataFrame({
    'Sales': [
        100, 120, 150, 170, 200,
        220, 250, 270, 300, 320,
        310, 330, 360, 390, 410,
        430, 450, 470, 490, 520,
        540, 560, 580, 600, 620,
        640, 660, 680, 700, 720
    ]
})

print(sales_df)
```

## 1 — Large Sales Dataset

### 1. Create EWM mean using span=3.
Recent values get higher importance.

```python
sales_df['EWM_Mean_Span3'] = sales_df['Sales'].ewm(span=3).mean()
```

### 2. Create EWM mean using span=5.

```python
sales_df['EWM_Mean_Span5'] = sales_df['Sales'].ewm(span=5).mean()
# Smoother trend because larger span gives less weight to sudden changes.
```

### 3. Compare EWM mean vs rolling mean.

```python
sales_df['Rolling_Mean_3'] = sales_df['Sales'].rolling(window=3).mean()

sales_df['EWM_Mean_3'] = sales_df['Sales'].ewm(span=3).mean()

print(sales_df[['Sales', 'Rolling_Mean_3', 'EWM_Mean_3']])
```

### 4. Compare EWM mean vs expanding mean.

```python
sales_df['Expanding_Mean'] = sales_df['Sales'].expanding().mean()

sales_df['EWM_Mean'] = sales_df['Sales'].ewm(span=5).mean()

print(sales_df[['Sales', 'Expanding_Mean', 'EWM_Mean']])
```

### 5. Detect recent sales trend.

```python
sales_df['Recent_Trend'] = sales_df['Sales'].ewm(span=3).mean()

print(sales_df[['Sales', 'Recent_Trend']])
```

### 6. Detect sales acceleration periods.

```python
sales_df['EWM_Trend'] = sales_df['Sales'].ewm(span=3).mean()

sales_df['Acceleration'] = sales_df['EWM_Trend'].diff()

print(sales_df[['Sales', 'EWM_Trend', 'Acceleration']])
```

### 7. Detect sales slowdown periods.

```python
sales_df['Slowdown'] = sales_df['Acceleration'] < 5

print(sales_df[['Sales', 'Acceleration', 'Slowdown']])
```

### 8. Create EWM std.

```python
sales_df['EWM_STD'] = sales_df['Sales'].ewm(span=5).std()

print(sales_df)
```

### 9. Detect unstable sales periods.

```python
sales_df['Unstable'] = sales_df['EWM_STD'] > 25

print(sales_df[['Sales', 'EWM_STD', 'Unstable']])
```

### 10. Detect stable sales periods.

```python
sales_df['Stable'] = sales_df['EWM_STD'] < 20

print(sales_df[['Sales', 'EWM_STD', 'Stable']])
```

### 11. Use adjust=True.

```python
sales_df['Adjust_True'] = sales_df['Sales'].ewm(
    span=5,
    adjust=True
).mean()

print(sales_df[['Sales', 'Adjust_True']])
```

### 12. Use adjust=False.

```python
sales_df['Adjust_False'] = sales_df['Sales'].ewm(
    span=5,
    adjust=False
).mean()

print(sales_df[['Sales', 'Adjust_False']])
```

## 2 — Large Stock Market Dataset

```python
stock_df = pd.DataFrame({
    'Price': [
        100, 102, 101, 105, 108,
        110, 115, 112, 118, 120,
        122, 125, 123, 128, 130,
        135, 138, 140, 145, 148,
        150, 152, 149, 155, 158,
        160, 162, 165, 168, 170
    ]
})

print(stock_df)
```

### 1. Create EWM moving average.

```python
stock_df['EWM_Mean'] = stock_df['Price'].ewm(span=5).mean()

print(stock_df)
```

### 2. Create EWM volatility.

```python
stock_df['EWM_Volatility'] = stock_df['Price'].ewm(span=5).std()

print(stock_df)
```

### 3. Detect high volatility periods.

```python
stock_df['High_Volatility'] = (
    stock_df['EWM_Volatility'] > 4
)

print(
    stock_df[
        ['Price', 'EWM_Volatility', 'High_Volatility']
    ]
)
```

### 4. Detect high volatility periods.

```python
stock_df['Low_Volatility'] = (
    stock_df['EWM_Volatility'] < 3
)

print(
    stock_df[
        ['Price', 'EWM_Volatility', 'Low_Volatility']
    ]
)
```

### 5. Detect risky market periods.

```python
stock_df['Risky_Market'] = (
    stock_df['EWM_Volatility'] > 5
)

print(
    stock_df[
        ['Price', 'EWM_Volatility', 'Risky_Market']
    ]
)
```

### 6. Detect stable market periods.

```python
stock_df['Stable_Market'] = (
    stock_df['EWM_Volatility'] < 2
)

print(
    stock_df[
        ['Price', 'EWM_Volatility', 'Stable_Market']
    ]
)
```

### 7. Compare EWM std vs rolling std.

```python
stock_df['Rolling_STD'] = (
    stock_df['Price']
    .rolling(window=5)
    .std()
)

stock_df['EWM_STD'] = (
    stock_df['Price']
    .ewm(span=5)
    .std()
)

print(
    stock_df[
        ['Price', 'Rolling_STD', 'EWM_STD']
    ]
)
```

### 8. Compare EWM mean vs rolling mean.

```python
stock_df['Rolling_Mean'] = (
    stock_df['Price']
    .rolling(window=5)
    .mean()
)

stock_df['EWM_Mean'] = (
    stock_df['Price']
    .ewm(span=5)
    .mean()
)

print(
    stock_df[
        ['Price', 'Rolling_Mean', 'EWM_Mean']
    ]
)
```

### 9. Use span=10.

```python
stock_df['EWM_Span10'] = (
    stock_df['Price']
    .ewm(span=10)
    .mean()
)

print(
    stock_df[
        ['Price', 'EWM_Span10']
    ]
)
```

### 10. Use span=3.

```python
stock_df['EWM_Span3'] = (
    stock_df['Price']
    .ewm(span=3)
    .mean()
)

print(
    stock_df[
        ['Price', 'EWM_Span3']
    ]
)
```

## 3 — Large Website Traffic Dataset

```python
traffic_df = pd.DataFrame({
    'Visitors': [
        1000, 1200, 1500, 1700, 1600,
        1800, 2000, 2200, 2100, 2300,
        2500, 2700, 2600, 2800, 3000,
        3200, 3100, 3300, 3500, 3700,
        3600, 3800, 4000, 4200, 4100,
        4300, 4500, 4700, 4900, 5000
    ]
})

print(traffic_df)
```

### 1. Create EWM average visitors.

```python
traffic_df['EWM_Visitors'] = (
    traffic_df['Visitors']
    .ewm(span=5)
    .mean()
)

print(traffic_df)
```

### 2. Detect recent traffic trend.

```python
traffic_df['Recent_Trend'] = (
    traffic_df['Visitors']
    .ewm(span=5)
    .mean()
)

print(
    traffic_df[
        ['Visitors', 'Recent_Trend']
    ]
)
```

### 3. Detect traffic spikes.

```python
traffic_df['Traffic_Change'] = (
    traffic_df['Visitors']
    .diff()
)

traffic_df['Traffic_Spike'] = (
    traffic_df['Traffic_Change'] > 300
)

print(
    traffic_df[
        ['Visitors', 'Traffic_Change', 'Traffic_Spike']
    ]
)
```

### 4. Detect traffic drops.

```python
traffic_df['Traffic_Drop'] = (
    traffic_df['Traffic_Change'] < -200
)

print(
    traffic_df[
        ['Visitors', 'Traffic_Change', 'Traffic_Drop']
    ]
)
```

### 5. Create EWM std.

```python
traffic_df['EWM_STD'] = (
    traffic_df['Visitors']
    .ewm(span=5)
    .std()
)

print(traffic_df)
```

### 6. Detect unstable traffic periods.

```python
traffic_df['Unstable_Traffic'] = (
    traffic_df['EWM_STD'] > 250
)

print(
    traffic_df[
        ['Visitors', 'EWM_STD', 'Unstable_Traffic']
    ]
)
```

### 7. Detect stable traffic periods.

```python
traffic_df['Stable_Traffic'] = (
    traffic_df['EWM_STD'] < 150
)

print(
    traffic_df[
        ['Visitors', 'EWM_STD', 'Stable_Traffic']
    ]
)
```

### 8. Compare EWM vs rolling visitors.

```python
traffic_df['Rolling_Mean'] = (
    traffic_df['Visitors']
    .rolling(window=5)
    .mean()
)

traffic_df['EWM_Mean'] = (
    traffic_df['Visitors']
    .ewm(span=5)
    .mean()
)

print(
    traffic_df[
        ['Visitors', 'Rolling_Mean', 'EWM_Mean']
    ]
)
```

### 9.Compare EWM vs expanding visitors.

```python
traffic_df['Expanding_Mean'] = (
    traffic_df['Visitors']
    .expanding()
    .mean()
)

traffic_df['EWM_Mean'] = (
    traffic_df['Visitors']
    .ewm(span=5)
    .mean()
)

print(
    traffic_df[
        ['Visitors', 'Expanding_Mean', 'EWM_Mean']
    ]
)
```

### 10. Create weighted traffic trend.

```python
traffic_df['EWM_Span7'] = (
    traffic_df['Visitors']
    .ewm(span=7)
    .mean()
)

print(
    traffic_df[
        ['Visitors', 'EWM_Span7']
    ]
)
```
