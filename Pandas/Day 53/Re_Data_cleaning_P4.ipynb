# Data Cleaning

```python
df.dropna(subset='Order_ID').shape
```

```python
df['Order_ID'].fillna(method='ffill').dropna().shape
```

```python
df = df.sort_values(by=['Order_Date', 'Customer_Name'])
```

##### Forward fill missing Order_ID

###### Because blank Order_ID rows belong to the previous order (line items).



```python
df['Order_ID'] = df['Order_ID'].fillna(method='ffill')
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset='Order_ID')
```

```python
df.isna().sum()
```

# 

```python
df['Customer_Name'] = df.groupby('Order_ID')['Customer_Name'].ffill()
```

```python
df.isna().sum()
```

```python
df.isna().mean().round(2)
```

```python
df[df['Customer_Name'].isna()].sort_values('Order_ID')
```

```python
df['Customer_Name'] = df['Customer_Name'].fillna('Unknown')
```

```python
df.isna().mean().round(2)
```

```python
df['Product'].unique()
```

```python
df[df['Product'].isna()]
```

```python
df.groupby('Order_ID')['Product'].unique()
```

```python
df['Category'].unique()
```

```python
product_category_map = (df.dropna(subset=['Product', 'Category']).drop_duplicates(subset=['Product']).set_index('Product')['Category'])
```

```python
product_category_map
```

```python
product_category_map = {
    'Phone': 'Electronics',
    'Monitor': 'Electronics',
    'Laptop': 'Electronics',
    'Tablet': 'Electronics',
    'Mouse': 'Accessories',
    'Keyboard': 'Accessories'
}
```

```python
product_category_map
```

##### Office can remain for products like: Printer, Paper, Pen, Chair

### Fill Missing Categories

```python
df['Category'] = df['Category'].fillna(
    df['Product'].map(product_category_map)
)
```

```python
df['Price'] = df.groupby(['Product','Quantity'])['Price'].transform(lambda p : p.fillna(p.median()))
```

```python
df['temp_price']= df['Price']
```

# 

```python
df['temp_Payment_Method'] = df['Payment_Method']
```

```python
filled_pay_method = (df.groupby(['Order_ID'])['Payment_Method'].transform(lambda x : x.mode()[0] if not x.mode().empty else np.nan))
```

```python
df['temp_Payment_Method'] = df['temp_Payment_Method'].fillna(filled_pay_method)
```

```python
df[['Payment_Method','temp_Payment_Method']].isna().mean().round(2)
```

```python
df[['Order_ID','Payment_Method','temp_Payment_Method']].head(20)
```

```python
df['Payment_Method'] = df['temp_Payment_Method'] 
```

```python
df.drop(columns='temp_Payment_Method',inplace=True)
```

```python
df['Customer_Rating'] = df['Customer_Rating'].fillna(df['Customer_Rating'].median())
```

```python
df['Customer_Rating'].unique()
```

```python
df.loc[(df['Customer_Rating']<0)| (df['Customer_Rating']> 5),'Customer_Rating'] = np.nan
```

```python
df['Customer_Rating'] = df['Customer_Rating'].fillna(df['Customer_Rating'].median())
```

```python
df.isna().mean().round(2)
```

```python
filled_order_date= (df.groupby(['Order_ID'])['Order_Date'].transform(lambda a : a.ffill().bfill()))           
```

```python
df['temp_Order_Date'] = df['Order_Date']
```

```python
df['temp_Order_Date'] = df['temp_Order_Date'].fillna(filled_order_date)
```

```python
df.isna().mean().round(2)
```

```python
df[df['Order_Date'].notna()].groupby('Order_ID').size()
```

```python
df.drop(columns='temp_Order_Date',inplace=True)
```

```python
df['Payment_Method'] = df['Payment_Method'].fillna('Unknown')
```

```python
df['City'] = df['City'].fillna('Unknown')
```

```python
mask = df['Email'].str.match(r'^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$',na=False)
```

```python
df.loc[~mask, 'Email'] = np.nan
```

```python
df['Email'] = df['Email'].fillna('Unknown')
```

```python
df.isna().mean().round(2)
```

```python
df['Discount_Percent'] = df['Discount_Percent'].fillna(0) 
```
