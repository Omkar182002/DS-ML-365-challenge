## Data Cleaning

```python
df['Employee_ID'].isnull().sum()
```

```python
df.dropna(subset='Employee_ID',inplace=True)
```

```python
df['Employee_ID'].isnull().sum()
```

```python
df['Employee_ID'].unique()[:20]
```

```python
import re

def standardize_id(emp_id):
    numbers = re.findall(r'\d+', str(emp_id))
    if numbers:
        return f"EMP-{int(numbers[0]):04d}"
        
df['Employee_ID'] = df['Employee_ID'].apply(standardize_id)
```

```python
df['Employee_ID'].unique()[:20]
```
