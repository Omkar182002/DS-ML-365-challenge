```python
df['City'].str.replace(r'[^a-zA-Z\s]','',regex=True).unique()
```

```python
df['City'] = df['City'].str.replace(r'[^a-zA-Z\s]','',regex=True)
```

```python
city_map = {
    'New Delhi': 'Delhi',
    'Bangalore': 'Bengaluru',
    'Bombay': 'Mumbai'
}
```

```python
df['City'] = df['City'].replace(city_map)
```

```python
df['City'].unique()
```

# 

```python
pattern = r'^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$'
```

```python
df['Email'] = df['Email'].where(df['Email'].str.match(pattern, na=False),np.nan)
```
