# Pandas Merging & Joining

```python
import pandas as pd
```

#  #Pandas merge()

```python
customers = pd.DataFrame({
    'customer_id': [101,102,103,104,105,106,107,108,109,110],
    'customer_name': ['Omkar','Rahul','Sneha','Amit','Pooja','Kiran','Neha','Vijay','Anjali','Rohit'],
    'city': ['Delhi','Mumbai','Pune','Delhi','Bangalore', 'Pune','Mumbai','Delhi','Hyderabad','Chennai'],
    'age': [22,25,24,30,27,29,23,31,26,28]
})
```

```python
customers.head()
```

```python
customers.info()
```

# 

```python
orders = pd.DataFrame({
    'order_id': [1001,1002,1003,1004,1005,1006,1007,1008,1009,1010,1011 ],
    'customer_id': [101,102,102,104,105,108,101,111,109,105,120],
    'product': ['Laptop','Phone','Keyboard','Mouse','Tablet','Monitor','Printer','Camera','Phone','Laptop','TV'],
    'amount': [55000,20000,3000,1500,25000,12000,8000,45000,22000,60000,70000]
})
```

```python
orders.head()
```

```python
orders.info()
```

# 

### 1. Merge customers and orders using customer_id.
(Create customer order report.)

```python
pd.merge(customers,orders,on = 'customer_id')
```

### 2.Perform INNER JOIN between customers and orders.
Question: Which customers actually placed orders?

```python
pd.merge(customers, orders, on='customer_id', how='inner')
```

### 3.Perform LEFT JOIN.
Question: Show all customers even if they never ordered.

```python
pd.merge(customers, orders, on='customer_id', how='left')
```

### 4 . Perform RIGHT JOIN.
Question:  Show all orders even if customer information is missing.

```python
pd.merge(customers, orders, on='customer_id', how='right')
```

### 5 . Perform FULL OUTER JOIN.
Question: Show every customer and every order.

```python
pd.merge(customers,orders,how = 'outer', on='customer_id')
```

# 

### 6.Find customers who never placed orders.

```python
merge_data = pd.merge(customers,orders,on='customer_id',how='left')
merge_data[merge_data['order_id'].isna()]
```

# 

### 7.Find orders having invalid customer IDs.

```python
merge_data = pd.merge(customers, orders, on='customer_id', how='right')
merge_data[merge_data['customer_name'].isna()]
```
