# Pandas Expanding Function

```python
import pandas as pd
```

## 1 — Expanding Mean

```python
sales_df = pd.DataFrame({'Sales': [100, 200, 300, 400, 500]})
```

```python
sales_df
```

### 1. Create expanding mean.

```python
sales_df.expanding().mean()
```

### 2. Store result in column Expanding_Mean.

```python
sales_df['Expanding_Mean'] = sales_df['Sales'].expanding().mean()
```

```python
sales_df
```

### 3. Find average till each row.

```python
sales_df['Sales'].expanding().mean()
```

### 4. Compare expanding mean with normal mean.

```python
normal_mean = sales_df['Sales'].mean()
```

```python
normal_mean
```

```python
sales_df['Expanding_Mean']
```

### 5. Find final cumulative average.

```python
final_avg = sales_df['Expanding_Mean'].iloc[-1]
final_avg
```

### 6. Which rows become NaN?

```python
sales_df['Expanding_Mean_2'] = (sales_df['Sales'].expanding(min_periods=2).mean())
```

```python
sales_df
```

### 7. Create expanding mean using min_periods=3.

```python
sales_df['Sales'].expanding(min_periods=3).mean()
```

### 8. Compare expanding mean vs rolling mean.

```python
sales_df
```

```python
sales_df['Rolling_mean'] = sales_df['Sales'].rolling(2).mean()
```

```python
sales_df
```

## 5 — Expanding Standard Deviation

```python
stock_df = pd.DataFrame({'Price': [100, 105, 102, 110, 108, 115]})
```

```python
stock_df
```

### 1. Create expanding std.

```python
stock_df.expanding().std()
```

```python
stock_df.expanding(min_periods=0).std()
```

```python
stock_df.expanding(min_periods=1).std()
```

```python
stock_df.expanding(min_periods=3).std()
```

### 2.Store result in Volatility.

```python
stock_df['Volatility'] = stock_df['Price'].expanding().std()
```

```python
stock_df
```

### 3.Find periods of high volatility.

```python
high_volatility = stock_df[stock_df['Volatility'] > 4]
high_volatility
```

### 4.Find periods of low volatility.

```python
low_volatility = stock_df[stock_df['Volatility'] < 4]
```

```python
low_volatility
```

### 5. Compare expanding std vs rolling std.

```python
stock_df['Rolling_STD'] = stock_df['Price'].rolling(3).std()
stock_df[['Price', 'Volatility', 'Rolling_STD']]
```

## 7 — GroupBy + Expanding

```python
store_df = pd.DataFrame({
    'Store': ['A', 'A', 'A', 'B', 'B', 'B'],
    'Sales': [100, 200, 300, 400, 500, 600]
    })
```

```python
store_df
```

### 1. Create expanding sum separately for each store.

```python
store_df.groupby('Store')['Sales'].expanding().sum()
```

```python
store_df.groupby('Store')['Sales'].expanding().sum().reset_index()
```

### 2. Create expanding mean separately for each store.

```python
store_df.groupby('Store')['Sales'].expanding().mean()
```

### 3. Create expanding max separately for each store.

```python
store_df.groupby('Store')['Sales'].expanding().max().reset_index()
```

### 4. Create expanding min separately for each store.

```python
store_df.groupby('Store')['Sales'].expanding().min()
```

```python
store_df.groupby('Store')['Sales'].expanding().min().reset_index()
```

### 5. Create expanding std separately for each store.

```python
store_df.groupby('Store')['Sales'].expanding().std()
```

### 6. Find best-performing store.

```python
total_sales = store_df.groupby('Store')['Sales'].sum()

best_store = total_sales.idxmax()

print("\nBest Performing Store")
print("Store:", best_store)
print("Total Sales:", total_sales.max())
```

### 7. Use reset_index(level=0, drop=True).

```python
store_df.groupby('Store')['Sales'].expanding().min().reset_index(level=0,drop=True)
```

### 8. Find most stable store.

```python
store_stability = store_df.groupby('Store')['Sales'].std()
stable_store = store_stability.idxmin()
print()
print("Most Stable Store")
print("Store:", stable_store)
print("STD:", store_stability.min())
```
