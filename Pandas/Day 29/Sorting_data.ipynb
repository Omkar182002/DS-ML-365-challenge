## PART 1: Series Sorting

```python
s = pd.Series([120, 450, 300, 800, 150, 220, 90, 600, 700, 50])
```

```python
s
```

#### 1. Sort values ascending

```python
s.sort_values()
```

#### 2. Sort values descending

```python
s.sort_values(ascending=False)
```

#### 3. Get top 5 highest values

```python
s.sort_values(ascending=False).head()
```

```python
s.nlargest(5)
```

#### 4. Get bottom 3 smallest values

```python
s.sort_values().head(3)
```

```python
s.nsmallest(3)
```

#### 5. Sort values and reset index

```python
s.sort_values()
```

```python
s.sort_values().reset_index()
```

```python
s.sort_values().reset_index(drop=True)
```

#### 6. Sort values and keep original index

```python
s.sort_values()
```

#### 7. Sort values but place NaN at the end

```python
s.sort_values( na_position='last')
```

```python
s2 = pd.Series([22,1,13,24,17,None,22, 0, None,11])
```

```python
s2
```

```python
s2.sort_values(na_position='last')
```

#### 8. Sort values based on absolute difference from mean

```python
mean_value = s.mean()
sorted_s = s.sort_values(key=lambda x : abs(x-mean_value))
sorted_s
```

#### 9. Sort values based on remainder when divided by 10

```python
s.sort_values(key=lambda x:x%10)
```

#### 10. Sort values and remove duplicates

```python
s3 = pd.Series([40, 10, 20, 10, 30, 40, 50])
s3
```

```python
s3.sort_values().drop_duplicates()
```

```python
s3.sort_values().drop_duplicates().reset_index(drop=True)
```

#### 11. Sort values based on custom condition (even numbers first, then odd)

```python
s.sort_values(key=lambda x: x % 2)
```

```python
s4 = pd.Series([7, 2, 9, 4, 11, 6, 3, 8])
```

```python
s4.sort_values(key=lambda x: x%2)
```

#### 12. Sort values in descending order but keep index aligned

```python
s.sort_values(ascending=False)
```

#### 13. Rank values and sort based on rank

```python
s.rank().sort_values()
```

#### 14. Sort values within a rolling window

```python
s.rolling(window=3).apply(lambda x: sorted(x)[-1])
```

#### 15. Filter employees earning above department average salary

```python
mean_val = s.mean()
s[s > mean_val].sort_values()
```

## 16. Sort values based on distance from mean

```python
s.sort_values(key=lambda x: abs(x - x.mean()))
```

#### 17. Sort values closest to median first

```python
s.sort_values(key = lambda x : abs(x- s.median()))
```

#### 18. Sort only values greater than 200

```python
s[s>200].sort_values()
```

#### 19. Sort based on square of values

```python
s.sort_values(key =lambda a : a**2)
```

#### 20.  Sort values and return top 25%

```python
s.sort_values(ascending=False).head(int(len(s) * 0.25))
```

## 🔹 PART 2: DataFrame Sorting

```python
df = pd.read_csv("col_row-dataset.csv")
```

```python
df.sample(5)
```

#### 1. Sort by age ascending

```python
df['age'].sort_values()
```

```python
df.sort_values(by='age').head(10)
```

#### 2. Sort by salary descending

```python
df.sort_values(by ='salary',ascending= False).head(154)
```

#### 3. Sort by name alphabetically

```python
df.sort_values(by="name").head()
```

```python
df.sort_values(by="name").tail()
```

#### 4. Sort by city

```python
df.sort_values(by = 'city')
```

#### 5. Sort by experience

```python
df.sort_values(by ='experience')
```

#### 6. Sort by salary then age

```python
df.sort_values(by =["salary",'age']).head(15)
```

```python
df.sort_values(by =["salary",'age']).tail(15)
```

#### 7. Sort by department then experience

```python
df.sort_values(by=['department','experience'])
```

#### 8. Sort by city then name

```python
df.sort_values(by=['city','name'])
```

#### 9. Sort by salary (NaN last)

```python
df.sort_values(by='salary').tail()
```

#### 10. Sort by multiple columns with different order

```python
df.sort_values(by=['department', 'salary'],ascending=[True, False])
```

#### 11. Get top 10 highest salary rows

```python
df.nlargest(10,columns='salary')
```

#### 12. Get lowest 5 experience rows

```python
df.nsmallest(5,columns='experience')
```

#### 13. Sort employees by salary within department

```python
df.sort_values(by=['department', 'salary'])
```

#### 14. Sort employees by experience within city

```python
df.sort_values(by=['experience','city'])
```

#### 15. Sort dataset by length of names

```python
df.sort_values(by='name', key=lambda x: x.str.len())
```

#### 16. Sort dataset by salary difference from mean

```python
df.sort_values(by='salary', key=lambda x: abs(x-x.mean()))
```

#### 17. Sort by salary/experience ratio

```python
df.sort_values(by='salary',key=lambda x: df['salary'] / df['experience'])
```

#### 18. Sort departments by average salary

```python
df.groupby('department')['salary'].mean().sort_values(ascending=False)
```

#### 19. Sort cities by employee count

```python
df.groupby('city')['name'].count()
```

#### 20. Sort by most frequent names

```python
df['name'].value_counts().sort_values(ascending=False)
```
