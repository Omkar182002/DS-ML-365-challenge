```python
import numpy as np
```

##### 1.	Use np.logical_and() on a = [True, False, True], b = [False, False, True].

```python
a1 = np.array([True,False,True])
b1 = np.array([False,False,True])
print(a1& b1)
print(np.logical_and(a1,b1))
```

##### 2.	Use np.logical_or() on a = [0, 1, 2], b = [1, 0, 3].
###### & , |,^ - use only boolean type data not numberic 
###### (& , |,^) = In Boolean -> Logaical operation
###### (& , |,^) = In Numeric -> Bitwise operation
###### (& , |,^) = In Str -> Error

```python
a2 = np.array([0, 1, 2])
b2 = np.array([1, 0, 3])
print(a2.astype(bool)|b2.astype(bool))
print(a2|b2)
print(np.logical_or(a2,b2))
```

##### 5.	Create a condition using logical_or():
a > 3 OR b < 2 where a = [2, 4, 1], b = [1, 2, 3].


```python
a5 = np.array([2, 4, 1])
b5 = np.array([1,2,3])
print(np.logical_or(a5,b5,where=(a5>3)|(b5<2)))

print(np.logical_or(a5>3,b5<2))
```

##### 20.	Chain logical_and() and logical_or() to create a complex filter.

```python
a20 = np.random.randint(1,100,10)
b20 = np.random.randint(1,100,10)
cond1 = np.logical_and(a20 > 3, b20 < 6)
cond2 = a20 < 2
final = np.logical_or(cond1, cond2)
print(final)
```
