#### 21. Check if two arrays have same shape.

## 

```python
a21 = np.arange(20).reshape(-1,10)
b21 = np.arange(20).reshape(2,-1)
c21 = np.arange(20).reshape(2,-1).T
print("a21 and b21 same shape :",a21.shape == b21.shape)
print("b21 and c21 same shape :",b21.shape == c21.shape)
```

```python
print("Shape of a21 array :",a21.shape)
print("Shape of b21 array :",b21.shape)
print("Shape of c21 array :",c21.shape)
```
