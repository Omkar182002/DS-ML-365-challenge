```python
a1 = np.array([1,2,3,4,5,6])
print(a1.ndim)
print(a1.size)
print(a1.dtype)
print(a1.nbytes)
print(a1.itemsize)
print(a1.shape)
print(a1.T)
```

```python
c8.flags
```

###### Flag	        Meaning
###### OWNDATA	 - ---Own memory or borrowed
###### WRITEABL  - --Can modify or not
###### C_CONTIGUOUS- --	Row-wise storage
###### F_CONTIGUOUS ----	Column-wise storage
###### ALIGNED ----Memory is properly aligned (faster access)
###### WRITEBACKIFCOPY ---Temporary array for copy-back (internal use)

```python
arrr = np.arange(10)
c9 = arrr[::2]        # view → OWNDATA False
c9.flags.writeable = False

c9.flags
```

```python
c9.flags.writeable = True
c9[0]=999
```
